## 📖 내장 함수(Built-in Functions)란 무엇인가?

### 내장 함수의 정의

내장 함수는 Python이 기본으로 제공하는 함수들입니다. **import 없이 바로 사용**할 수 있으며, 우리가 이미 사용해본 `print()`, `type()`, `len()` 같은 함수들이 모두 내장 함수입니다.

```python
# 이미 사용해본 내장 함수들
print("Hello")      # 출력
type(10)            # 타입 확인
len([1, 2, 3])      # 길이
int("10")           # 형변환
```

### 왜 내장 함수를 배워야 할까요?

내장 함수는 **Python 커뮤니티가 수십 년간 최적화**한 함수들입니다. 우리가 직접 코드를 작성하는 것보다:

1. **훨씬 빠릅니다** (C언어로 구현되어 있음)
2. **안전합니다** (엣지 케이스가 모두 처리됨)
3. **읽기 쉽습니다** (다른 개발자도 바로 이해)

예를 들어, 리스트의 최댓값을 찾는다고 생각해보세요:

```python
# 우리가 직접 작성한다면...
numbers = [5, 2, 8, 1, 9]
maximum = numbers[0]
for num in numbers:
    if num > maximum:
        maximum = num
print(maximum)  # 9

# 내장 함수를 사용하면...
print(max(numbers))  # 9
```

한 줄이면 충분합니다! ✨

### FastAPI/실무에서는 어떻게 사용될까요?

API를 개발할 때 데이터를 처리하는 경우가 아주 많습니다:

```python
# FastAPI 엔드포인트 예시
@app.get("/products/stats")
def get_product_stats():
    prices = [15000, 23000, 18000, 32000, 27000]
    
    return {
        "total_products": len(prices),      # 상품 개수
        "max_price": max(prices),           # 최고가
        "min_price": min(prices),           # 최저가
        "total_value": sum(prices),         # 총 가치
        "average_price": sum(prices) / len(prices)  # 평균가
    }
```

이처럼 **통계 정보를 제공하는 API**를 만들 때 내장 함수가 필수입니다!

## 📖 순수 함수(Pure Function) vs 메서드(Method)

오늘 배울 내용 중 가장 중요한 개념입니다. 이 차이를 이해하면 Python 코드가 보이기 시작합니다!

### 메서드(Method)란?

메서드는 **객체에 붙어있는 함수**입니다. 점(`.`)을 찍어서 호출하며, **원본 객체를 직접 변경**하는 경우가 많습니다.

```python
numbers = [3, 1, 4, 1, 5]
numbers.append(9)  # 메서드: 원본을 변경
print(numbers)     # [3, 1, 4, 1, 5, 9]
```

### 순수 함수(Pure Function)란?

순수 함수는 **원본을 변경하지 않고** 새로운 값을 반환합니다. 함수 이름만으로 호출합니다.

```python
numbers = [3, 1, 4, 1, 5]
result = sorted(numbers)  # 순수 함수: 새 리스트 반환
print(numbers)            # [3, 1, 4, 1, 5] (원본 유지!)
print(result)             # [1, 1, 3, 4, 5] (새 리스트)
```

### 왜 이 차이가 중요할까요?

실무에서 **원본 데이터가 예기치 않게 변경**되면 버그가 발생합니다:

```python
# 위험한 코드
original_prices = [10000, 20000, 30000]
sorted_prices = original_prices.sort()  # ⚠️ sort()는 None을 반환!

print(sorted_prices)     # None (원하는 결과가 아님!)
print(original_prices)   # [10000, 20000, 30000] (정렬됨, 원본 변경됨!)

# 안전한 코드
original_prices = [10000, 20000, 30000]
sorted_prices = sorted(original_prices)  # ✅ sorted()는 새 리스트 반환

print(sorted_prices)     # [10000, 20000, 30000]
print(original_prices)   # [10000, 20000, 30000] (원본 유지!)
```

### 핵심 정리

| 구분 | 메서드 (예: sort) | 순수 함수 (예: sorted) |
|------|------------------|----------------------|
| 호출 방식 | `list.sort()` | `sorted(list)` |
| 원본 변경 | ✅ 변경됨 | ❌ 유지됨 |
| 반환값 | `None` | 새로운 리스트 |
| 사용 시기 | 원본을 바꿔도 될 때 | 원본을 보존해야 할 때 |

**실무 팁**: 대부분의 경우 **순수 함수를 사용하는 것이 안전**합니다! 원본 데이터는 최대한 보존하는 것이 좋습니다.

## 1️⃣ 핵심 내장 함수 - 기본편

### len(), max(), min(), sum()

가장 자주 사용하는 4가지 내장 함수를 알아봅시다. 이 함수들은 **데이터 분석의 기본**입니다!

#### 각 함수의 역할

- `len()`: 리스트의 **길이**(요소 개수)를 반환
- `max()`: 리스트에서 **가장 큰 값**을 반환
- `min()`: 리스트에서 **가장 작은 값**을 반환
- `sum()`: 리스트의 **모든 값을 더한 합계**를 반환

#### 실행 전 주목 포인트

- 이 함수들은 모두 **원본을 변경하지 않습니다**
- 단순히 **값을 계산해서 반환**만 합니다
- `sum()`과 `len()`을 조합하면 **평균**을 구할 수 있습니다

````python
# 학생 점수 리스트
scores = [85, 92, 78, 95, 88]

# len(): 리스트의 길이(요소 개수) 반환
# 학생이 몇 명인지 알 수 있습니다
print("학생 수:", len(scores))  # 출력: 학생 수: 5

# max(): 리스트에서 가장 큰 값 반환
# 최고 점수를 찾습니다
print("최고 점수:", max(scores))  # 출력: 최고 점수: 95

# min(): 리스트에서 가장 작은 값 반환
# 최저 점수를 찾습니다
print("최저 점수:", min(scores))  # 출력: 최저 점수: 78

# sum(): 리스트의 모든 값을 더한 합계 반환
# 모든 학생의 점수를 더합니다
print("총점:", sum(scores))  # 출력: 총점: 438

# 평균 계산: sum()과 len()을 조합!
# 총점을 학생 수로 나누면 평균이 나옵니다
average = sum(scores) / len(scores)
print("평균:", average)  # 출력: 평균: 87.6
````

**예상 출력 결과:**
```
학생 수: 5
최고 점수: 95
최저 점수: 78
총점: 438
평균: 87.6
```

**결과 해석:**
- 5명의 학생이 있고
- 가장 잘한 학생은 95점
- 가장 못한 학생은 78점
- 모든 점수를 합치면 438점
- 평균은 87.6점입니다

### 💡 실무 팁

1. **평균 구하기**: `sum() / len()`은 매우 자주 사용하는 패턴입니다
2. **빈 리스트 주의**: 빈 리스트에 `max()`, `min()`, `sum()`을 사용하면 에러가 발생하거나 0을 반환합니다
   ```python
   empty = []
   print(len(empty))   # 0
   print(sum(empty))   # 0
   # print(max(empty)) # ❌ ValueError 발생!
   ```
3. **문자열 리스트**: 문자열 리스트에도 `max()`와 `min()`을 사용할 수 있습니다 (알파벳 순서)
   ```python
   names = ["Alice", "Charlie", "Bob"]
   print(max(names))  # "Charlie"
   print(min(names))  # "Alice"
   ```

## 2️⃣ sorted() - 정렬된 새 리스트 반환

### sorted()는 순수 함수입니다!

`sorted()` 함수는 **원본 리스트를 건드리지 않고**, 정렬된 **새로운 리스트**를 만들어서 반환합니다.

#### 왜 원본을 유지하는 게 중요할까요?

실무에서는 같은 데이터를 **여러 방식으로 보여줘야** 하는 경우가 많습니다:

```python
# API 응답 예시
product_prices = [15000, 23000, 12000, 28000]

# 1. 입력된 순서대로 보여주기
print("등록 순:", product_prices)

# 2. 가격 낮은 순으로 보여주기
print("저가순:", sorted(product_prices))

# 3. 가격 높은 순으로 보여주기
print("고가순:", sorted(product_prices, reverse=True))

# 원본은 그대로!
print("원본:", product_prices)
```

#### 실행 전 주목 포인트

- `sorted(리스트)`: **오름차순** 정렬 (작은 것부터)
- `sorted(리스트, reverse=True)`: **내림차순** 정렬 (큰 것부터)
- 원본 리스트는 **절대 변경되지 않습니다**

````python
# 정렬되지 않은 숫자 리스트
numbers = [5, 2, 8, 1, 9]

# sorted()는 정렬된 **새로운** 리스트를 반환합니다
# 원본은 전혀 건드리지 않습니다!
sorted_numbers = sorted(numbers)

# 원본과 정렬된 리스트를 비교해봅시다
print("원본:", numbers)         # 출력: 원본: [5, 2, 8, 1, 9]
print("정렬:", sorted_numbers)  # 출력: 정렬: [1, 2, 5, 8, 9]

# reverse=True 옵션: 내림차순 정렬 (큰 것부터)
sorted_desc = sorted(numbers, reverse=True)
print("내림차순:", sorted_desc)  # 출력: 내림차순: [9, 8, 5, 2, 1]
````

**예상 출력 결과:**
```
원본: [5, 2, 8, 1, 9]
정렬: [1, 2, 5, 8, 9]
내림차순: [9, 8, 5, 2, 1]
```

**결과 해석:**
- 원본 `numbers`는 `[5, 2, 8, 1, 9]` 그대로 유지됩니다
- `sorted_numbers`는 오름차순으로 정렬된 새 리스트 `[1, 2, 5, 8, 9]`
- `sorted_desc`는 내림차순으로 정렬된 새 리스트 `[9, 8, 5, 2, 1]`

### 💡 실무 활용

FastAPI에서 상품 목록을 여러 방식으로 제공할 때:

```python
@app.get("/products")
def get_products(sort_by: str = "price_asc"):
    products = get_all_products()  # DB에서 가져온 원본
    
    if sort_by == "price_asc":
        # 가격 낮은 순
        return sorted(products, key=lambda x: x["price"])
    elif sort_by == "price_desc":
        # 가격 높은 순
        return sorted(products, key=lambda x: x["price"], reverse=True)
    else:
        # 원본 순서 (등록 순)
        return products
```

원본 데이터를 유지하면서 다양한 정렬 옵션을 제공할 수 있습니다!

## 3️⃣ sort() vs sorted() - 결정적 차이

### 가장 헷갈리는 부분!

Python 초보자가 가장 많이 실수하는 부분입니다. 이름은 비슷하지만 **완전히 다르게 동작**합니다!

### sort() - 메서드 (원본 변경)

- **메서드**이므로 `리스트.sort()` 형태로 호출
- **원본 리스트를 직접 정렬** (원본이 바뀜!)
- **None을 반환** (아무것도 돌려주지 않음)
- 메모리 효율적 (새 리스트를 만들지 않음)

### sorted() - 순수 함수 (원본 유지)

- **함수**이므로 `sorted(리스트)` 형태로 호출
- **원본은 그대로 두고 새 리스트 반환**
- **정렬된 새 리스트를 반환**
- 안전함 (원본 데이터 보존)

### 언제 무엇을 사용할까?

| 상황 | 추천 | 이유 |
|------|------|------|
| 원본이 필요 없을 때 | `sort()` | 메모리 절약 |
| 원본을 유지해야 할 때 | `sorted()` | 안전함 |
| 모르겠을 때 | `sorted()` | 실수 방지 |

**일반적으로는 `sorted()`를 사용하는 것이 더 안전합니다!**

#### 실행 전 주목 포인트

- `sort()`는 `None`을 반환하므로 변수에 할당하면 안 됩니다
- `sorted()`는 반드시 변수에 할당해서 사용해야 합니다

````python
# ============================================
# 1. sort() 메서드 - 원본 변경, None 반환
# ============================================

numbers1 = [5, 2, 8, 1, 9]

# sort()는 원본을 정렬하고 None을 반환합니다
result = numbers1.sort()  # 원본을 정렬, None 반환

print("sort() 후 numbers1:", numbers1)  # 출력: [1, 2, 5, 8, 9] (원본이 바뀜!)
print("sort() 반환값:", result)         # 출력: None (아무것도 안 돌려줌!)

print()  # 빈 줄

# ============================================
# 2. sorted() 함수 - 원본 유지, 새 리스트 반환
# ============================================

numbers2 = [5, 2, 8, 1, 9]

# sorted()는 원본을 유지하고 새 리스트를 반환합니다
result = sorted(numbers2)  # 원본 유지, 새 리스트 반환

print("sorted() 후 numbers2:", numbers2)  # 출력: [5, 2, 8, 1, 9] (원본 그대로!)
print("sorted() 반환값:", result)         # 출력: [1, 2, 5, 8, 9] (새 리스트!)
````

**예상 출력 결과:**
```
sort() 후 numbers1: [1, 2, 5, 8, 9]
sort() 반환값: None

sorted() 후 numbers2: [5, 2, 8, 1, 9]
sorted() 반환값: [1, 2, 5, 8, 9]
```

**결과 분석:**

1. **sort() 메서드**:
   - `numbers1`이 `[1, 2, 5, 8, 9]`로 변경됨 (원본 변경!)
   - 반환값은 `None` (아무것도 돌려주지 않음)
   - ⚠️ `result = numbers1.sort()`라고 쓰면 `result`는 `None`입니다!

2. **sorted() 함수**:
   - `numbers2`는 `[5, 2, 8, 1, 9]` 그대로 유지 (원본 보존!)
   - 반환값은 `[1, 2, 5, 8, 9]` (정렬된 새 리스트)
   - ✅ `result`에 정렬된 리스트가 저장됨

### ⚠️ 흔한 실수

```python
# 실수 1: sort()의 반환값을 사용하려고 함
numbers = [5, 2, 8, 1, 9]
sorted_numbers = numbers.sort()  # ❌ sorted_numbers는 None!
print(sorted_numbers)            # None 출력

# 올바른 방법
numbers = [5, 2, 8, 1, 9]
numbers.sort()          # 원본을 정렬
print(numbers)          # [1, 2, 5, 8, 9]

# 또는
numbers = [5, 2, 8, 1, 9]
sorted_numbers = sorted(numbers)  # ✅ 새 리스트 반환
print(sorted_numbers)             # [1, 2, 5, 8, 9]
```

### 💡 선택 가이드

```python
# 원본이 필요 없다면 → sort() 사용
temp_data = [5, 2, 8, 1, 9]
temp_data.sort()
process(temp_data)  # 정렬된 데이터만 필요

# 원본을 유지해야 한다면 → sorted() 사용
user_input = [5, 2, 8, 1, 9]
sorted_for_display = sorted(user_input)
save_to_db(user_input)           # 원본 저장
show_to_user(sorted_for_display) # 정렬된 것 표시
```

## 4️⃣ 얕은 복사(Shallow Copy) - 리스트 복사의 비밀

### 왜 리스트 복사를 배워야 할까요?

Python에서 리스트를 다룰 때 **가장 많이 하는 실수**가 바로 복사와 관련된 것입니다!

```python
# 이 코드의 문제점은?
original = [1, 2, 3]
copy = original
copy.append(4)

print(original)  # [1, 2, 3, 4] ← 어? 원본도 바뀜!
```

### 참조 복사 vs 얕은 복사

#### 1. 참조 복사 (Reference Copy) - 위험!

`copy = original`처럼 단순 대입하면 **같은 리스트를 가리킵니다**:

```
메모리:
┌─────────────┐
│ [1, 2, 3]   │ ← original과 copy가 둘 다 이것을 가리킴
└─────────────┘
   ↑       ↑
original  copy
```

한쪽을 수정하면 **둘 다 변경**됩니다!

#### 2. 얕은 복사 (Shallow Copy) - 안전!

`copy = original[:]` 또는 `copy = original.copy()`를 사용하면 **새로운 리스트를 생성**합니다:

```
메모리:
┌─────────────┐
│ [1, 2, 3]   │ ← original
└─────────────┘

┌─────────────┐
│ [1, 2, 3]   │ ← copy (별도의 리스트)
└─────────────┘
```

한쪽을 수정해도 **다른 쪽에 영향 없음**!

### 실무에서 왜 중요한가?

```python
# FastAPI 예시
@app.post("/cart/checkout")
def checkout(cart_items: list):
    # 원본 장바구니 유지 (취소할 수도 있으니)
    processing_items = cart_items[:]  # 얕은 복사!
    
    # 처리 중 수정
    for item in processing_items:
        item["status"] = "processing"
    
    # 원본은 그대로 유지됨!
    return {"original": cart_items, "processing": processing_items}
```

#### 실행 전 주목 포인트

- `reference = original`: 같은 리스트를 가리킴 (위험)
- `copy = original[:]`: 새로운 리스트 생성 (안전)
- `copy = original.copy()`: 위와 동일 (명시적)

````python
# ============================================
# 1. 참조 복사 (위험한 방법)
# ============================================

# 원본 리스트
original = [1, 2, 3]

# 단순 대입: 같은 리스트를 가리킴!
reference = original  # original과 reference는 같은 객체!

# reference를 수정하면...
reference.append(4)

# 원본도 함께 변경됩니다! (위험!)
print("original:", original)    # 출력: [1, 2, 3, 4] (함께 변경!)
print("reference:", reference)  # 출력: [1, 2, 3, 4]

print()  # 빈 줄

# ============================================
# 2. 얕은 복사 - 슬라이싱 방법 (안전한 방법)
# ============================================

# 새로운 원본 리스트
original = [1, 2, 3]

# 슬라이싱으로 새로운 리스트 생성!
# [:]는 "처음부터 끝까지 전부"를 의미
copy = original[:]  # 완전히 새로운 리스트 생성!

# copy를 수정하면...
copy.append(4)

# 원본은 변경되지 않습니다! (안전!)
print("original:", original)  # 출력: [1, 2, 3] (변경 안됨!)
print("copy:", copy)          # 출력: [1, 2, 3, 4]
````

**예상 출력 결과:**
```
original: [1, 2, 3, 4]
reference: [1, 2, 3, 4]

original: [1, 2, 3]
copy: [1, 2, 3, 4]
```

**결과 분석:**

1. **참조 복사의 문제**:
   - `reference = original`로 대입하면 같은 리스트를 가리킴
   - `reference.append(4)`를 하면 `original`도 함께 `[1, 2, 3, 4]`로 변경
   - 이것은 **의도하지 않은 버그**를 만들 수 있습니다!

2. **얕은 복사의 장점**:
   - `copy = original[:]`로 새로운 리스트 생성
   - `copy.append(4)`를 해도 `original`은 `[1, 2, 3]` 그대로
   - 원본 데이터를 안전하게 보호할 수 있습니다!

### 💡 복사 방법 총정리

```python
original = [1, 2, 3]

# 방법 1: 슬라이싱 (가장 많이 사용)
copy1 = original[:]

# 방법 2: copy() 메서드 (명시적)
copy2 = original.copy()

# 방법 3: list() 함수
copy3 = list(original)

# 모두 같은 결과: 새로운 리스트 생성
```

### ⚠️ 주의: "얕은" 복사의 한계

"얕은 복사"는 **1단계만 복사**합니다. 중첩 리스트는 조심해야 합니다:

```python
# 중첩 리스트 (리스트 안에 리스트)
original = [[1, 2], [3, 4]]
copy = original[:]  # 얕은 복사

# 안쪽 리스트를 수정하면?
copy[0].append(999)

print(original)  # [[1, 2, 999], [3, 4]] ← 원본도 변경!
print(copy)      # [[1, 2, 999], [3, 4]]

# 중첩 리스트는 copy.deepcopy()를 사용해야 합니다 (나중에 배움)
```

**하지만 걱정하지 마세요!** 우리가 주로 사용하는 단순 리스트에서는 `[:]`로 충분합니다.

## 5️⃣ 실전 예제: 성적 처리 시스템

### 배운 내용을 모두 활용해봅시다!

지금까지 배운 내장 함수들을 조합하면 **실제 성적 처리 시스템**을 만들 수 있습니다.

#### 이 예제에서 사용하는 기술

1. `len()` - 학생 수 계산
2. `max()` - 최고 점수 찾기
3. `min()` - 최저 점수 찾기
4. `sum()` - 총점 계산
5. `sum() / len()` - 평균 계산
6. `sorted()` - 점수 순위 매기기 (원본 유지)

#### 실행 전 주목 포인트

- 모든 통계를 **한 번에 계산**합니다
- `sorted()`를 사용하므로 원본 점수는 유지됩니다
- f-string을 사용해서 깔끔하게 출력합니다
- 이런 패턴은 **FastAPI에서 통계 API를 만들 때** 그대로 사용됩니다!

````python
# 학생 점수 데이터 (실제 시험 점수라고 가정)
student_scores = [88, 92, 76, 95, 84, 90, 78]

# ============================================
# 통계 계산 (내장 함수 활용)
# ============================================

# 1. 총 학생 수
total_students = len(student_scores)

# 2. 최고 점수 (1등 학생의 점수)
highest = max(student_scores)

# 3. 최저 점수 (꼴찌 학생의 점수)
lowest = min(student_scores)

# 4. 총점 (모든 학생 점수의 합)
total = sum(student_scores)

# 5. 평균 점수 (총점 ÷ 학생 수)
average = total / total_students

# 6. 점수 순위 (높은 점수부터 정렬)
# reverse=True로 내림차순 정렬
# 원본은 유지됩니다!
sorted_scores = sorted(student_scores, reverse=True)

# ============================================
# 결과 출력 (보기 좋게 정리)
# ============================================

print("\n=== 성적 처리 결과 ===")
print(f"총 학생 수: {total_students}명")
print(f"최고 점수: {highest}점")
print(f"최저 점수: {lowest}점")
print(f"평균 점수: {average:.1f}점")  # .1f는 소수점 1자리
print(f"점수 순위: {sorted_scores}")
````

**예상 출력 결과:**
```
=== 성적 처리 결과 ===
총 학생 수: 7명
최고 점수: 95점
최저 점수: 76점
평균 점수: 86.1점
점수 순위: [95, 92, 90, 88, 84, 78, 76]
```

**결과 해석:**
- 총 7명의 학생이 시험을 봤습니다
- 가장 잘한 학생은 95점
- 가장 못한 학생은 76점
- 반 평균은 86.1점 (상당히 높은 편!)
- 점수 순위는 95점부터 76점까지 내림차순으로 정렬되었습니다

1. **내장 함수의 개념**:
   - Python이 기본 제공하는 함수들
   - import 없이 바로 사용 가능
   - C로 구현되어 있어 빠르고 안전함

2. **핵심 내장 함수 4종 세트**:
   - `len()`: 리스트 길이
   - `max()`: 최댓값
   - `min()`: 최솟값
   - `sum()`: 합계
   - 조합: `sum() / len()` = 평균

3. **sorted() - 순수 함수**:
   - 원본을 유지하고 새 리스트 반환
   - `sorted(리스트)`: 오름차순

4. **sort() vs sorted() 차이**:
   - `sort()`: 메서드, 원본 변경, None 반환
   - `sorted()`: 함수, 원본 유지, 새 리스트 반환
   - **헷갈릴 땐 sorted()를 사용하세요!**

5. **얕은 복사**:
   - `copy = original`: 참조 복사 (위험)
   - `copy = original[:]`: 얕은 복사 (안전)
   - 원본을 보호하고 싶을 때 필수!

#### ⭐ 내장 함수는 항상 순수 함수
