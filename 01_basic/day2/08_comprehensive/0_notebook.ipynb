---

### 핵심 개념

이 문제에서 연습하는 것:

1. ✅ **딕셔너리 리스트**: `[{}, {}, {}]`
2. ✅ **리스트 순회**: `for product in products:`
3. ✅ **중첩 구조 접근**: `products[0]["name"]`
4. ✅ **조건부 필터링**: `if product["stock"] < 10:`
5. ✅ **합계 계산**: `total += price * quantity`

---

### 💡 팁

**딕셔너리의 리스트 구조**를 이해하세요:

```python
products = [
    {"name": "노트북", "stock": 15},  # 첫 번째 상품
    {"name": "마우스", "stock": 5},   # 두 번째 상품
]

# 접근 방법
products[0]          # 첫 번째 딕셔너리 전체
products[0]["name"]  # 첫 번째 상품의 이름
```

---

### ⚠️ 자주 하는 실수

```python
# ❌ 잘못된 방법
for product in products:
    print(products["name"])  # products가 아니라 product!

# ✅ 올바른 방법
for product in products:
    print(product["name"])   # 현재 아이템은 product
```

---

### 핵심 개념

이 문제에서 연습하는 것:

1. ✅ **데이터 그룹화**: 고객별 합계 계산
2. ✅ **집합 활용**: 중복 제거 (상품 종류)
3. ✅ **통계 계산**: 합계, 평균, 최댓값
4. ✅ **딕셔너리 업데이트**: 기존 값에 더하기
5. ✅ **최댓값 찾기**: 반복문으로 비교

---

### 💡 핵심 패턴

**고객별 합계를 구하는 패턴** (자주 사용됨!):

```python
customer_total = {}

for order in orders:
    customer = order["customer"]
    amount = order["price"] * order["quantity"]
    
    # 키가 없으면 0으로 초기화
    if customer not in customer_total:
        customer_total[customer] = 0
    
    # 누적
    customer_total[customer] += amount
```

---

### 🎯 도전 포인트

- 같은 고객의 여러 주문을 **합산**해야 합니다
- 집합을 사용하여 **중복 없는 상품 목록**을 만듭니다
- 딕셔너리에서 **최댓값**을 가진 키를 찾아야 합니다
