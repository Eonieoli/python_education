## 📖 튜플(Tuple)이란 무엇인가?

### 튜플의 정의

튜플은 **여러 개의 값을 순서대로 저장**할 수 있는 Python의 자료구조입니다. 리스트와 매우 비슷하지만, **한 번 만들면 수정할 수 없다**는 중요한 차이점이 있습니다.

```python
# 리스트는 수정 가능
my_list = [1, 2, 3]
my_list[0] = 999  # OK!

# 튜플은 수정 불가능
my_tuple = (1, 2, 3)
# my_tuple[0] = 999  # 에러 발생!
```

### 왜 튜플이 필요할까요?

"수정할 수 없다면 왜 사용하나요?"라고 생각할 수 있습니다. 하지만 바로 그 **불변성(Immutability)** 때문에 튜플이 유용한 경우가 많습니다!

**1. 데이터 보호**  
좌표, RGB 색상값처럼 **절대 바뀌면 안 되는 데이터**는 튜플로 저장하면 실수로 수정하는 것을 방지할 수 있습니다.

```python
# 화면의 한 점 (x, y)
position = (100, 200)
# position[0] = 999  # 에러! 실수로 수정 불가
```

**2. 함수에서 여러 값 반환**  
함수가 여러 개의 값을 반환할 때 튜플을 사용합니다. (Day 4에서 자세히 배웁니다)

```python
def get_user_info():
    name = "홍길동"
    age = 25
    return name, age  # 튜플로 반환! (name, age)

user_name, user_age = get_user_info()  # 언패킹!
```

**3. 성능 최적화**  
튜플은 리스트보다 메모리를 적게 사용하고 속도도 빠릅니다. 수정이 필요 없는 데이터라면 튜플이 더 효율적입니다.

### FastAPI에서 튜플은 어떻게 쓰일까요?

FastAPI 개발에서 튜플은 주로 **여러 값을 묶어서 전달**할 때 사용됩니다.

```python
# API에서 좌표 데이터 처리
def calculate_distance(point1, point2):
    x1, y1 = point1  # 튜플 언패킹!
    x2, y2 = point2
    # 거리 계산...
```

하지만 더 중요한 것은 **튜플 언패킹(Tuple Unpacking)** 개념입니다! 이것은 FastAPI의 함수 매개변수를 다룰 때 매우 자주 사용되는 패턴이므로, 오늘 확실히 익혀두세요. 🎯

## 📖 집합(Set)이란 무엇인가?

### 집합의 정의

집합은 **중복을 허용하지 않는** 순서가 없는 자료구조입니다. 수학의 집합 개념과 동일합니다.

```python
# 리스트는 중복 허용
my_list = [1, 2, 2, 3, 3, 3]
print(my_list)  # [1, 2, 2, 3, 3, 3]

# 집합은 중복 자동 제거!
my_set = {1, 2, 2, 3, 3, 3}
print(my_set)  # {1, 2, 3}
```

### 왜 집합이 필요할까요?

**1. 중복 제거**  
데이터에서 중복을 제거할 때 집합만큼 간단한 방법이 없습니다!

```python
# 설문 응답에서 중복 제거
responses = ["파이썬", "자바", "파이썬", "C++", "자바"]
unique_languages = set(responses)
# 결과: {"파이썬", "자바", "C++"}
```

**2. 빠른 검색 (O(1))**  
집합은 `in` 연산자로 검색할 때 **매우 빠릅니다**. 리스트보다 수백~수천 배 빠를 수 있습니다!

```python
# 10,000개의 데이터에서 검색
allowed_users_list = ["user1", "user2", ..., "user10000"]  # 느림!
allowed_users_set = {"user1", "user2", ..., "user10000"}  # 빠름!

if "admin" in allowed_users_set:  # 거의 즉시 확인!
    print("접근 허용")
```

### FastAPI에서 집합은 어떻게 쓰일까요?

**1. 권한 관리**  
허용된 사용자, 역할, 권한을 집합으로 관리하면 빠르게 확인할 수 있습니다.

```python
# 관리자 권한이 있는 사용자들
admin_users = {"alice", "bob", "charlie"}

# API 엔드포인트에서 권한 확인
if current_user in admin_users:
    # 관리자 기능 실행
    pass
```

**2. 중복 데이터 정리**  
API 요청에서 받은 태그, 카테고리 등의 중복을 제거할 때 유용합니다.

```python
# 사용자가 입력한 태그 (중복 포함)
tags = ["python", "coding", "python", "web", "coding"]

# 중복 제거 후 저장
unique_tags = set(tags)  # {"python", "coding", "web"}
```

### 주의: 집합은 순서가 없습니다!

집합은 순서를 보장하지 않으므로, 인덱싱이 불가능합니다.

```python
my_set = {1, 2, 3}
# my_set[0]  # 에러! 집합은 인덱싱 불가
```

순서가 필요하면 리스트를 사용하고, 중복 제거나 빠른 검색이 필요하면 집합을 사용하세요!

## 1️⃣ 튜플 기본 - 생성과 인덱싱

### 튜플은 소괄호 ()로 만듭니다

튜플을 만드는 방법은 매우 간단합니다. 소괄호 `()` 안에 값들을 쉼표로 구분해서 넣으면 됩니다.

리스트는 대괄호 `[]`, 튜플은 소괄호 `()`라고 기억하세요!

### 인덱싱은 리스트와 동일합니다

튜플도 리스트처럼 인덱스로 요소에 접근할 수 있습니다. 0부터 시작하고, 음수 인덱스도 사용 가능합니다.

````python
# 튜플은 괄호()로 생성합니다 (수정 불가능!)
point = (10, 20)
print(point)  # 출력: (10, 20)
print(type(point))  # 출력: <class 'tuple'>

# 인덱싱은 리스트와 동일합니다
print(point[0])  # 출력: 10 (첫 번째 요소)
print(point[1])  # 출력: 20 (두 번째 요소)

# 여러 타입을 섞을 수 있습니다
user = ("홍길동", 25, "서울")
print(user)  # 출력: ('홍길동', 25, '서울')
````

**실행 결과:**
```
(10, 20)
<class 'tuple'>
10
20
('홍길동', 25, '서울')
```

**결과 해석:**
- `point`는 튜플 타입입니다 (`<class 'tuple'>`)
- 인덱싱으로 개별 요소에 접근할 수 있습니다
- 튜플 안에는 문자열, 숫자 등 다양한 타입을 함께 저장할 수 있습니다

### 💡 알아두기

- 튜플은 리스트와 달리 **소괄호 ()**를 사용합니다
- 요소가 1개인 튜플을 만들 때는 쉼표를 꼭 붙여야 합니다: `(1,)`
- 괄호 없이도 쉼표만으로 튜플을 만들 수 있습니다: `1, 2, 3`은 `(1, 2, 3)`과 같습니다

## 2️⃣ 튜플 언패킹 (핵심!) ⭐⭐⭐

### 언패킹이란?

**언패킹(Unpacking)**은 튜플(또는 리스트)의 값들을 **한 번에 여러 변수에 할당**하는 것을 말합니다. Python의 매우 강력한 기능이며, FastAPI를 배울 때 자주 사용하게 됩니다!

```python
# 언패킹 없이 (번거로움)
point = (100, 200)
x = point[0]
y = point[1]

# 언패킹 사용 (간단!)
point = (100, 200)
x, y = point  # 한 줄로 끝!
```

### 왜 언패킹이 중요한가요?

1. **코드가 간결해집니다**: 여러 줄의 코드를 한 줄로 줄일 수 있습니다
2. **가독성이 좋아집니다**: 의도가 명확하게 드러납니다
3. **FastAPI에서 필수적입니다**: 함수의 여러 반환값을 받을 때 항상 사용합니다

### 언패킹의 다양한 활용

언패킹은 튜플뿐만 아니라 **리스트, 문자열 등 순서가 있는 모든 자료구조**에 사용할 수 있습니다!

````python
# 튜플의 값을 여러 변수에 한 번에 할당할 수 있습니다
coordinates = (100, 200)
x, y = coordinates  # 언패킹!
print(f"x={x}, y={y}")  # 출력: x=100, y=200

# 괄호 없이도 가능합니다
name, age, city = "김철수", 30, "부산"
print(f"{name}님은 {age}살이고 {city}에 삽니다")
# 출력: 김철수님은 30살이고 부산에 삽니다

# RGB 색상값 언패킹
color = (255, 128, 0)  # 주황색
r, g, b = color
print(f"R={r}, G={g}, B={b}")  # 출력: R=255, G=128, B=0
````

**실행 결과:**
```
x=100, y=200
김철수님은 30살이고 부산에 삽니다
R=255, G=128, B=0
```

**결과 해석:**
- `coordinates`의 두 값이 자동으로 `x`, `y`에 할당되었습니다
- 괄호 없이 쉼표로 구분된 값들도 자동으로 튜플이 되어 언패킹됩니다
- RGB 색상값처럼 고정된 개수의 값을 다룰 때 매우 유용합니다

### 💡 언패킹 활용 팁

**1. 함수에서 여러 값 반환 (Day 4에서 자세히 배웁니다)**
```python
def get_min_max(numbers):
    return min(numbers), max(numbers)  # 튜플로 반환

minimum, maximum = get_min_max([1, 5, 3, 9, 2])  # 언패킹!
```

**2. 설정 값 관리**
```python
# 화면 크기 설정
screen_size = (1920, 1080)
width, height = screen_size
```

**3. 좌표 계산**
```python
start = (0, 0)
end = (100, 50)
start_x, start_y = start
end_x, end_y = end
```

### ⚠️ 주의사항

언패킹할 때 **변수 개수와 값의 개수가 반드시 같아야** 합니다!

```python
# 에러 발생!
x, y = (1, 2, 3)  # ValueError: too many values to unpack
x, y, z = (1, 2)  # ValueError: not enough values to unpack
```

## 3️⃣ 스와핑 (값 교환) - 언패킹의 멋진 활용

### 스와핑이란?

**스와핑(Swapping)**은 두 변수의 값을 서로 바꾸는 것을 말합니다. 다른 프로그래밍 언어에서는 임시 변수를 사용해야 하지만, Python에서는 **언패킹으로 한 줄에 해결**할 수 있습니다!

### 다른 언어에서는 이렇게 합니다 (복잡함)

```python
# Java, C++ 등의 방식
a = 10
b = 20

# 임시 변수 필요!
temp = a
a = b
b = temp
```

### Python에서는 이렇게 합니다 (간단!)

```python
a = 10
b = 20

# 한 줄로 끝!
a, b = b, a
```

이것이 바로 Python의 우아함입니다! ✨

````python
# 파이썬에서는 언패킹으로 값을 쉽게 교환할 수 있습니다!
a = 10
b = 20
print(f"교환 전: a={a}, b={b}")  # 출력: 교환 전: a=10, b=20

# 스와핑 (한 줄로!)
a, b = b, a
print(f"교환 후: a={a}, b={b}")  # 출력: 교환 후: a=20, b=10
````

**실행 결과:**
```
교환 전: a=10, b=20
교환 후: a=20, b=10
```

**어떻게 동작하나요?**

```python
a, b = b, a
```

이 코드는 다음과 같이 동작합니다:
1. 먼저 우변 `b, a`가 튜플 `(20, 10)`으로 평가됩니다
2. 그 다음 좌변의 `a, b`에 언패킹됩니다
3. 결과적으로 `a = 20`, `b = 10`이 됩니다

### 💡 스와핑 활용 사례

**1. 정렬 알고리즘**  
버블 정렬, 선택 정렬 등에서 값을 교환할 때 사용합니다.

```python
numbers = [5, 2, 8, 1, 9]
# 첫 번째와 두 번째 값 교환
numbers[0], numbers[1] = numbers[1], numbers[0]
```

**2. 변수 초기화**  
시작값과 끝값을 바꿀 때 유용합니다.

```python
start = 100
end = 0
# 범위를 반대로
start, end = end, start
```

**3. 게임 개발**  
플레이어 순서를 바꿀 때 사용합니다.

```python
player1 = "Alice"
player2 = "Bob"
# 턴 교환
player1, player2 = player2, player1
```

## 4️⃣ 튜플의 불변성 (수정 불가!)

### 튜플은 왜 수정할 수 없나요?

튜플의 가장 큰 특징은 **불변성(Immutability)**입니다. 한 번 만들어진 튜플은 절대 수정할 수 없습니다.

이것은 버그가 아니라 **의도된 기능**입니다! 불변성 덕분에:
- 데이터가 실수로 변경되는 것을 방지할 수 있습니다
- 딕셔너리의 키로 사용할 수 있습니다 (리스트는 불가능!)
- 메모리를 더 효율적으로 사용합니다
- 여러 곳에서 동시에 사용해도 안전합니다

### 수정이 필요하면 어떻게 하나요?

튜플을 "수정"할 수는 없지만, **새로운 튜플을 만들** 수는 있습니다!

````python
# 튜플은 생성 후 수정할 수 없습니다
position = (50, 100)
# position[0] = 999  # 에러! 튜플은 수정 불가
# TypeError: 'tuple' object does not support item assignment

# 새로운 튜플을 만들어야 합니다
position = (999, 100)  # 가능! (새로운 튜플 생성)
print(position)  # 출력: (999, 100)
````

**실행 결과:**
```
(999, 100)
```

**결과 해석:**
- 기존 튜플을 수정하는 것은 불가능합니다 (주석 처리된 코드)
- 하지만 변수에 **새로운 튜플**을 할당하는 것은 가능합니다
- 이것은 수정이 아니라 "교체"입니다!

### 💡 리스트 vs 튜플 선택 기준

**리스트를 사용하세요:**
- 데이터를 추가/삭제/수정해야 할 때
- 학생 목록, 장바구니처럼 계속 변하는 데이터

```python
shopping_cart = ["사과", "바나나"]  # 리스트 사용
shopping_cart.append("우유")  # 추가 가능!
```

**튜플을 사용하세요:**
- 데이터가 고정되어 있을 때
- 좌표, RGB 색상처럼 절대 바뀌면 안 되는 데이터
- 함수에서 여러 값을 반환할 때

```python
screen_resolution = (1920, 1080)  # 튜플 사용
# 화면 해상도는 바뀌면 안 됨!
```

### ⚠️ 헷갈리는 포인트

튜플 안에 리스트가 있으면, **리스트는 수정 가능**합니다!

```python
data = ([1, 2, 3], "고정값")
data[0].append(4)  # 가능! 리스트는 수정 가능
# data[1] = "변경"  # 에러! 튜플 자체는 수정 불가
```

## 5️⃣ 집합 기본 - 생성과 중복 제거

### 집합은 set()으로 만듭니다

집합을 만드는 방법은 두 가지가 있습니다:

1. **중괄호 `{}`** 사용 (단, 빈 집합은 `set()`으로)
2. **`set()` 함수** 사용

```python
# 방법 1: 중괄호
my_set = {1, 2, 3}

# 방법 2: set() 함수
my_set = set([1, 2, 3])

# 주의! 빈 집합은 반드시 set()으로
empty_set = set()  # 올바름
# empty_dict = {}  # 이것은 딕셔너리!
```

### 집합의 슈퍼파워: 자동 중복 제거!

집합에 중복된 값을 넣으면 **자동으로 하나만** 남습니다. 이것이 집합의 가장 강력한 기능입니다!

````python
# 집합은 set()으로 생성합니다 (중복 자동 제거!)
tags = set(["파이썬", "코딩", "파이썬", "개발", "코딩"])
print(tags)  # 출력: {'파이썬', '코딩', '개발'} (순서 무작위)

# 리스트에서 중복 제거할 때 유용합니다
numbers = [1, 2, 2, 3, 3, 3, 4, 4, 4, 4]
unique_numbers = set(numbers)
print(unique_numbers)  # 출력: {1, 2, 3, 4}
````

**실행 결과:**
```
{'파이썬', '코딩', '개발'}
{1, 2, 3, 4}
```

**결과 해석:**
- 원본 리스트에 있던 "파이썬"과 "코딩"의 중복이 자동으로 제거되었습니다
- `numbers` 리스트의 중복된 숫자들이 하나씩만 남았습니다
- 집합은 순서가 없으므로, 출력 순서는 입력 순서와 다를 수 있습니다

## 6️⃣ in 연산자로 빠른 검색

### in 연산자란?

`in` 연산자는 "~에 포함되어 있는가?"를 확인하는 연산자입니다. 리스트, 튜플, 집합 모두에서 사용할 수 있지만, **집합에서 가장 빠릅니다**!

### 왜 집합의 in이 빠른가요?

**기술적 배경:**
- 리스트/튜플: 처음부터 끝까지 하나씩 확인 → **O(n)** 시간 복잡도
- 집합: 해시 테이블 사용 → **O(1)** 시간 복잡도 (거의 즉시!)

```python
# 1,000개 데이터에서 검색
my_list = [1, 2, 3, ..., 1000]  # 평균 500번 확인
my_set = {1, 2, 3, ..., 1000}   # 1번만 확인!
```

데이터가 많을수록 차이가 극명합니다. 10,000개 데이터라면 집합이 **수천 배 빠를 수** 있습니다!

### 실무에서는 언제 사용하나요?

**권한 확인, 사용자 검증, 필터링** 등 "~가 있는지 확인"하는 모든 경우에 사용합니다.

````python
# 집합은 in 연산자로 매우 빠르게 검색할 수 있습니다!
allowed_users = {"admin", "user1", "user2"}

# 사용자 확인
if "admin" in allowed_users:
    print("관리자입니다")  # 출력: 관리자입니다

if "hacker" in allowed_users:
    print("허가된 사용자입니다")
else:
    print("접근 거부!")  # 출력: 접근 거부!
````

**실행 결과:**
```
관리자입니다
접근 거부!
```

**결과 해석:**
- `"admin"`은 `allowed_users`에 있으므로 첫 번째 조건이 True
- `"hacker"`는 없으므로 두 번째 조건이 False → else 블록 실행

## 7️⃣ 집합 메서드 - add(), remove()

### 집합은 수정 가능합니다!

튜플과 달리 집합은 **수정 가능(Mutable)**합니다. 요소를 추가하거나 제거할 수 있습니다.

### 주요 메서드

- **`add(값)`**: 집합에 요소를 추가합니다 (중복은 무시됨)
- **`remove(값)`**: 집합에서 요소를 제거합니다 (없으면 에러)
- **`discard(값)`**: 집합에서 요소를 제거합니다 (없어도 에러 없음)

### add vs append vs insert

```python
# 리스트: append() - 끝에 추가
my_list = [1, 2, 3]
my_list.append(4)  # [1, 2, 3, 4]

# 집합: add() - 순서 없음, 중복 무시
my_set = {1, 2, 3}
my_set.add(4)  # {1, 2, 3, 4}
my_set.add(2)  # {1, 2, 3, 4} (중복 무시!)
```

````python
# 빈 집합 생성
categories = set()

# add()로 요소 추가
categories.add("IT")
categories.add("과학")
categories.add("IT")  # 중복은 무시됨
print(categories)  # 출력: {'IT', '과학'}

# remove()로 요소 제거
categories.remove("과학")
print(categories)  # 출력: {'IT'}
````

**실행 결과:**
```
{'IT', '과학'}
{'IT'}
```

**결과 해석:**
- `add("IT")`를 두 번 호출했지만, 중복은 자동으로 무시되어 "IT"는 하나만 존재합니다
- `remove("과학")`으로 "과학"을 제거했습니다

### 💡 remove vs discard

**remove()**: 없는 요소를 제거하려고 하면 **에러 발생**
```python
my_set = {1, 2, 3}
my_set.remove(999)  # KeyError 발생!
```

**discard()**: 없는 요소를 제거해도 **에러 없음**
```python
my_set = {1, 2, 3}
my_set.discard(999)  # 에러 없음, 조용히 무시
```

**언제 뭘 쓸까요?**
- 확실히 있는 요소 → `remove()` 사용 (없으면 에러로 알려줌)
- 있을 수도 없을 수도 → `discard()` 사용 (안전함)

### 🎯 실전 활용 예제

**동적으로 권한 관리**
```python

online_users.add("bob")

# 사용자 로그아웃

    print("bob은 현재 온라인입니다")
```

## 8️⃣ 실전 예제: 중복 태그 제거 시스템

### 실무 시나리오

블로그나 쇼핑몰에서 게시글에 태그를 달 때, 사용자가 실수로 같은 태그를 여러 번 입력하는 경우가 있습니다. 이때 집합을 사용하면 **자동으로 중복을 제거**할 수 있습니다!

### FastAPI에서는?

```python
from fastapi import FastAPI
from pydantic import BaseModel

class BlogPost(BaseModel):
    title: str
    content: str
    tags: list[str]  # 사용자 입력 (중복 가능)

@app.post("/posts")

    # 태그 중복 제거!
    unique_tags = list(set(post.tags))
    # DB에 저장...
    return {"tags": unique_tags}
```

### 지금 배운 내용의 종합 활용

아래 예제는 오늘 배운 **집합의 모든 기능**을 보여줍니다:
- 중복 제거
- 길이 확인 (`len()`)
- 데이터 비교 (원본 vs 정리된 데이터)

````python
# 블로그 게시글 태그 (중복 있음)
post_tags = ["파이썬", "웹개발", "FastAPI", "파이썬", "API", "웹개발"]

# 중복 제거
unique_tags = set(post_tags)
print(f"원본: {post_tags}")
print(f"중복 제거: {unique_tags}")
# 출력: 원본: ['파이썬', '웹개발', 'FastAPI', '파이썬', 'API', '웹개발']
# 출력: 중복 제거: {'파이썬', '웹개발', 'FastAPI', 'API'}

# 태그 개수
print(f"원본 태그 개수: {len(post_tags)}")  # 출력: 원본 태그 개수: 6
print(f"고유 태그 개수: {len(unique_tags)}")  # 출력: 고유 태그 개수: 4
````

**실행 결과:**
```
원본: ['파이썬', '웹개발', 'FastAPI', '파이썬', 'API', '웹개발']
중복 제거: {'파이썬', '웹개발', 'FastAPI', 'API'}
원본 태그 개수: 6
고유 태그 개수: 4
```

**결과 해석:**
- 원본에는 "파이썬" 2개, "웹개발" 2개가 있었지만, 집합으로 변환하면서 각각 1개씩만 남았습니다
- 원본 태그는 6개였지만, 고유 태그는 4개입니다
- 이렇게 중복을 제거하면 DB에 저장할 때도 효율적이고, 사용자 경험도 좋아집니다

### 💡 실전 개선 버전

실무에서는 중복 제거 후 **다시 리스트로 변환**하는 경우가 많습니다.

```python
# 중복 제거 + 정렬 + 리스트 변환
post_tags = ["파이썬", "웹개발", "FastAPI", "파이썬", "API", "웹개발"]
clean_tags = sorted(set(post_tags))  # 집합 → 정렬 → 리스트
print(clean_tags)  # ['API', 'FastAPI', '웹개발', '파이썬']

# API 응답
return {"tags": clean_tags}
```

왜 리스트로 변환할까요?
- JSON은 집합을 지원하지 않습니다 (리스트만 가능)
- 정렬된 순서로 보여주기 위해
- DB에 저장할 때 리스트 형태가 필요한 경우
