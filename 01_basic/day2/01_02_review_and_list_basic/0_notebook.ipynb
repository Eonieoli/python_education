# 📖 리스트(List)란 무엇인가?

## 리스트의 정의

리스트는 **여러 개의 값을 순서대로 저장**할 수 있는 Python의 가장 강력하고 유연한 자료구조입니다. 프로그래밍에서 가장 자주 사용되는 데이터 타입 중 하나죠!

## 왜 리스트가 필요할까요?

잠깐, 리스트 없이 프로그래밍을 한다고 상상해볼까요?

학생 10명의 점수를 저장해야 한다면 이렇게 해야 할까요?

```python
score1 = 85
score2 = 90
score3 = 78
score4 = 92
score5 = 88
score6 = 76
score7 = 95
score8 = 89
score9 = 84
score10 = 91
```

이렇게 하면 문제가 뭘까요?
- 변수가 너무 많아서 관리가 어렵습니다 😫
- 평균을 구하려면 10개를 다 더해야 합니다
- 학생이 100명이라면? 변수 100개를 만들 건가요?
- 새로운 학생이 추가되면? 코드를 다시 작성해야 합니다

**리스트를 사용하면 이렇게 간단해집니다:**

```python
scores = [85, 90, 78, 92, 88, 76, 95, 89, 84, 91]
```

단 한 줄로 모든 데이터를 관리할 수 있습니다! ✨

## 실생활 비유

리스트는 **서랍장**이나 **책장**과 비슷합니다:

- **서랍장**: 여러 개의 칸(요소)이 있고, 각 칸에는 번호(인덱스)가 붙어있습니다
- **첫 번째 칸**: 인덱스 0번
- **두 번째 칸**: 인덱스 1번
- **순서가 중요**: 각 칸의 위치가 정해져 있습니다

## FastAPI에서 리스트는 어떻게 쓰일까요?

FastAPI로 웹 API를 만들 때, 리스트는 **필수적**입니다!

### 예시 1: 여러 사용자 정보 반환

```python
@app.get("/users")
def get_users():
    # 여러 사용자를 리스트로 반환!
    return [
        {"name": "Alice", "age": 25},
        {"name": "Bob", "age": 30},
        {"name": "Charlie", "age": 35}
    ]
```

### 예시 2: 상품 목록 API

```python
@app.get("/products")
def get_products():
    # 쇼핑몰 상품들을 리스트로!
    return [
        {"id": 1, "name": "노트북", "price": 1200000},
        {"id": 2, "name": "마우스", "price": 30000},
        {"id": 3, "name": "키보드", "price": 80000}
    ]
```

보시다시피, **실무에서는 리스트 없이는 API를 만들 수 없습니다!**

# 🔍 인덱싱과 슬라이싱 개념

## 인덱스(Index)란?

리스트의 각 요소는 **위치를 나타내는 번호**를 가지고 있습니다. 이것을 **인덱스**라고 합니다.

### 중요한 규칙: 인덱스는 0부터 시작!

```
fruits = ["사과", "바나나", "체리", "포도"]
          ↑        ↑        ↑       ↑
        인덱스 0    1        2       3
```

처음 프로그래밍을 배우는 분들이 가장 헷갈려하는 부분이 바로 이겁니다!
- 첫 번째 요소의 인덱스는 **0**
- 두 번째 요소의 인덱스는 **1**
- 세 번째 요소의 인덱스는 **2**

왜 0부터 시작할까요? 컴퓨터 메모리의 시작 위치가 0이기 때문입니다. 처음에는 어색하지만 곧 익숙해질 거예요!

### 음수 인덱스: 뒤에서부터 접근하기

Python의 멋진 기능 중 하나는 **음수 인덱스**입니다!

```
fruits = ["사과", "바나나", "체리", "포도"]
          ↑        ↑        ↑       ↑
        -4       -3       -2      -1
```

- `fruits[-1]` → 마지막 요소 ("포도")
- `fruits[-2]` → 뒤에서 두 번째 ("체리")

마지막 요소에 접근할 때 리스트 길이를 몰라도 `-1`로 간단히 접근할 수 있습니다!

## 슬라이싱(Slicing)이란?

슬라이싱은 리스트의 **일부분**을 잘라내는 기능입니다. 마치 빵을 자르듯이 리스트를 원하는 부분만 추출할 수 있어요!

### 기본 문법: `[시작:끝]`

```python
numbers = [10, 20, 30, 40, 50]
numbers[1:4]  # 인덱스 1부터 3까지 → [20, 30, 40]
```

**중요한 규칙:**
- 시작 인덱스는 **포함**됩니다
- 끝 인덱스는 **포함되지 않습니다** (끝 - 1까지)

### 다양한 슬라이싱 패턴

```python
numbers = [10, 20, 30, 40, 50]

numbers[2:]    # 인덱스 2부터 끝까지 → [30, 40, 50]
numbers[:3]    # 처음부터 인덱스 2까지 → [10, 20, 30]
numbers[::2]   # 처음부터 끝까지 2칸씩 → [10, 30, 50]
numbers[::-1]  # 역순! → [50, 40, 30, 20, 10]
```

### 문자열과 똑같아요!

기억나시나요? Day 1에서 배운 문자열 인덱싱과 슬라이싱!

```python
# 문자열
text = "Hello"
text[0]     # 'H'
text[1:4]   # 'ell'

# 리스트 (똑같은 방식!)
items = ["H", "e", "l", "l", "o"]
items[0]    # 'H'
items[1:4]  # ['e', 'l', 'l']
```

문자열과 리스트 모두 **순서가 있는 자료형**이기 때문에 같은 방식으로 접근할 수 있습니다!

# 📏 len() 함수의 역할

## len()이란?

`len()`은 **length(길이)**의 줄임말로, 리스트에 몇 개의 요소가 있는지 알려주는 내장 함수입니다.

```python
students = ["김철수", "이영희", "박민수"]
len(students)  # 3
```

## 왜 중요할까요?

### 1. 데이터 개수 확인

```python
# 사용자가 몇 명인지 확인
users = [...]
print(f"총 {len(users)}명의 사용자가 있습니다")
```

### 2. 평균 계산

```python
scores = [85, 90, 78, 92, 88]
average = sum(scores) / len(scores)  # 합계를 개수로 나누기
```

### 3. 반복문과의 관계 (미리보기)

내일 배울 `for`문에서 len()은 매우 중요합니다:

```python
# 곧 배울 내용!
for i in range(len(students)):
    print(f"{i+1}번째 학생: {students[i]}")
```

### 4. FastAPI에서의 활용

```python
@app.get("/users")
def get_users():
    users = get_all_users()  # 데이터베이스에서 사용자 가져오기
    return {
        "total": len(users),  # 전체 사용자 수
        "users": users
    }
```

## 빈 리스트의 길이는?

```python
empty = []
len(empty)  # 0
```

빈 리스트의 길이는 당연히 0입니다. 이것을 이용해서 리스트가 비어있는지 확인할 수 있습니다:

```python
if len(my_list) == 0:
    print("리스트가 비어있습니다")
```

# ➕ 리스트 연산

리스트도 숫자처럼 **연산**을 할 수 있습니다! 하지만 의미는 조금 다릅니다.

## + 연산자: 리스트 합치기 (Concatenation)

`+` 연산자는 두 리스트를 **하나로 합쳐**줍니다:

```python
list1 = [1, 2, 3]
list2 = [4, 5, 6]
combined = list1 + list2
# 결과: [1, 2, 3, 4, 5, 6]
```

### 실무 예시: 데이터 병합

```python
# 상반기 매출
first_half = [100, 120, 110, 130, 125, 140]

# 하반기 매출
second_half = [150, 145, 160, 155, 170, 180]

# 연간 전체 매출
yearly = first_half + second_half
```

## * 연산자: 리스트 반복 (Repetition)

`*` 연산자는 리스트를 **여러 번 반복**해줍니다:

```python
[0] * 5
# 결과: [0, 0, 0, 0, 0]

["★"] * 3
# 결과: ['★', '★', '★']
```

### 언제 사용할까요?

#### 1. 초기값 설정

```python
# 10개의 0으로 채워진 리스트 만들기
scores = [0] * 10
```

#### 2. 구분선 만들기

```python
separator = ["-"] * 20
print("".join(separator))  # "--------------------"
```

#### 3. 패턴 생성

```python
# 체스판 패턴
row = ["□", "■"] * 4
# 결과: ['□', '■', '□', '■', '□', '■', '□', '■']
```

## ⚠️ 주의사항

리스트 연산은 **새로운 리스트**를 만듭니다. 원본은 변하지 않습니다:

```python
list1 = [1, 2, 3]
list2 = [4, 5, 6]
combined = list1 + list2

print(list1)     # [1, 2, 3] - 원본 그대로!
print(list2)     # [4, 5, 6] - 원본 그대로!
print(combined)  # [1, 2, 3, 4, 5, 6] - 새로운 리스트
```

## 2. 인덱싱 - 특정 요소 접근하기

### 대괄호 `[인덱스]`로 특정 위치의 값을 가져올 수 있습니다

리스트의 각 요소는 **인덱스**라는 번호를 가지고 있습니다. 이 번호로 원하는 요소에 접근할 수 있어요!

**기억하세요**: 인덱스는 **0부터 시작**합니다!

````python
# 과일 리스트 만들기
fruits = ["사과", "바나나", "체리", "포도"]

# 양수 인덱스로 접근 (앞에서부터)
print(fruits[0])  # 출력: 사과 (첫 번째 요소, 인덱스 0)
print(fruits[1])  # 출력: 바나나 (두 번째 요소, 인덱스 1)
print(fruits[2])  # 출력: 체리 (세 번째 요소, 인덱스 2)
````

````python
# 음수 인덱스로 접근 (뒤에서부터)
# 음수 인덱스는 Python의 멋진 기능입니다!
print(fruits[-1])  # 출력: 포도 (마지막 요소)
print(fruits[-2])  # 출력: 체리 (뒤에서 두 번째)
````

### 🔍 인덱스 시각화

```
fruits = ["사과", "바나나", "체리", "포도"]

양수 인덱스:   0       1        2       3
음수 인덱스:  -4      -3       -2      -1
```

### 💡 실무 팁

- 첫 번째 요소: `list[0]`
- 마지막 요소: `list[-1]` (리스트 길이를 몰라도 OK!)
- 뒤에서 두 번째: `list[-2]`

## 3. 슬라이싱 기본 - 부분 추출하기

### `[시작:끝]` 형식으로 리스트의 일부분을 잘라낼 수 있습니다

슬라이싱은 리스트의 **연속된 부분**을 추출하는 강력한 기능입니다!

**중요**: 끝 인덱스는 포함되지 않습니다! (끝 - 1까지만 추출)

````python
# 숫자 리스트 만들기
numbers = [10, 20, 30, 40, 50, 60, 70]

# [시작:끝] - 시작 인덱스부터 (끝-1) 인덱스까지
print(numbers[1:4])  
# 출력: [20, 30, 40]
# 인덱스 1(20), 2(30), 3(40)까지 포함 (4는 미포함!)

print(numbers[0:3])  
# 출력: [10, 20, 30]
# 인덱스 0, 1, 2까지
````

````python
# [시작:] - 시작 인덱스부터 끝까지 전부
print(numbers[3:])  
# 출력: [40, 50, 60, 70]
# 인덱스 3부터 끝까지 모두
````

````python
# [:끝] - 처음부터 (끝-1) 인덱스까지
print(numbers[:3])  
# 출력: [10, 20, 30]
# 처음부터 인덱스 2까지
````

### 📝 슬라이싱 패턴 정리

```python
numbers = [10, 20, 30, 40, 50, 60, 70]

numbers[1:4]   # [20, 30, 40] - 인덱스 1부터 3까지
numbers[3:]    # [40, 50, 60, 70] - 인덱스 3부터 끝까지
numbers[:3]    # [10, 20, 30] - 처음부터 인덱스 2까지
numbers[:]     # [10, 20, 30, 40, 50, 60, 70] - 전체 복사
```

### 💡 왜 끝 인덱스는 포함 안 될까요?

이렇게 하면 **길이 계산이 쉽습니다**:

```python
numbers[1:4]  # 4 - 1 = 3개 요소
numbers[0:5]  # 5 - 0 = 5개 요소
```

## 4. 슬라이싱 심화 - 간격과 역순

### `[시작:끝:간격]` 형식으로 더 다양한 추출이 가능합니다

슬라이싱에 **세 번째 숫자**를 추가하면 간격(step)을 지정할 수 있습니다!

````python
numbers = [10, 20, 30, 40, 50, 60, 70]

# [::간격] - 전체 범위에서 간격을 두고 추출
print(numbers[::2])  
# 출력: [10, 30, 50, 70]
# 인덱스 0, 2, 4, 6 (짝수 인덱스)

print(numbers[1::2])  
# 출력: [20, 40, 60]
# 인덱스 1, 3, 5 (홀수 인덱스)
````

````python
# [::-1] - 역순으로 추출 (매우 자주 사용!)
# 간격을 -1로 하면 뒤에서부터 순회합니다
print(numbers[::-1])  
# 출력: [70, 60, 50, 40, 30, 20, 10]
# 리스트를 거꾸로 뒤집기!
````

### 🎯 실무 활용 예시

#### 1. 최근 데이터 역순 조회

```python
# 최신 뉴스를 먼저 보여주기

recent_news = news[::-1]  # [최신 뉴스 ... 오래된 뉴스]
```

#### 2. 격일 데이터 추출

```python
# 주간 매출에서 월, 수, 금 데이터만
weekly_sales = [100, 120, 110, 130, 125, 140, 135]
odd_days = weekly_sales[::2]  # [100, 110, 125, 135]
```

## 5. 리스트 연산 - 합치기와 반복하기

### `+` 연산자로 리스트를 합치고, `*` 연산자로 반복할 수 있습니다

리스트도 연산을 할 수 있습니다! 하지만 숫자 연산과는 의미가 다릅니다.

````python
# + 연산자: 두 리스트를 하나로 합치기 (concatenation)
list1 = [1, 2, 3]
list2 = [4, 5, 6]
combined = list1 + list2

print("합친 리스트:", combined)  
# 출력: 합친 리스트: [1, 2, 3, 4, 5, 6]

# 원본은 변하지 않습니다!
print("list1:", list1)  # 출력: list1: [1, 2, 3]
print("list2:", list2)  # 출력: list2: [4, 5, 6]
````

````python
# * 연산자: 리스트를 여러 번 반복 (repetition)
repeated = [0] * 5
print("반복 리스트:", repeated)  
# 출력: 반복 리스트: [0, 0, 0, 0, 0]

stars = ["★"] * 3
print("별 리스트:", stars)  
# 출력: 별 리스트: ['★', '★', '★']
````

## 6. len() 함수 - 리스트 길이 확인하기

### `len(리스트)`로 요소의 개수를 확인할 수 있습니다

`len()`은 리스트에 몇 개의 요소가 있는지 알려주는 매우 유용한 함수입니다!

````python
# 학생 명단
students = ["김철수", "이영희", "박민수", "정수진"]

# len()으로 학생 수 확인
count = len(students)
print(f"학생 수: {count}명")  
# 출력: 학생 수: 4명
````

````python
# 빈 리스트의 길이는 0
empty = []
print(f"빈 리스트 길이: {len(empty)}")  
# 출력: 빈 리스트 길이: 0
````

## 7. 실전 예제 - 상품 재고 관리 시스템

### 지금까지 배운 내용을 모두 활용해봅시다!

쇼핑몰의 상품 재고를 관리하는 간단한 시스템을 만들어보겠습니다.

````python
# 상품 목록 (재고가 있는 상품들)
products = ["노트북", "마우스", "키보드", "모니터", "헤드셋"]

# 1. 전체 상품 개수 확인 (len 활용)
total_products = len(products)
print(f"전체 상품 개수: {total_products}개")  
# 출력: 전체 상품 개수: 5개
````

````python
# 2. 처음 3개 상품 (인기 상품으로 표시)
popular = products[:3]  # 슬라이싱 활용
print("인기 상품:", popular)  
# 출력: 인기 상품: ['노트북', '마우스', '키보드']
````

````python
# 3. 마지막 상품 (가장 최근 입고된 상품)
last_product = products[-1]  # 음수 인덱싱 활용
print(f"마지막 상품: {last_product}")  
# 출력: 마지막 상품: 헤드셋
````

````python
# 4. 새 상품 추가 (리스트 연산 활용)
new_items = ["스피커", "웹캠"]
new_products = products + new_items  # + 연산자로 합치기

print("추가 후:", new_products)
# 출력: 추가 후: ['노트북', '마우스', '키보드', '모니터', '헤드셋', '스피커', '웹캠']

print(f"총 {len(new_products)}개 상품으로 증가!")  
# 출력: 총 7개 상품으로 증가!
````

### 🎯 이 예제에서 배운 것

1. **len()**: 전체 상품 개수 파악
2. **슬라이싱**: 처음 N개 추출 (인기 상품)
3. **음수 인덱스**: 마지막 요소 접근 (최신 상품)
4. **+ 연산**: 새 상품 추가 (재고 확장)

# 📌 핵심 정리

## 오늘 배운 내용

### 1. 리스트(List)의 개념
- 여러 개의 값을 순서대로 저장하는 자료구조
- `[값1, 값2, 값3]` 형식으로 생성
- FastAPI에서 여러 데이터를 반환할 때 필수!

### 2. 인덱싱(Indexing)
- `list[0]`: 첫 번째 요소 (인덱스는 0부터!)
- `list[-1]`: 마지막 요소
- 특정 위치의 값에 접근

### 3. 슬라이싱(Slicing)
- `list[시작:끝]`: 부분 추출 (끝은 미포함)
- `list[:3]`: 처음부터 3개
- `list[2:]`: 3번째부터 끝까지
- `list[::-1]`: 역순

### 4. 리스트 연산
- `list1 + list2`: 두 리스트 합치기
- `list * 3`: 리스트 3번 반복

### 5. len() 함수
- `len(list)`: 리스트의 요소 개수 반환
- 평균 계산, 반복문 등에 활용

## 🎯 기억할 점

### ✅ 꼭 기억하세요!

1. **인덱스는 0부터 시작!**
   - 첫 번째 = `[0]`, 두 번째 = `[1]`

2. **슬라이싱의 끝은 포함 안 됨!**
