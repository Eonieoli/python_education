## 📖 딕셔너리 메서드란 무엇인가?

딕셔너리는 Python에서 가장 강력한 자료구조 중 하나입니다. 단순히 `dict["key"]`로 값을 가져오는 것 외에도, **다양한 메서드**를 제공하여 딕셔너리를 효율적으로 다룰 수 있습니다.

### 왜 딕셔너리 메서드를 배워야 할까요?

실무에서는 다음과 같은 상황을 자주 마주칩니다:

1. **API에서 받은 데이터를 처리할 때**: 어떤 필드는 있고, 어떤 필드는 없을 수 있습니다
2. **설정 파일을 읽을 때**: 선택적 설정값이 있거나 없을 수 있습니다
3. **사용자 입력을 처리할 때**: 필수 정보와 선택 정보가 섞여 있습니다

이런 상황에서 **안전하게** 딕셔너리를 다루려면 메서드를 잘 알아야 합니다!

### FastAPI에서는?

FastAPI로 API를 만들 때, 딕셔너리 메서드는 정말 자주 사용됩니다:

```python
# API 응답에서 선택적 필드 처리
user_data = {
    "name": "홍길동",
    "email": "hong@example.com"
    # bio는 선택적 필드
}

# get()으로 안전하게 접근
bio = user_data.get("bio", "소개가 없습니다")
```

이제 딕셔너리를 **안전하고 효율적으로** 다루는 방법을 배워봅시다! 🚀

## 1. keys(), values(), items() 메서드

### 딕셔너리의 모든 키, 값, 키-값 쌍을 가져오기

딕셔너리에는 **세 가지 핵심 메서드**가 있습니다:

- `keys()`: 모든 **키**들을 가져옵니다
- `values()`: 모든 **값**들을 가져옵니다
- `items()`: 모든 **키-값 쌍**을 튜플로 가져옵니다

이 메서드들은 **for문**과 함께 사용하면 정말 강력합니다! (for문은 Day 3에서 배웁니다)

### 언제 사용하나요?

- `keys()`: "이 딕셔너리에 어떤 키들이 있지?" 확인할 때
- `values()`: "모든 값들만 필요해!" 할 때
- `items()`: "키와 값을 **함께** 순회하고 싶어!" 할 때 (가장 많이 씀!)

````python
# 사용자 정보 딕셔너리
user = {
    "name": "홍길동",
    "age": 25,
    "email": "hong@example.com"
}

# keys(): 모든 키 가져오기
# dict_keys 객체가 반환됩니다 (리스트처럼 사용 가능)
print(user.keys())  # 출력: dict_keys(['name', 'age', 'email'])

# values(): 모든 값 가져오기
# dict_values 객체가 반환됩니다
print(user.values())  # 출력: dict_values(['홍길동', 25, 'hong@example.com'])

# items(): 모든 키-값 쌍 가져오기 (튜플로)
# dict_items 객체가 반환되며, 각 요소는 (키, 값) 튜플입니다
print(user.items())  # 출력: dict_items([('name', '홍길동'), ('age', 25), ('email', 'hong@example.com')])
````

**실행 결과:**
```
dict_keys(['name', 'age', 'email'])
dict_values(['홍길동', 25, 'hong@example.com'])
dict_items([('name', '홍길동'), ('age', 25), ('email', 'hong@example.com')])
```

### 💡 알아두기

- `dict_keys`, `dict_values`, `dict_items`는 **특별한 객체**입니다
- 리스트처럼 보이지만, 메모리를 더 효율적으로 사용합니다
- 필요하면 `list()`로 변환할 수 있습니다: `list(user.keys())`
- **items()는 Day 3의 for문에서 정말 유용합니다!**

## 2. get() 메서드 - KeyError 방지 (🌟 핵심!)

### 안전하게 딕셔너리 값 가져오기

딕셔너리에서 값을 가져올 때, `dict["key"]` 방식은 **위험**합니다!

**왜 위험할까요?**

```python
product = {"name": "노트북", "price": 1200000}
print(product["stock"])  # ❌ KeyError 발생! 프로그램 중단!
```

만약 API에서 받은 데이터에 특정 키가 없다면? **프로그램이 죽습니다!** 💥

### get() 메서드가 해결책입니다!

`get()` 메서드는 다음과 같이 동작합니다:

```python
dict.get(key)           # 키가 없으면 None 반환 (에러 안남!)
dict.get(key, default)  # 키가 없으면 default 반환
```

### FastAPI에서 왜 중요한가요?

API 응답에서 **선택적 필드**를 다룰 때 필수입니다:

```python
# 사용자 프로필 (bio는 선택적)
user = {"name": "Alice", "email": "alice@example.com"}

# get()으로 안전하게 접근
bio = user.get("bio", "소개가 없습니다")  # ✅ 에러 없이 기본값 반환
```

````python
# 일반 접근 방식의 문제점
product = {"name": "노트북", "price": 1200000}

# 존재하는 키는 문제 없음
print(product["name"])  # 출력: 노트북

# 존재하지 않는 키는 에러 발생!
# print(product["stock"])  # ❌ KeyError: 'stock' 발생!

print("--- get() 메서드 사용 ---")

# get() 메서드로 안전하게 접근
print(product.get("name"))  # 출력: 노트북 (키가 있으면 값 반환)
print(product.get("stock"))  # 출력: None (키가 없으면 None 반환, 에러 없음!)

print("--- get()의 두 번째 인자: 기본값 설정 ---")

# get()의 두 번째 인자: 기본값 설정
# 키가 없으면 지정한 기본값을 반환합니다
print(product.get("stock", 0))  # 출력: 0 (키가 없으면 0 반환)
print(product.get("discount", 0.1))  # 출력: 0.1 (키가 없으면 0.1 반환)
````

**실행 결과:**
```
노트북
--- get() 메서드 사용 ---
노트북
None
--- get()의 두 번째 인자: 기본값 설정 ---
0
0.1
```

### 💡 핵심 정리

| 방법 | 키가 있을 때 | 키가 없을 때 |
|------|--------------|---------------|
| `dict["key"]` | 값 반환 | ❌ **KeyError 발생** |
| `dict.get("key")` | 값 반환 | ✅ `None` 반환 |
| `dict.get("key", default)` | 값 반환 | ✅ `default` 반환 |

### ⚠️ 실무 팁

**언제 `dict["key"]`를 사용하나요?**
- 키가 **반드시** 있어야 하는 경우 (없으면 에러가 맞음)
- 예: `user["id"]` (사용자 ID는 필수)

**언제 `dict.get()`을 사용하나요?**
- 키가 **선택적**인 경우 (있어도 되고 없어도 됨)
- 예: `user.get("bio", "소개 없음")` (소개는 선택)
- **API 응답 처리할 때 거의 항상 get() 사용!**

## 3. pop() 메서드 - 키 제거 및 값 반환

### 딕셔너리에서 키를 제거하면서 값 받아오기

`pop()` 메서드는 **두 가지 일**을 동시에 합니다:

1. 딕셔너리에서 키를 **제거**합니다
2. 제거된 키의 **값을 반환**합니다

```python
value = dict.pop(key)           # 키 제거 + 값 반환 (키 없으면 에러)
value = dict.pop(key, default)  # 키 제거 + 값 반환 (키 없으면 default)
```

### 언제 사용하나요?

1. **민감한 정보 제거**: 비밀번호를 응답에서 제거
2. **장바구니 관리**: 구매한 상품을 장바구니에서 제거
3. **임시 데이터 처리**: 처리 후 제거가 필요한 임시 필드

### del과의 차이점은?

```python
del dict["key"]    # 키만 제거, 값은 받지 못함
value = dict.pop("key")  # 키 제거 + 값도 받음 ✅
```

````python
# 장바구니 딕셔너리
cart = {
    "사과": 3,
    "바나나": 2,
    "우유": 1
}

print("원래 장바구니:", cart)  # 출력: 원래 장바구니: {'사과': 3, '바나나': 2, '우유': 1}

# pop()으로 키 제거하고 값 받기
# 바나나를 장바구니에서 제거하면서, 몇 개였는지 알아냅니다
banana_count = cart.pop("바나나")
print("제거된 바나나 개수:", banana_count)  # 출력: 제거된 바나나 개수: 2
print("남은 장바구니:", cart)  # 출력: 남은 장바구니: {'사과': 3, '우유': 1}

print("--- pop()에도 기본값 설정 가능 ---")

# pop()에도 기본값 설정 가능 (키가 없어도 에러 안남)
# 포도는 장바구니에 없지만, 기본값 0이 반환됩니다
grape_count = cart.pop("포도", 0)
print("제거된 포도 개수:", grape_count)  # 출력: 제거된 포도 개수: 0
print("최종 장바구니:", cart)  # 출력: 최종 장바구니: {'사과': 3, '우유': 1}
````

**실행 결과:**
```
원래 장바구니: {'사과': 3, '바나나': 2, '우유': 1}
제거된 바나나 개수: 2
남은 장바구니: {'사과': 3, '우유': 1}
--- pop()에도 기본값 설정 가능 ---
제거된 포도 개수: 0
최종 장바구니: {'사과': 3, '우유': 1}
```

### 💡 실무 활용 예시

**API 응답에서 비밀번호 제거:**
```python
# 사용자 데이터 (DB에서 가져옴)
user_data = {
    "id": 1,
    "username": "hong123",
    "password": "secret123"  # 응답에 포함하면 안 됨!
}

# pop()으로 비밀번호 제거 (값은 받지 않아도 됨)
user_data.pop("password")

# 이제 안전하게 응답 가능
return user_data  # {"id": 1, "username": "hong123"}
```

## 4. update() 메서드 - 딕셔너리 병합

### 두 개의 딕셔너리를 합치기

`update()` 메서드는 **다른 딕셔너리의 내용**을 현재 딕셔너리에 추가합니다.

```python
dict1.update(dict2)  # dict2의 모든 키-값을 dict1에 추가/업데이트
```

### 동작 방식:

1. **새로운 키**가 있으면 → **추가**합니다
2. **기존 키**가 있으면 → **값을 업데이트**합니다 (덮어씀)

### 언제 사용하나요?

1. **설정 병합**: 기본 설정 + 사용자 설정
2. **정보 추가**: 기본 정보 + 추가 정보
3. **데이터 업데이트**: 기존 데이터를 새 데이터로 부분 갱신

### FastAPI에서는?

```python
# 기본 응답 템플릿
response = {"status": "success"}

# 사용자 데이터 추가
user_data = {"name": "Alice", "email": "alice@example.com"}
response.update(user_data)

# 결과: {"status": "success", "name": "Alice", "email": "alice@example.com"}
```

````python
# 기본 사용자 정보
user_info = {
    "name": "김철수",
    "age": 30
}

# 추가 정보
extra_info = {
    "email": "kim@example.com",
    "phone": "010-1234-5678"
}

print("원래 정보:", user_info)  # 출력: 원래 정보: {'name': '김철수', 'age': 30}

# update()로 병합
# extra_info의 모든 키-값이 user_info에 추가됩니다
user_info.update(extra_info)
print("병합 후:", user_info)
# 출력: 병합 후: {'name': '김철수', 'age': 30, 'email': 'kim@example.com', 'phone': '010-1234-5678'}

print("--- 기존 키가 있으면 값이 업데이트됨 ---")

# 기존 키가 있으면 값이 업데이트됨
# age는 이미 있으므로 31로 업데이트되고, city는 새로 추가됩니다
user_info.update({"age": 31, "city": "서울"})
print("업데이트 후:", user_info)
# 출력: 업데이트 후: {'name': '김철수', 'age': 31, 'email': 'kim@example.com', 'phone': '010-1234-5678', 'city': '서울'}
````

**실행 결과:**
```
원래 정보: {'name': '김철수', 'age': 30}
병합 후: {'name': '김철수', 'age': 30, 'email': 'kim@example.com', 'phone': '010-1234-5678'}
--- 기존 키가 있으면 값이 업데이트됨 ---
업데이트 후: {'name': '김철수', 'age': 31, 'email': 'kim@example.com', 'phone': '010-1234-5678', 'city': '서울'}
```

### 💡 주의사항

**update()는 원본을 수정합니다!**

```python
dict1 = {"a": 1}
dict2 = {"b": 2}

dict1.update(dict2)  # dict1이 변경됨
print(dict1)  # {'a': 1, 'b': 2}
print(dict2)  # {'b': 2} (변경 없음)
```

원본을 유지하고 싶다면:
```python
# 방법 1: 복사본 만들기
dict3 = dict1.copy()
dict3.update(dict2)

# 방법 2: 언패킹 (Day 4에서 배움)
dict3 = {**dict1, **dict2}
```

## 5. in 연산자 - 키 존재 확인

### 딕셔너리에 특정 키가 있는지 확인하기

`in` 연산자는 딕셔너리에 **키가 존재하는지** 확인합니다.

```python
"key" in dict     # True 또는 False 반환
"key" not in dict # True 또는 False 반환
```

### 왜 필요한가요?

**키가 있는지 먼저 확인**하고, 있을 때만 접근하고 싶을 때 사용합니다!

```python
# 방법 1: in 연산자 (Day 3 if문에서 활용)
if "bio" in user:
    print(user["bio"])

# 방법 2: get() (더 간단!)
bio = user.get("bio", "소개 없음")
```

### in과 get()의 조합

`in` 연산자와 `get()` 메서드는 **함께 사용**하면 더욱 강력합니다:

- `in`: 키 존재 여부만 확인 (True/False)
- `get()`: 키가 없어도 안전하게 값 가져오기

````python
# 재고 딕셔너리
inventory = {
    "노트북": 5,
    "마우스": 20,
    "키보드": 15
}

# in 연산자로 키 존재 확인 (True/False 반환)
print("노트북" in inventory)  # 출력: True (있음)
print("모니터" in inventory)  # 출력: False (없음)

print("--- not in 연산자 ---")

# not in 연산자 (키가 없는지 확인)
print("모니터" not in inventory)  # 출력: True (없음)
print("노트북" not in inventory)  # 출력: False (있음)

print("--- in 연산자 + get() 조합 ---")

# in 연산자 + get() 조합 (안전한 패턴)
# 방법 1: get()으로 기본값 제공 (더 간단!)
print(inventory.get("노트북", 0))  # 출력: 5 (있으면 값 반환)
print(inventory.get("모니터", 0))  # 출력: 0 (없으면 기본값 반환)

# 방법 2: in 연산자로 먼저 확인
has_laptop = "노트북" in inventory
has_monitor = "모니터" in inventory
print(f"노트북 있음: {has_laptop}")  # 출력: 노트북 있음: True
print(f"모니터 있음: {has_monitor}")  # 출력: 모니터 있음: False
````

**실행 결과:**
```
True
False
--- not in 연산자 ---
True
False
--- in 연산자 + get() 조합 ---
5
0
노트북 있음: True
모니터 있음: False
```

### 💡 실무 팁: 언제 뭘 사용하나요?

| 상황 | 방법 | 예시 |
|------|------|------|
| 값이 필요 없고 존재만 확인 | `in` | `if "admin" in roles:` |
| 값도 필요하고 기본값 있음 | `get()` | `bio = user.get("bio", "없음")` |
| 있을 때만 특별한 처리 | `in` + `dict[]` | `if "key" in dict: print(dict["key"])` |

### ⚠️ 주의: in은 키만 확인합니다!

```python
user = {"name": "Alice", "age": 25}

print("name" in user)    # ✅ True (키 확인)
print("Alice" in user)   # ❌ False (값은 확인 안 됨!)
print("Alice" in user.values())  # ✅ True (값 확인하려면 이렇게)
```

## 6. 실전 예제: API 응답 처리

### FastAPI에서 실제로 사용하는 패턴

실무에서 API를 만들 때, **모든 필드가 항상 있는 건 아닙니다**.

예를 들어:
- 사용자 프로필: 이름, 이메일은 필수, 소개(bio)와 프로필 사진(avatar)은 선택
- 상품 정보: 이름, 가격은 필수, 할인율과 재고는 선택

이런 경우 `get()` 메서드가 정말 유용합니다!

### 없는 필드 때문에 에러가 나면 안 돼요!

API 서버가 에러로 중단되면 **서비스 전체가 멈춥니다**. 그래서 `get()`으로 **안전하게** 처리해야 합니다.

````python
# API에서 받은 사용자 데이터 (일부 정보가 없을 수 있음)
api_response = {
    "user_id": 12345,
    "username": "alice",
    "email": "alice@example.com"
    # bio와 avatar는 선택적 필드 (없을 수 있음)
}

# get()으로 안전하게 접근
# 키가 있으면 값을 사용하고, 없으면 기본값을 사용합니다
user_id = api_response.get("user_id")  # 필수 필드는 기본값 없이
username = api_response.get("username", "익명")  # 혹시 모르니 기본값 제공
bio = api_response.get("bio", "소개가 없습니다")  # 없는 키는 기본값 사용
avatar = api_response.get("avatar", "default.png")  # 없는 키는 기본값 사용

# 결과 출력 - 에러 없이 안전하게 실행됩니다!
print(f"사용자: {username} (ID: {user_id})")  # 출력: 사용자: alice (ID: 12345)
print(f"소개: {bio}")  # 출력: 소개: 소개가 없습니다
print(f"프로필 사진: {avatar}")  # 출력: 프로필 사진: default.png
````

**실행 결과:**
```
사용자: alice (ID: 12345)
소개: 소개가 없습니다
프로필 사진: default.png
```

### 💡 실전 패턴

**FastAPI에서 이렇게 사용합니다:**

```python
@app.get("/users/{user_id}")
def get_user(user_id: int):
    # DB에서 사용자 정보 가져오기 (딕셔너리로 반환)
    user_data = db.get_user(user_id)
    
    # 선택적 필드를 안전하게 처리
    response = {
        "id": user_data["id"],  # 필수 필드
        "name": user_data["name"],  # 필수 필드
        "bio": user_data.get("bio", "소개 없음"),  # 선택 필드
        "avatar": user_data.get("avatar", "default.png")  # 선택 필드
    }
    
    return response
```

### ⚠️ 실무에서 꼭 기억하세요!

1. **필수 필드**: `dict["key"]` 사용 (없으면 에러가 맞음)
2. **선택 필드**: `dict.get("key", default)` 사용 (없어도 괜찮음)
3. **민감 정보**: `dict.pop("password")` 같이 제거 후 응답
4. **설정 병합**: `dict.update()` 로 기본값 + 사용자 설정

## 📌 핵심 정리

이번 시간에 배운 딕셔너리 메서드들:

1. **`keys()`, `values()`, `items()`**: 딕셔너리의 모든 키, 값, 키-값 쌍 가져오기
   - for문과 함께 사용하면 강력함 (Day 3에서 배움)

2. **`get(key, default)`**: 🌟 **가장 중요!** 안전하게 값 가져오기
   - KeyError 방지
   - API 응답 처리의 필수 패턴

3. **`pop(key, default)`**: 키 제거 + 값 반환
   - 민감 정보 제거할 때 유용
   - 장바구니 관리 같은 상황에서 활용

4. **`update(dict)`**: 딕셔너리 병합
   - 새 키는 추가, 기존 키는 업데이트
   - 설정 병합, 정보 추가에 활용

5. **`in` 연산자**: 키 존재 확인
   - `"key" in dict` → True/False
   - get()과 함께 사용하면 더 강력
