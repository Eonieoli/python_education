## 📖 딕셔너리(Dictionary)란 무엇인가?

딕셔너리는 **key(키)와 value(값)의 쌍**으로 데이터를 저장하는 Python의 핵심 자료구조입니다.

### 왜 딕셔너리가 필요할까요?

학생 정보를 저장한다고 생각해봅시다. 리스트로는 이렇게 저장할 수 있습니다:

```python
student = ["김철수", 20, "컴퓨터공학"]
```

하지만 이 방법의 문제점은 무엇일까요?

- `student[0]`이 이름인지, `student[1]`이 나이인지 **기억해야 합니다**
- 순서를 바꾸면 **의미가 완전히 달라집니다**
- 데이터가 무엇을 의미하는지 **코드만 봐서는 알기 어렵습니다**

**딕셔너리를 사용하면**:

```python
student = {
    "name": "김철수",
    "age": 20,
    "major": "컴퓨터공학"
}
```

- 각 데이터가 무엇을 의미하는지 **명확합니다** ✨
- 순서를 바꿔도 **의미가 그대로 유지됩니다**
- 필요한 정보를 `student["name"]`처럼 **직관적으로 접근**할 수 있습니다

### 실생활 비유

딕셔너리는 실제 **사전(Dictionary)**과 비슷합니다:

- 📖 사전: **단어(key)** → **뜻(value)**
- 💾 딕셔너리: **키(key)** → **값(value)**

사전에서 "사과"라는 단어를 찾으면 그 뜻을 얻을 수 있듯이, 딕셔너리에서 "name" 키를 찾으면 "김철수"라는 값을 얻을 수 있습니다.

### FastAPI에서는 어떻게 사용될까요?

FastAPI로 웹 API를 만들 때, **거의 모든 데이터가 딕셔너리 형태**로 주고받습니다!

```python
# API가 반환하는 JSON 데이터 (사실은 딕셔너리!)
{
    "user_id": 1,
    "username": "alice",
    "email": "alice@example.com",
    "is_active": true
}
```

이것이 바로 **JSON(JavaScript Object Notation)** 형식이고, Python에서는 딕셔너리로 표현됩니다. 즉:

> **딕셔너리를 마스터하는 것 = FastAPI의 핵심을 이해하는 것!** 🎯

### 핵심 개념

1. **Key(키)**: 데이터를 찾기 위한 이름 (문자열, 숫자 등)
2. **Value(값)**: 실제 저장되는 데이터 (어떤 타입이든 가능)
3. **중괄호 `{}`**: 딕셔너리를 만들 때 사용
4. **콜론 `:`**: 키와 값을 연결
5. **쉼표 `,`**: 여러 쌍을 구분

## 1. 딕셔너리 생성과 기본 구조

딕셔너리를 만드는 방법은 매우 간단합니다. 중괄호 `{}` 안에 `key: value` 형태로 데이터를 넣으면 됩니다.

### 기본 문법

```python
딕셔너리명 = {
    "키1": 값1,
    "키2": 값2,
    "키3": 값3
}
```

### 주의사항

- 키는 보통 **문자열**을 사용합니다 (따옴표로 감싸기)
- 값은 **어떤 타입이든** 가능합니다 (숫자, 문자열, 리스트, 심지어 또 다른 딕셔너리!)
- 각 쌍은 **쉼표**로 구분합니다
- 마지막 쌍 뒤의 쉼표는 있어도 되고 없어도 됩니다

실제 코드로 확인해봅시다! 👇

````python
# 빈 딕셔너리 만들기
# 아무것도 들어있지 않은 빈 딕셔너리를 만들 수 있습니다
empty_dict = {}
print(empty_dict)  # 출력: {}
print(type(empty_dict))  # 출력: <class 'dict'>
# type() 함수로 확인하면 'dict' 타입임을 알 수 있습니다
````

**실행 결과:**
```
{}
<class 'dict'>
```

빈 딕셔너리는 `{}`로 표시되며, 타입은 `dict`입니다. 나중에 데이터를 추가할 수 있습니다.

---

````python
# 학생 정보를 딕셔너리로 저장
# 실제로 의미 있는 데이터를 담은 딕셔너리를 만들어봅시다
student = {
    "name": "김철수",      # 키: "name", 값: "김철수"
    "age": 20,            # 키: "age", 값: 20 (숫자)
    "major": "컴퓨터공학"  # 키: "major", 값: "컴퓨터공학"
}

# 딕셔너리 전체를 출력하면 모든 key-value 쌍이 보입니다
print(student)  
# 출력: {'name': '김철수', 'age': 20, 'major': '컴퓨터공학'}
````

**실행 결과:**
```python
{'name': '김철수', 'age': 20, 'major': '컴퓨터공학'}
```

### 결과 해석

- 딕셔너리는 중괄호 `{}`로 감싸져 있습니다
- 각 쌍이 `'키': 값` 형태로 표시됩니다
- 문자열 키는 작은따옴표로 표시됩니다
- 숫자 값은 따옴표 없이 그대로 표시됩니다

### 💡 알아두기

- **순서는 중요하지 않습니다**: `{"name": "김철수", "age": 20}`과 `{"age": 20, "name": "김철수"}`는 동일합니다
- Python 3.7+ 버전부터는 삽입 순서가 유지되지만, 딕셔너리를 사용할 때 순서에 의존하면 안 됩니다
- 키는 **고유(unique)**해야 합니다. 같은 키를 두 번 사용하면 나중 값으로 덮어씌워집니다

## 2. 딕셔너리 접근과 수정

딕셔너리에서 값을 가져오거나 바꾸는 방법을 배워봅시다.

### 값 가져오기

리스트는 인덱스(숫자)로 접근했지만, 딕셔너리는 **키(key)**로 접근합니다:

```python
딕셔너리["키"]
```

### 값 수정하기

기존 키에 새로운 값을 할당하면 값이 변경됩니다:

```python
딕셔너리["키"] = 새로운_값
```

코드로 확인해봅시다! 👇

````python
# 대괄호 []로 key를 사용해 value에 접근
student = {"name": "김철수", "age": 20, "major": "컴퓨터공학"}

# "name" 키로 이름 가져오기
print(student["name"])  # 출력: 김철수

# "age" 키로 나이 가져오기
print(student["age"])   # 출력: 20
````

**실행 결과:**
```
김철수
20
```

대괄호 안에 키를 넣으면 해당하는 값을 가져올 수 있습니다. 리스트의 인덱싱과 비슷하지만, 숫자 대신 **의미 있는 키**를 사용한다는 점이 다릅니다!

---

````python
# 값 수정 (기존 key에 새로운 value 할당)
# 학생이 생일을 맞아 나이가 한 살 증가했습니다
student["age"] = 21

# 변경된 값 확인
print(student["age"])   # 출력: 21

# 여러 값을 한 번에 확인 (f-string 활용)
# f-string으로 딕셔너리 값들을 조합해서 문장을 만들 수 있습니다
print(f"{student['name']}님은 {student['age']}세입니다")
# 출력: 김철수님은 21세입니다
````

**실행 결과:**
```
21
김철수님은 21세입니다
```

### 핵심 포인트

- `student["age"] = 21`은 **이미 존재하는 키**의 값을 수정합니다
- 수정 후 다시 접근하면 새로운 값이 나옵니다
- f-string 안에서도 딕셔너리 접근이 가능합니다: `{딕셔너리['키']}`

### 💡 f-string에서 딕셔너리 사용 시 주의

f-string 안에서 딕셔너리를 사용할 때는 **작은따옴표**와 **큰따옴표**를 구분해야 합니다:

- ✅ 올바른 예: `f"{student['name']}"`  (f-string은 큰따옴표, 키는 작은따옴표)
- ❌ 잘못된 예: `f"{student["name"]}"`  (따옴표가 겹침)

## 3. 딕셔너리 추가와 삭제

딕셔너리는 **가변(mutable)** 자료형입니다. 즉, 생성 후에도 항목을 추가하거나 삭제할 수 있습니다.

### 새 항목 추가하기

**존재하지 않는 키**에 값을 할당하면 새로운 항목이 추가됩니다:

```python
딕셔너리["새로운_키"] = 값
```

### 항목 삭제하기

`del` 키워드를 사용합니다:

```python
del 딕셔너리["삭제할_키"]
```

### 추가 vs 수정 구분하기

- **키가 이미 존재** → 값이 **수정**됩니다
- **키가 존재하지 않음** → 새 항목이 **추가**됩니다

실습해봅시다! 👇

````python
# 새로운 key-value 추가 (존재하지 않는 key에 값 할당)
student = {"name": "김철수", "age": 20}
print("초기 상태:", student)

# "major" 키는 아직 없습니다 → 새로운 항목 추가!
student["major"] = "컴퓨터공학"
print("전공 추가 후:", student)

# "grade" 키도 없습니다 → 또 추가!
student["grade"] = 3
print("학년 추가 후:", student)
# 출력: {'name': '김철수', 'age': 20, 'major': '컴퓨터공학', 'grade': 3}
````

**실행 결과:**
```python
초기 상태: {'name': '김철수', 'age': 20}
전공 추가 후: {'name': '김철수', 'age': 20, 'major': '컴퓨터공학'}
학년 추가 후: {'name': '김철수', 'age': 20, 'major': '컴퓨터공학', 'grade': 3}
```

딕셔너리에 항목이 점점 늘어나는 것을 확인할 수 있습니다! 처음에 2개였던 항목이 4개로 늘어났습니다.

---

````python
# del 키워드로 항목 삭제
# grade 정보는 더 이상 필요하지 않다고 가정해봅시다
del student["grade"]  # "grade" 키와 그 값을 딕셔너리에서 완전히 제거

print(student)
# 출력: {'name': '김철수', 'age': 20, 'major': '컴퓨터공학'}
# grade가 사라진 것을 확인할 수 있습니다
````

**실행 결과:**
```python
{'name': '김철수', 'age': 20, 'major': '컴퓨터공학'}
```

### 주의사항 ⚠️

- `del`로 **존재하지 않는 키**를 삭제하려고 하면 **KeyError**가 발생합니다
- 삭제는 **영구적**입니다. 삭제한 항목은 복구할 수 없습니다
- `del`은 딕셔너리의 항목뿐만 아니라 변수 자체도 삭제할 수 있습니다: `del student`

### 💡 실무 팁

실제 프로그래밍에서는 데이터를 삭제하기보다는:

- 값을 `None`으로 설정하거나
- 별도의 "삭제됨" 플래그를 사용하는 경우가 많습니다

```python
student["grade"] = None  # 삭제보다는 "값 없음"으로 표시
```

## 4. KeyError와 안전한 접근

딕셔너리에서 **가장 흔한 에러**가 바로 `KeyError`입니다. 이것은 무엇이고, 어떻게 방지할까요?

### KeyError란?

딕셔너리에 **존재하지 않는 키**로 접근하려고 할 때 발생하는 에러입니다:

```python
student = {"name": "김철수", "age": 20}
print(student["email"])  # KeyError! (email 키가 없음)
```

### 왜 문제일까요?

프로그램이 **중단**됩니다! 특히 웹 API에서는 치명적입니다:

- 사용자가 선택적으로 입력한 정보 (이메일, 전화번호 등)
- 외부 API에서 가져온 불완전한 데이터
- 데이터베이스에서 일부 필드가 없는 경우

### 안전하게 접근하는 방법

#### 방법 1: `in` 연산자로 확인 후 접근

```python
if "email" in student:
    print(student["email"])
else:
    print("이메일 정보가 없습니다")
```

#### 방법 2: `get()` 메서드 사용 (다음 교시에 배울 예정)

```python
email = student.get("email", "없음")  # 키가 없으면 "없음" 반환
```

실습해봅시다! 👇

````python
# 존재하지 않는 key에 접근하면 KeyError 발생!
student = {"name": "김철수", "age": 20}

# 주의: 아래 줄의 주석을 해제하면 에러가 발생합니다!
# print(student["email"])  # KeyError: 'email'

print("에러를 방지하기 위해 주석 처리했습니다.")
print("실제로는 KeyError: 'email' 에러가 발생하고 프로그램이 멈춥니다.")
````

### KeyError 예시

만약 위 코드에서 주석을 해제하고 실행하면:

```
KeyError: 'email'
```

이런 에러가 발생하며 프로그램이 중단됩니다. 이것은:

- 딕셔너리에 "email"이라는 키가 없다는 의미입니다
- Python이 "이 키로 찾을 값이 없어요!"라고 알려주는 것입니다

---

````python
# 안전하게 확인하는 방법: in 연산자 사용
# 키가 딕셔너리에 존재하는지 먼저 확인합니다

if "email" in student:  # "email" 키가 student 딕셔너리에 있는가?
    print(student["email"])  # 있으면 출력
else:
    print("이메일 정보가 없습니다")  # 없으면 친절한 메시지
    # 출력: 이메일 정보가 없습니다
````

**실행 결과:**
```
이메일 정보가 없습니다
```

에러 없이 안전하게 처리됩니다! 프로그램이 중단되지 않고 계속 실행됩니다.

---

````python
# key 존재 확인
# in 연산자는 True/False(불린값)를 반환합니다

print("name" in student)   # "name" 키가 있나요? → True
# 출력: True

print("phone" in student)  # "phone" 키가 있나요? → False
# 출력: False
````

**실행 결과:**
```
True
False
```

### in 연산자 활용법

`in` 연산자는 매우 유용합니다:

- **단독 사용**: `"키" in 딕셔너리` → True 또는 False 반환
- **if문과 함께**: 키가 있을 때만 실행
- **not in**: 키가 **없는지** 확인 → `"키" not in 딕셔너리`

```python
if "email" not in student:
    print("이메일을 입력해주세요")
```

### 💡 실무에서의 안전한 접근

FastAPI 개발 시 이런 패턴을 자주 사용합니다:

```python
# API에서 받은 데이터 처리
user_data = {"username": "alice", "age": 25}

# 선택적 필드(optional field) 안전하게 처리
if "email" in user_data:
    email = user_data["email"]
else:
    email = "미등록"  # 기본값 설정
```

이제 배운 내용을 종합해서 **실무에 가까운 예제**를 다뤄봅시다!

딕셔너리를 사용하면 이 모든 정보를 **하나의 변수**로 관리할 수 있고, **의미가 명확**합니다!

코드로 확인해봅시다! 👇

````python
# 온라인 쇼핑몰의 상품 정보
# 실제 쇼핑몰 API에서 이런 형태의 데이터를 주고받습니다
product = {
    "id": 1001,           # 상품 고유 번호
    "name": "노트북",      # 상품명
    "price": 1200000,     # 가격 (원)
    "stock": 15           # 재고 수량
}

print("=== 상품 정보 ===")
print(product)
````

**실행 결과:**
```python
=== 상품 정보 ===
{'id': 1001, 'name': '노트북', 'price': 1200000, 'stock': 15}
```

이것이 바로 실제 쇼핑몰 API에서 사용하는 데이터 형식입니다! FastAPI에서는 이런 딕셔너리가 자동으로 JSON으로 변환됩니다.

---

````python
# 상품 정보 출력
# 고객에게 보여줄 형태로 예쁘게 포맷팅합니다

print(f"상품명: {product['name']}")
# f-string의 :, 포맷 지정자를 사용하면 천 단위 구분 기호를 추가할 수 있습니다
print(f"가격: {product['price']:,}원")  # :,는 천 단위 콤마
print(f"재고: {product['stock']}개")
````

**실행 결과:**
```
상품명: 노트북
가격: 1,200,000원
재고: 15개
```

### f-string 포맷팅 팁

- `:,`는 **천 단위 구분 기호**를 추가합니다
- `1200000`이 `1,200,000`으로 표시되어 읽기 쉽습니다
- 실제 웹사이트나 앱에서 보는 것과 똑같은 형식입니다!

---

````python
# 재고 업데이트 (판매 발생)
# 고객이 노트북 1대를 구매했습니다

print("\n=== 판매 처리 ===")
print(f"판매 전 재고: {product['stock']}개")

# 재고에서 1을 빼서 업데이트
product["stock"] = product["stock"] - 1

print(f"판매 후 재고: {product['stock']}개")  
# 출력: 판매 후 재고: 14개
````

**실행 결과:**
```
=== 판매 처리 ===
판매 전 재고: 15개
판매 후 재고: 14개
```

### 재고 관리 로직

실제 쇼핑몰에서는 이런 식으로 재고를 관리합니다:

1. 주문이 들어옴
2. 현재 재고 확인 (`product["stock"]`)
3. 재고가 충분하면 감소 (`product["stock"] = product["stock"] - 1`)
4. 재고가 부족하면 "품절" 메시지

### 💡 복합 할당 연산자

`product["stock"] = product["stock"] - 1`은 다음과 같이 줄일 수 있습니다:

```python
product["stock"] -= 1  # 같은 의미, 더 간결!
```

---

````python
# 할인 정보 추가
# 특별 세일 기간에 10% 할인을 적용합니다

print("\n=== 할인 이벤트 ===")
product["discount_rate"] = 0.1  # 10% 할인 (0.1 = 10/100)

# 할인가 계산
# 원가 × (1 - 할인율) = 할인가
# 1,200,000 × (1 - 0.1) = 1,200,000 × 0.9 = 1,080,000
discounted_price = product["price"] * (1 - product["discount_rate"])

print(f"정가: {product['price']:,}원")
print(f"할인율: {product['discount_rate'] * 100:.0f}%")  # 0.1 → 10%
print(f"할인가: {discounted_price:,.0f}원")  
# 출력: 할인가: 1,080,000원
# :.0f는 소수점 없이 정수로 표시
````

**실행 결과:**
```
=== 할인 이벤트 ===
정가: 1,200,000원
할인율: 10%
할인가: 1,080,000원
```

### 할인가 계산 공식 이해하기

할인가를 계산하는 방법:

```
할인가 = 원가 × (1 - 할인율)
```

예시:
- 원가: 1,200,000원
- 할인율: 10% (0.1)
- 계산: 1,200,000 × (1 - 0.1) = 1,200,000 × 0.9 = 1,080,000원

### f-string 포맷팅 고급

- `:.0f`: 소수점 0자리 (정수로 표시)
- `:,.0f`: 천 단위 구분 + 소수점 0자리
- `* 100`: 0.1을 10으로 변환 (퍼센트 표시용)

## 📌 핵심 정리

이번 시간에 배운 내용을 정리해봅시다!

### 1. 딕셔너리의 핵심 개념

- **key-value 구조**: 키로 값을 저장하고 찾는 자료구조
- **중괄호 `{}`**: 딕셔너리를 만들 때 사용
- **콜론 `:`**: 키와 값을 연결
- **가변(mutable)**: 생성 후에도 수정, 추가, 삭제 가능

### 2. 딕셔너리 기본 문법

```python
# 생성
딕셔너리 = {"키1": 값1, "키2": 값2}

# 접근
딕셔너리["키"]

# 수정
딕셔너리["기존_키"] = 새_값

# 추가
딕셔너리["새_키"] = 값

# 삭제
del 딕셔너리["키"]

# 안전한 확인
if "키" in 딕셔너리:
    # 키가 존재할 때만 실행
```

### 3. KeyError 방지하기

- **문제**: 없는 키로 접근 → `KeyError` 발생 → 프로그램 중단
- **해결**: `in` 연산자로 먼저 확인
- **패턴**: `if "키" in 딕셔너리:` 사용

### 4. FastAPI와의 연결

- 딕셔너리 = JSON 구조
- API 요청/응답 데이터는 대부분 딕셔너리 형태
- 웹 개발의 핵심 자료구조

### 중요한 개념

1. **딕셔너리는 순서가 중요하지 않습니다**
   - 키로 값을 찾기 때문에 순서는 의미 없음

   - 각 키는 하나의 값만 가질 수 있음

3. **KeyError를 항상 조심하세요**
   - 없는 키로 접근하면 에러!
   - `in` 연산자로 먼저 확인하는 습관
   - 다음 교시에서 배울 `get()` 메서드가 더 편리함

4. **FastAPI의 핵심은 딕셔너리입니다**
   - API 데이터 = JSON = 딕셔너리
   - 딕셔너리를 잘 다루면 FastAPI도 쉬워집니다!
