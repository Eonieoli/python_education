## 📖 리스트는 왜 '수정 가능'할까요?

### 가변성(Mutable)이란?

지난 시간에 배운 **문자열**을 기억하시나요? 문자열은 한 번 만들면 **절대 내용을 바꿀 수 없었습니다** (불변성, Immutable). 

```python
text = "Hello"
text[0] = "h"  # ❌ 에러! 문자열은 수정 불가
```

하지만 **리스트는 다릅니다!** 리스트는 생성 후에도 자유롭게 내용을 변경할 수 있습니다. 이것을 **가변성(Mutable)**이라고 합니다.

```python
fruits = ["사과", "바나나", "딸기"]
fruits[1] = "포도"  # ✅ 가능! 리스트는 수정 가능
print(fruits)  # ['사과', '포도', '딸기']
```

### 왜 이게 중요할까요?

실무에서는 데이터가 **끊임없이 변합니다**:
- 쇼핑몰 장바구니: 상품 추가/삭제
- TODO 앱: 할 일 추가/완료 처리
- 재고 시스템: 입고/출고/반품
- 사용자 목록: 회원가입/탈퇴

이런 변화를 **효율적으로** 처리하려면 리스트의 수정 기능이 필수입니다!

### FastAPI에서는?

FastAPI로 API를 만들 때, 서버는 데이터를 **계속 추가하고 삭제**해야 합니다:

```python
# 사용자 목록 관리
users = [{"id": 1, "name": "Alice"}, {"id": 2, "name": "Bob"}]

@app.post("/users")  # 새 사용자 추가
def create_user(user):
    users.append(user)  # ← 리스트에 추가!
    return user

@app.delete("/users/{user_id}")  # 사용자 삭제
def delete_user(user_id):
    users.pop(user_id)  # ← 리스트에서 삭제!
```

리스트 수정 메서드를 잘 다루면, 이런 API를 자유자재로 만들 수 있습니다! 💪

### 메서드(Method)란?

메서드는 **특정 자료형에 딸려있는 함수**입니다. 점(`.`)을 찍어서 사용합니다:

```python
리스트.메서드()  # 예: fruits.append("사과")
```

오늘은 리스트를 수정하는 **6가지 핵심 메서드**를 배워봅시다!

## 1. 리스트 가변성 확인하기

### 문자열 vs 리스트 비교

먼저 문자열과 리스트의 차이를 직접 확인해봅시다. 문자열은 수정이 불가능하지만, 리스트는 수정이 가능합니다!

````python
# 리스트는 생성 후에도 내용을 수정할 수 있습니다 (가변성)
# 문자열은 수정 불가능 (불변성)이지만, 리스트는 수정 가능!

fruits = ["사과", "바나나", "딸기"]
print("원본:", fruits)  # 출력: 원본: ['사과', '바나나', '딸기']

# 리스트 내용을 직접 수정 가능
fruits[1] = "포도"  # 인덱스 1번(두 번째) 요소를 "포도"로 변경
print("수정 후:", fruits)  # 출력: 수정 후: ['사과', '포도', '딸기']
````

**실행 결과:**
```
원본: ['사과', '바나나', '딸기']
수정 후: ['사과', '포도', '딸기']
```

### 💡 알아두기

- 리스트는 인덱스로 접근해서 **직접 값을 바꿀 수 있습니다**
- 문자열은 이렇게 하면 에러가 발생합니다 (TypeError)
- 이 차이점이 리스트를 강력하게 만들어줍니다!

**주의사항:**
- 존재하지 않는 인덱스에 값을 할당하면 에러 발생 (IndexError)
- 예: `fruits[100] = "수박"` ❌

## 2. append() - 리스트 끝에 요소 추가하기

### append()는 가장 많이 쓰는 메서드입니다!

`append()`는 리스트의 **맨 끝**에 새로운 요소를 추가합니다. "추가하다(append)"라는 뜻 그대로입니다.

**문법:**
```python
리스트.append(추가할_값)
```

**언제 사용하나요?**
- 쇼핑몰 장바구니에 상품 담기
- TODO 리스트에 새 할 일 추가
- 채팅 메시지 목록에 새 메시지 추가
- API에서 새 사용자 등록

````python
# append()는 리스트 끝에 새 요소를 추가합니다
shopping = ["우유", "빵"]
print("장바구니:", shopping)  # 출력: 장바구니: ['우유', '빵']

shopping.append("계란")  # "계란"을 끝에 추가
print("추가 후:", shopping)  # 출력: 추가 후: ['우유', '빵', '계란']

shopping.append("버터")  # "버터"도 끝에 추가
print("또 추가:", shopping)  # 출력: 또 추가: ['우유', '빵', '계란', '버터']
````

**실행 결과:**
```
장바구니: ['우유', '빵']
추가 후: ['우유', '빵', '계란']
또 추가: ['우유', '빵', '계란', '버터']
```

### 💡 핵심 포인트

- `append()`는 **항상 맨 끝**에만 추가합니다
- 중간에 넣고 싶다면? → `insert()` 사용 (다음에 배웁니다!)
- 한 번에 하나씩만 추가됩니다
- **원본 리스트를 직접 수정**합니다 (새 리스트를 만들지 않음)

**실전 팁:**
- FastAPI에서 사용자를 추가할 때: `users.append(new_user)`
- 로그를 기록할 때: `logs.append(new_log)`
- 검색 결과를 모을 때: `results.append(item)`

## 3. insert() - 원하는 위치에 요소 추가하기

### 특정 위치에 끼워넣기

`insert()`는 리스트의 **원하는 위치**에 요소를 삽입합니다. `append()`는 무조건 끝에만 추가하지만, `insert()`는 중간이나 맨 앞에도 넣을 수 있습니다!

**문법:**
```python
리스트.insert(인덱스, 추가할_값)
```

**언제 사용하나요?**
- 우선순위가 높은 할 일을 맨 앞에 추가
- 정렬된 리스트에서 올바른 위치에 삽입
- 특정 순서를 유지하며 요소 추가

````python
# insert(인덱스, 값)는 특정 위치에 요소를 삽입합니다
colors = ["빨강", "파랑", "노랑"]
print("원본:", colors)  # 출력: 원본: ['빨강', '파랑', '노랑']

# 인덱스 1번 위치에 "초록" 삽입
# 기존 인덱스 1번 이후의 요소들은 뒤로 밀려남
colors.insert(1, "초록")
print("삽입 후:", colors)  # 출력: 삽입 후: ['빨강', '초록', '파랑', '노랑']

# 맨 앞에 삽입 (인덱스 0)
colors.insert(0, "보라")
print("맨 앞 삽입:", colors)  # 출력: 맨 앞 삽입: ['보라', '빨강', '초록', '파랑', '노랑']
````

**실행 결과:**
```
원본: ['빨강', '파랑', '노랑']
삽입 후: ['빨강', '초록', '파랑', '노랑']
맨 앞 삽입: ['보라', '빨강', '초록', '파랑', '노랑']
```

### 💡 동작 원리 이해하기

```python
원본:    ["빨강", "파랑", "노랑"]
인덱스:     0       1       2

insert(1, "초록") 실행 후:
결과:    ["빨강", "초록", "파랑", "노랑"]
인덱스:     0       1       2       3
         ↑       ↑(새로)  ↑       ↑
         그대로   삽입됨   밀림    밀림
```

- 지정한 인덱스 **이후의 모든 요소**가 뒤로 밀립니다
- 리스트 길이가 1 증가합니다

**주의사항:**
- 범위를 벗어난 인덱스를 주면? → 자동으로 맨 끝에 추가됩니다 (에러 X)
- 예: 길이 3인 리스트에 `insert(100, "값")` → 맨 끝에 추가됨

**실전 예시:**
```python
# 긴급 할 일을 맨 앞에!
todos.insert(0, "긴급: 회의 준비")

# 정렬된 점수 리스트에서 올바른 위치에 삽입
scores = [60, 70, 90]
scores.insert(2, 80)  # [60, 70, 80, 90]
```

## 4. 인덱스로 요소 수정하기

### 직접 값을 바꾸기

리스트는 인덱스를 사용해서 **기존 값을 직접 변경**할 수 있습니다. 이것이 가변성의 핵심입니다!

**문법:**
```python
리스트[인덱스] = 새로운_값
```

**언제 사용하나요?**
- 사용자 정보 수정 (이름, 이메일 변경)
- 점수 업데이트
- 상품 가격 변경
- 잘못 입력된 데이터 수정

````python
# 인덱스를 사용해 특정 위치의 값을 변경할 수 있습니다
scores = [85, 90, 75, 88]
print("원본 점수:", scores)  # 출력: 원본 점수: [85, 90, 75, 88]

# 두 번째 점수(인덱스 1) 수정
scores[1] = 95  # 90 → 95로 변경
print("수정 후:", scores)  # 출력: 수정 후: [85, 95, 75, 88]

# 마지막 점수 수정 (음수 인덱스 사용)
scores[-1] = 92  # 88 → 92로 변경
print("마지막 수정:", scores)  # 출력: 마지막 수정: [85, 95, 75, 92]
````

**실행 결과:**
```
원본 점수: [85, 90, 75, 88]
수정 후: [85, 95, 75, 88]
마지막 수정: [85, 95, 75, 92]
```

### 💡 핵심 포인트

- **기존 값이 사라지고** 새 값으로 완전히 교체됩니다
- 리스트 길이는 변하지 않습니다 (추가/삭제가 아니라 수정)
- 음수 인덱스(`-1`, `-2` 등)도 사용 가능합니다

**주의사항:**
```python
numbers = [1, 2, 3]
numbers[5] = 100  # ❌ IndexError!
# 존재하지 않는 인덱스는 수정 불가
```

**실전 활용:**
```python
# FastAPI에서 사용자 정보 수정
users = [{"id": 1, "name": "Alice"}, {"id": 2, "name": "Bob"}]
users[0] = {"id": 1, "name": "Alice Smith"}  # 이름 변경

# 상품 가격 업데이트
prices = [1000, 2000, 3000]
prices[1] = 2500  # 할인 가격 적용
```

## 5. del - 인덱스로 요소 삭제하기

### Python의 del 키워드

`del`은 Python의 **키워드**(메서드가 아님!)로, 리스트의 특정 인덱스에 있는 요소를 삭제합니다.

**문법:**
```python
del 리스트[인덱스]
```

**언제 사용하나요?**
- 특정 위치의 데이터를 제거할 때
- 인덱스를 알고 있을 때
- 빠른 삭제가 필요할 때

````python
# del을 사용하면 특정 인덱스의 요소를 삭제합니다
numbers = [10, 20, 30, 40, 50]
print("원본:", numbers)  # 출력: 원본: [10, 20, 30, 40, 50]

# 인덱스 2번 요소 삭제 (30이 삭제됨)
del numbers[2]
print("삭제 후:", numbers)  # 출력: 삭제 후: [10, 20, 40, 50]
# 인덱스 2 이후의 요소들이 앞으로 당겨짐
````

**실행 결과:**
```
원본: [10, 20, 30, 40, 50]
삭제 후: [10, 20, 40, 50]
```

### 💡 동작 원리

```python
원본:     [10, 20, 30, 40, 50]
인덱스:     0   1   2   3   4

del numbers[2] 실행 후:
결과:     [10, 20, 40, 50]
인덱스:     0   1   2   3
           ↑   ↑   ↑   ↑
          유지 유지 당김 당김
```

- 삭제된 위치 **이후의 모든 요소**가 앞으로 당겨집니다
- 리스트 길이가 1 감소합니다
- **삭제된 값은 복구할 수 없습니다**

**del vs remove() vs pop() 비교:**

| 방법 | 삭제 기준 | 반환값 | 용도 |
|------|----------|--------|------|
| `del` | 인덱스 | 없음 | 위치를 알 때 빠른 삭제 |
| `remove()` | 값 | 없음 | 값을 알 때 삭제 |
| `pop()` | 인덱스 | 삭제된 값 | 삭제하면서 값도 사용 |

**주의사항:**
```python
data = [1, 2, 3]
del data[10]  # ❌ IndexError!
# 존재하지 않는 인덱스는 삭제 불가
```

## 6. remove() - 값으로 요소 삭제하기

### 특정 값을 찾아서 삭제

`remove()`는 리스트에서 **특정 값**을 찾아서 삭제합니다. `del`은 인덱스로 삭제하지만, `remove()`는 값으로 삭제합니다!

**문법:**
```python
리스트.remove(삭제할_값)
```

**언제 사용하나요?**
- 인덱스는 모르지만 **값을 알 때**
- 특정 항목을 제거할 때 (예: 장바구니에서 "우유" 제거)
- 사용자가 선택한 항목 삭제

````python
# remove(값)는 리스트에서 특정 값을 찾아 삭제합니다
# 같은 값이 여러 개 있으면 첫 번째 것만 삭제됩니다
animals = ["고양이", "강아지", "토끼", "강아지"]
print("원본:", animals)  # 출력: 원본: ['고양이', '강아지', '토끼', '강아지']

animals.remove("강아지")  # 첫 번째 "강아지"만 삭제됨
print("삭제 후:", animals)  # 출력: 삭제 후: ['고양이', '토끼', '강아지']
````

**실행 결과:**
```
원본: ['고양이', '강아지', '토끼', '강아지']
삭제 후: ['고양이', '토끼', '강아지']
```

### 💡 중요한 특징

**1. 첫 번째 매칭만 삭제**
```python
numbers = [1, 2, 3, 2, 4, 2]
numbers.remove(2)  # 첫 번째 2만 삭제
print(numbers)  # [1, 3, 2, 4, 2]
```

**2. 값이 없으면 에러 발생**
```python
fruits = ["사과", "바나나"]
fruits.remove("포도")  # ❌ ValueError: list.remove(x): x not in list
```

**안전하게 사용하기:**
```python
# 값이 있는지 먼저 확인
if "포도" in fruits:
    fruits.remove("포도")
else:
    print("포도가 없습니다")
```

**실전 활용:**
```python
# 장바구니에서 상품 제거
cart = ["우유", "빵", "계란"]
cart.remove("빵")  # 사용자가 "빵" 삭제 요청

# 차단 목록에서 해제
blocked_users = ["user1", "user2", "user3"]
blocked_users.remove("user2")  # user2 차단 해제
```

## 7. pop() - 삭제하면서 값도 가져오기

### 가장 강력한 삭제 메서드!

`pop()`은 요소를 **삭제하면서 동시에 그 값을 반환**합니다. 삭제된 값을 사용해야 할 때 매우 유용합니다!

**문법:**
```python
삭제된_값 = 리스트.pop()        # 마지막 요소 제거
삭제된_값 = 리스트.pop(인덱스)  # 특정 위치 제거
```

**언제 사용하나요?**
- 스택(Stack) 구현 (마지막에 넣은 것 먼저 꺼내기)
- 대기열(Queue) 구현
- 삭제한 값을 로그에 기록
- 작업 완료 처리하며 결과 저장

````python
# pop()은 요소를 삭제하면서 그 값을 반환합니다
# 인자 없으면 마지막 요소, 인덱스 주면 해당 위치 요소 삭제

tasks = ["청소", "빨래", "설거지", "장보기"]
print("할 일:", tasks)  # 출력: 할 일: ['청소', '빨래', '설거지', '장보기']

# 마지막 할 일 꺼내기
last_task = tasks.pop()  # "장보기"가 삭제되고 반환됨
print(f"완료: {last_task}")  # 출력: 완료: 장보기
print("남은 일:", tasks)  # 출력: 남은 일: ['청소', '빨래', '설거지']

# 첫 번째 할 일 꺼내기
first_task = tasks.pop(0)  # "청소"가 삭제되고 반환됨
print(f"완료: {first_task}")  # 출력: 완료: 청소
print("남은 일:", tasks)  # 출력: 남은 일: ['빨래', '설거지']
````

**실행 결과:**
```
할 일: ['청소', '빨래', '설거지', '장보기']
완료: 장보기
남은 일: ['청소', '빨래', '설거지']
완료: 청소
남은 일: ['빨래', '설거지']
```

### 💡 pop()의 특별한 점

**1. 반환값이 있습니다**
```python
numbers = [1, 2, 3]
removed = numbers.pop()  # removed = 3
print(f"삭제된 값: {removed}")  # 삭제된 값: 3
```

**2. 인덱스 생략 시 마지막 요소 제거**
```python
stack = ["A", "B", "C"]
stack.pop()  # "C" 제거 (LIFO - Last In First Out)
```

**3. 인덱스 지정 가능**
```python
queue = ["손님1", "손님2", "손님3"]
queue.pop(0)  # "손님1" 제거 (FIFO - First In First Out)
```

**del vs pop() 비교:**

```python
# del: 값을 사용하지 않을 때
del numbers[0]  # 그냥 삭제만

# pop(): 삭제된 값을 사용할 때
removed = numbers.pop(0)  # 삭제하면서 값도 저장
print(f"로그: {removed} 삭제됨")
```

**실전 활용:**
```python
# 작업 큐 (Queue)
jobs = ["작업1", "작업2", "작업3"]
current_job = jobs.pop(0)  # 첫 번째 작업 처리
print(f"처리 중: {current_job}")

# 되돌리기 기능 (Undo)
history = ["액션1", "액션2", "액션3"]
last_action = history.pop()  # 마지막 액션 취소
print(f"취소됨: {last_action}")
```

**주의사항:**
```python
empty_list = []
empty_list.pop()  # ❌ IndexError: pop from empty list
# 빈 리스트에서 pop() 불가!
```

## 8. 실전 예제: 재고 관리 시스템

### 배운 내용을 모두 활용해봅시다!

이제 지금까지 배운 **모든 메서드**를 활용해서 실제로 동작하는 재고 관리 시스템을 만들어봅시다. 이 예제는 실무에서 자주 만나는 패턴입니다.

**사용할 메서드:**
- `append()`: 신규 상품 입고
- `insert()`: 긴급 입고
- `pop()`: 판매 완료 처리
- `remove()`: 단종 상품 제거

````python
# 상품 재고 리스트
inventory = ["노트북", "마우스", "키보드"]
print("초기 재고:", inventory)  # 출력: 초기 재고: ['노트북', '마우스', '키보드']

# 신규 상품 입고
inventory.append("모니터")
inventory.append("헤드셋")
print("입고 후:", inventory)  
# 출력: 입고 후: ['노트북', '마우스', '키보드', '모니터', '헤드셋']

# 긴급으로 맨 앞에 추가 (인기 상품)
inventory.insert(0, "웹캠")
print("긴급 입고:", inventory)  
# 출력: 긴급 입고: ['웹캠', '노트북', '마우스', '키보드', '모니터', '헤드셋']

# 판매 완료 (pop으로 제거하며 기록)
sold_item = inventory.pop(1)  # "노트북"이 판매됨
print(f"판매 완료: {sold_item}")  # 출력: 판매 완료: 노트북
print("현재 재고:", inventory)  
# 출력: 현재 재고: ['웹캠', '마우스', '키보드', '모니터', '헤드셋']

# 단종 상품 제거
inventory.remove("웹캠")  # 단종 결정
print("최종 재고:", inventory)  
# 출력: 최종 재고: ['마우스', '키보드', '모니터', '헤드셋']
````

**실행 결과:**
```
초기 재고: ['노트북', '마우스', '키보드']
입고 후: ['노트북', '마우스', '키보드', '모니터', '헤드셋']
긴급 입고: ['웹캠', '노트북', '마우스', '키보드', '모니터', '헤드셋']
판매 완료: 노트북
현재 재고: ['웹캠', '마우스', '키보드', '모니터', '헤드셋']
최종 재고: ['마우스', '키보드', '모니터', '헤드셋']
```

### 💡 실전 분석

이 예제가 실무와 어떻게 연결되는지 봅시다:

**1. 입고 처리 (append)**
```python
# FastAPI 엔드포인트
@app.post("/inventory")
def add_product(product: str):
    inventory.append(product)  # ← 리스트에 추가!
    return {"message": f"{product} 입고 완료"}
```

**2. 긴급 처리 (insert)**
```python
# 우선순위 큐
tasks.insert(0, "긴급 작업")  # 맨 앞에 추가
```

**3. 판매 기록 (pop)**
```python
# 판매하면서 로그 저장
sold = inventory.pop(index)
sales_log.append({"item": sold, "time": now()})
```

**4. 단종 처리 (remove)**
```python
# 관리자가 특정 상품 단종
@app.delete("/inventory/{product}")
def discontinue(product: str):
    inventory.remove(product)
    return {"message": f"{product} 단종됨"}
```

**왜 이런 메서드들이 중요한가요?**

실무에서 데이터는 **항상 변합니다**:
- 사용자가 추가/삭제/수정을 요청합니다
- 서버는 이를 **즉시** 처리해야 합니다
- 리스트 메서드가 없다면? 매번 새 리스트를 만들어야 합니다 (비효율적!)

리스트의 가변성 덕분에 효율적으로 데이터를 관리할 수 있습니다! 🚀

## 📌 핵심 정리

이번 시간에 배운 내용:

1. **리스트 가변성(Mutable)**: 리스트는 생성 후에도 내용을 자유롭게 수정할 수 있습니다
2. **append(value)**: 리스트 끝에 요소를 추가합니다 (가장 많이 사용!)
3. **insert(index, value)**: 특정 위치에 요소를 삽입합니다
4. **list[index] = value**: 인덱스로 직접 값을 수정합니다
5. **del list[index]**: 인덱스로 요소를 삭제합니다 (반환값 없음)
6. **remove(value)**: 값으로 요소를 찾아 삭제합니다 (첫 번째만!)
7. **pop(index)**: 요소를 삭제하면서 그 값을 반환합니다 (가장 강력!)

### 메서드 선택 가이드

| 상황 | 사용할 메서드 | 예시 |
|------|--------------|------|
| 끝에 추가 | `append()` | `todos.append("새 할 일")` |
| 특정 위치에 추가 | `insert()` | `tasks.insert(0, "긴급")` |
| 값 수정 | `list[index] = value` | `scores[0] = 100` |
| 인덱스로 삭제 | `del` 또는 `pop()` | `del items[2]` |
| 값으로 삭제 | `remove()` | `cart.remove("우유")` |
| 삭제하면서 값 사용 | `pop()` | `item = queue.pop(0)` |
