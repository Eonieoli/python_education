## 📖 while문이란 무엇인가?

while문은 **조건이 True인 동안 계속 반복**하는 반복문입니다. for문이 "정해진 횟수만큼" 반복한다면, while문은 "조건이 만족될 때까지" 반복합니다.

### 왜 while문이 필요할까요?

실생활에서 생각해봅시다:

- 🔐 **로그인 시스템**: 비밀번호가 맞을 때까지 계속 입력받기
- 🎮 **게임**: 사용자가 '종료'를 선택할 때까지 계속 진행
- 📱 **메뉴 시스템**: 사용자가 '나가기'를 선택할 때까지 메뉴 표시
- 💬 **챗봇**: 사용자가 '종료'라고 말할 때까지 대화 계속

이런 경우들은 **반복 횟수를 미리 알 수 없습니다**. "언제 끝날지 모르지만, 특정 조건이 만족되면 끝난다" - 이것이 바로 while문이 필요한 이유입니다!

### for문 vs while문 - 언제 무엇을 쓸까?

| 상황 | 사용할 반복문 | 예시 |
|------|--------------|------|
| 반복 횟수를 정확히 알 때 | **for문** | 학생 10명의 점수 처리, 1~100까지 합계 |
| 반복 횟수를 모를 때 | **while문** | 로그인 성공까지, 정답 맞출 때까지 |
| 리스트/딕셔너리 순회 | **for문** | 상품 목록 출력, 사용자 정보 처리 |
| 조건 만족까지 반복 | **while문** | 입력값 검증, 게임 진행 |

### FastAPI/웹 개발에서의 활용

실무에서 while문은 이렇게 사용됩니다:

```python
# 1. 재시도 로직 (네트워크 요청)
retry_count = 0
max_retries = 3
while retry_count < max_retries:
    try:
        response = requests.get(api_url)
        if response.status_code == 200:
            break  # 성공하면 종료
    except:
        retry_count += 1
        time.sleep(1)  # 1초 대기 후 재시도

# 2. 데이터베이스 연결 재시도
connected = False
while not connected:
    try:
        db.connect()
        connected = True
    except ConnectionError:
        print("재연결 시도 중...")
        time.sleep(5)

# 3. 페이지네이션 (모든 데이터 가져오기)
page = 1
all_users = []
while True:
    users = get_users(page=page, limit=100)
    if not users:  # 더 이상 데이터가 없으면
        break
    all_users.extend(users)
    page += 1
```

### ⚠️ 주의사항: 무한 루프

while문을 사용할 때 가장 주의해야 할 것은 **무한 루프(Infinite Loop)**입니다:

```python
# 위험! 영원히 반복됨 ⚠️
count = 1
while count <= 5:
    print(count)
    # count += 1 을 깜빡했어요! 😱
```

**해결 방법**:
1. 조건을 변경하는 코드를 반드시 넣기 (`count += 1`)
2. 또는 `break`로 빠져나올 방법 준비하기
3. 최대 반복 횟수 제한하기

```python
# 안전한 패턴 ✅
max_attempts = 100
attempts = 0
while True:
    attempts += 1
    if attempts >= max_attempts:
        print("최대 시도 횟수 초과")
        break
    # 작업 수행...
```

### 핵심 포인트 정리

> 💡 **while문의 핵심**:
> 1. 조건이 True인 동안 계속 반복
> 2. 반복 횟수를 모를 때 사용
> 3. 반드시 조건이 False가 될 수 있어야 함 (무한루프 방지)
> 4. break로 즉시 종료, continue로 건너뛰기 가능

자, 이제 실제 코드로 while문을 익혀봅시다! 🚀

## 1. while문 기본 - 조건이 True인 동안 반복하기

while문의 가장 기본적인 형태를 배워봅시다. while문은 `while 조건:` 형태로 시작하며, 조건이 True인 동안 들여쓰기된 코드를 계속 실행합니다.

### 문법 구조

```python
while 조건:
    # 조건이 True일 때 실행되는 코드
    # 반드시 조건을 변경하는 코드가 있어야 함!
```

**실행 흐름**:
1. 조건을 확인합니다
2. 조건이 True면 코드 블록을 실행합니다
3. 다시 1번으로 돌아갑니다
4. 조건이 False가 되면 while문을 빠져나옵니다

````python
# 기본 예제: 1부터 5까지 출력하기
count = 1  # 시작값 설정

while count <= 5:  # count가 5 이하인 동안 반복
    print(f"{count}번째 반복")  # 현재 count 출력
    count += 1  # ⭐ 중요! count를 1 증가시켜야 언젠가 조건이 False가 됨

# 실행 과정:
# count=1: 1<=5 (True) → "1번째 반복" 출력 → count=2
# count=2: 2<=5 (True) → "2번째 반복" 출력 → count=3
# count=3: 3<=5 (True) → "3번째 반복" 출력 → count=4
# count=4: 4<=5 (True) → "4번째 반복" 출력 → count=5
# count=5: 5<=5 (True) → "5번째 반복" 출력 → count=6
# count=6: 6<=5 (False) → 반복 종료!
````

**예상 출력:**
```
1번째 반복
2번째 반복
3번째 반복
4번째 반복
5번째 반복
```

### 💡 핵심 포인트

- `count += 1`이 없으면 count는 계속 1이고, 조건은 영원히 True → **무한 루프**!
- while문 안에서 조건을 변경하는 코드는 **필수**입니다
- 조건이 처음부터 False면 한 번도 실행되지 않습니다

## 2. for문과 while문 비교

같은 작업을 for문과 while문으로 각각 구현하면서 차이점을 명확히 이해해봅시다.

````python
# for문: 반복 횟수를 알 때 (더 간결!)
print("=== for문으로 1~5 출력 ===")
for i in range(1, 6):  # range(1, 6)은 1, 2, 3, 4, 5를 생성
    print(f"for: {i}")

print()  # 빈 줄 출력

# while문: 조건이 만족될 때까지
print("=== while문으로 1~5 출력 ===")
num = 1  # 시작값 직접 설정
while num <= 5:  # 조건 직접 작성
    print(f"while: {num}")
    num += 1  # 증가 코드 직접 작성
````

**예상 출력:**
```
=== for문으로 1~5 출력 ===
for: 1
for: 2
for: 3
for: 4
for: 5

=== while문으로 1~5 출력 ===
while: 1
while: 2
while: 3
while: 4
while: 5
```

### 비교 분석

| 특징 | for문 | while문 |
|------|-------|--------|
| 코드 길이 | 짧고 간결 | 조금 더 김 |
| 반복 횟수 | 명확함 (range로 정의) | 조건에 따라 다름 |
| 사용 시기 | 횟수를 알 때 | 조건이 중요할 때 |
| 실수 가능성 | 낮음 | 높음 (무한루프 가능) |

**결론**: 반복 횟수를 알면 for문이 더 안전하고 간결합니다. 하지만 "조건이 만족될 때까지"가 중요한 경우에는 while문이 더 자연스럽습니다!

## 3. break - 반복문 즉시 종료하기

`break`는 반복문을 **즉시 종료**하는 명령어입니다. 조건과 상관없이 break를 만나면 while문을 빠져나옵니다.

### 언제 사용하나요?

- 원하는 값을 찾았을 때
- 에러가 발생했을 때
- 최대 시도 횟수에 도달했을 때
- 사용자가 "종료"를 선택했을 때

### 무한 루프 + break 패턴

`while True:`와 `break`를 조합하면 "조건이 만족될 때까지" 반복하는 안전한 패턴을 만들 수 있습니다.

````python
print("=== break 예제 ===")

# 시나리오: 5번 시도하면 종료하는 시스템
count = 0

while True:  # 무한 루프! (조건이 항상 True)
    count += 1  # 시도 횟수 증가
    print(f"{count}번째 시도")
    
    # 5번 시도했으면 종료
    if count >= 5:
        print("5번 시도 완료!")
        break  # ⭐ while문을 즉시 종료!
    
# break를 만나면 이 줄로 바로 이동합니다
print("while문 종료됨")
````

**예상 출력:**
```
=== break 예제 ===
1번째 시도
2번째 시도
3번째 시도
4번째 시도
5번째 시도
5번 시도 완료!
while문 종료됨
```

### 💡 break 사용 팁

```python
# ✅ 좋은 패턴: while True + break
while True:
    user_input = input("종료하려면 'quit' 입력: ")
    if user_input == 'quit':
        break
    # 작업 수행...

# ❌ 복잡한 패턴 (권장하지 않음)
quit_flag = False
while not quit_flag:
    user_input = input("종료하려면 'quit' 입력: ")
    if user_input == 'quit':
        quit_flag = True
    # 작업 수행...
```

첫 번째 방법이 더 간결하고 읽기 쉽습니다!

## 4. continue - 다음 반복으로 건너뛰기

`continue`는 **현재 반복을 건너뛰고** 다음 반복으로 바로 이동합니다. break와 다르게 반복문을 종료하지는 않습니다.

### 언제 사용하나요?

- 특정 조건일 때 처리를 건너뛰고 싶을 때
- 예외 상황을 먼저 처리하고 싶을 때
- 필터링이 필요할 때 (홀수만, 짝수만 등)

````python
print("=== continue 예제: 홀수만 출력하기 ===")

num = 0  # 시작값

while num < 10:  # 0부터 9까지
    num += 1  # ⚠️ 주의: continue 전에 num을 증가시켜야 함!
    
    # 짝수인 경우 건너뛰기
    if num % 2 == 0:  # num을 2로 나눈 나머지가 0이면 짝수
        continue  # ⭐ 아래 코드를 실행하지 않고 다음 반복으로!
    
    # 이 줄은 홀수일 때만 실행됨
    print(num)
````

**예상 출력:**
```
=== continue 예제: 홀수만 출력하기 ===
1
3
5
7
9
```

### 실행 과정 상세 분석

```python
num=0 → num=1: 1%2=1 (홀수) → print(1)
num=1 → num=2: 2%2=0 (짝수) → continue (print 건너뜀)
num=2 → num=3: 3%2=1 (홀수) → print(3)
num=3 → num=4: 4%2=0 (짝수) → continue
...
```

### ⚠️ 중요한 주의사항!

```python
# ❌ 잘못된 코드 - 무한 루프!
num = 0
while num < 10:
    if num % 2 == 0:
        continue  # num을 증가시키기 전에 continue!
    num += 1  # 짝수일 때는 여기까지 오지 못함 → 무한 루프!
    print(num)

# ✅ 올바른 코드
num = 0
while num < 10:
    num += 1  # continue 전에 먼저 증가!
    if num % 2 == 0:
        continue
    print(num)
```

**핵심**: continue를 사용할 때는 반드시 **조건을 변경하는 코드**가 continue 이전에 있어야 합니다!

## 5. 실전 예제: 로그인 시스템 (재시도 로직)

이제 배운 내용을 모두 활용해서 실무에서 자주 사용하는 **재시도 로직**을 구현해봅시다. 이것은 FastAPI를 포함한 모든 웹 개발에서 핵심적인 패턴입니다!

### 시나리오

- 사용자가 비밀번호를 입력합니다
- 최대 3번까지 시도할 수 있습니다
- 비밀번호가 맞으면 로그인 성공
- 3번 모두 틀리면 로그인 실패

### 실무 활용

이 패턴은 다음과 같은 곳에서 사용됩니다:
- 🔐 로그인/인증 시스템
- 🌐 네트워크 요청 재시도
- 💾 데이터베이스 연결 재시도
- 📧 이메일 전송 재시도

````python
print("=== 실전 예제: 로그인 시스템 ===")

# 설정
correct_password = "1234"  # 정답 비밀번호
max_attempts = 3  # 최대 시도 횟수

# 로그인 시도
attempts = 0  # 현재까지 시도한 횟수

while attempts < max_attempts:  # 최대 시도 횟수 이내인 동안
    # 비밀번호 입력 (실제로는 input() 사용)
    # 시연을 위해 하드코딩된 값 사용
    if attempts == 0:
        password = "0000"  # 첫 번째 시도: 틀린 비밀번호
    elif attempts == 1:
        password = "9999"  # 두 번째 시도: 틀린 비밀번호
    else:
        password = "1234"  # 세 번째 시도: 맞는 비밀번호
    
    attempts += 1  # 시도 횟수 증가 (중요!)
    print(f"\n{attempts}번째 시도: {password}")
    
    # 비밀번호 검증
    if password == correct_password:
        print("✅ 로그인 성공!")
        break  # 성공했으므로 반복문 종료
    else:
        # 비밀번호가 틀렸을 때
        remaining = max_attempts - attempts  # 남은 시도 횟수 계산
        
        if remaining > 0:
            # 아직 기회가 남았을 때
            print(f"❌ 비밀번호가 틀렸습니다. (남은 횟수: {remaining}회)")
        else:
            # 마지막 시도까지 실패했을 때
            print("❌ 로그인 실패! 횟수 초과")
````

**예상 출력:**
```
=== 실전 예제: 로그인 시스템 ===

1번째 시도: 0000
❌ 비밀번호가 틀렸습니다. (남은 횟수: 2회)

2번째 시도: 9999
❌ 비밀번호가 틀렸습니다. (남은 횟수: 1회)

3번째 시도: 1234
✅ 로그인 성공!
```

### 코드 분석: 핵심 포인트

#### 1. 최대 시도 횟수 제한
```python
while attempts < max_attempts:
```
- 무한 시도를 방지합니다
- 보안을 위해 필수적입니다 (무차별 대입 공격 방지)

#### 2. 성공 시 즉시 종료
```python
if password == correct_password:
    print("✅ 로그인 성공!")
    break  # 더 이상 시도할 필요 없음
```
- 조건을 만족하면 바로 종료
- 불필요한 반복을 줄입니다

#### 3. 사용자 친화적인 피드백
```python
remaining = max_attempts - attempts
print(f"남은 횟수: {remaining}회")
```
- 사용자에게 남은 기회를 알려줍니다
- UX(사용자 경험) 향상

### 🔒 실무 응용: FastAPI 로그인

```python
from fastapi import HTTPException

@app.post("/login")
def login(username: str, password: str):
    # 데이터베이스에서 사용자 조회
    user = db.get_user(username)
    
    # 로그인 시도 횟수 확인
    if user.failed_attempts >= 3:
        raise HTTPException(status_code=429, detail="계정이 잠겼습니다")
    
    # 비밀번호 검증
    if not verify_password(password, user.hashed_password):
        user.failed_attempts += 1  # 실패 횟수 증가
        db.update_user(user)
        raise HTTPException(status_code=401, detail="비밀번호가 틀렸습니다")
    
    # 로그인 성공
    user.failed_attempts = 0  # 카운터 리셋
    return {"token": create_token(user)}
```

이처럼 while문의 재시도 로직 패턴은 실무에서 매우 자주 사용됩니다!

3. **break**: 반복문을 즉시 종료 (조건과 무관)
4. **continue**: 현재 반복을 건너뛰고 다음 반복으로
5. **재시도 로직**: 실무에서 가장 많이 사용하는 while문 패턴

```python
# 기존 방식
numbers = []
for i in range(10):
    numbers.append(i * 2)

# 리스트 컴프리헨션 (한 줄!)
numbers = [i * 2 for i in range(10)]
```
