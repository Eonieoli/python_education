## 📖 Boolean(불리언) 타입이란 무엇인가?

### True와 False, 단 두 가지 값

Boolean(불리언)은 **True(참)** 또는 **False(거짓)** 단 두 가지 값만 가지는 데이터 타입입니다. 이름은 수학자 조지 불(George Boole)에서 유래했어요.

일상생활에서 우리는 항상 "예/아니오" 질문에 답합니다:
- 로그인 되었나요? → 예 (True) / 아니오 (False)
- 성인인가요? → 예 (True) / 아니오 (False)
- 재고가 있나요? → 예 (True) / 아니오 (False)

프로그램도 마찬가지입니다! Boolean 타입은 **프로그램이 의사결정을 내리고 다른 경로로 분기하는 데** 사용됩니다.

### 왜 Boolean이 필요한가?

프로그램은 상황에 따라 다르게 동작해야 합니다:

```python
# 로그인 체크
if 아이디와_비밀번호가_맞다:  # True면
    메인_페이지_보여주기()
else:  # False면
    에러_메시지_보여주기()
```

이처럼 조건에 따라 프로그램의 흐름을 제어하려면 "참/거짓"을 판단할 수 있어야 합니다. 이것이 바로 Boolean의 역할입니다! 🎯

### FastAPI에서는?

FastAPI로 API를 만들 때도 Boolean은 필수입니다:

```python
# 사용자 권한 체크
if user.is_admin:  # Boolean 값
    # 관리자만 접근 가능한 기능
    return admin_dashboard()
else:
    # 권한 없음 에러
    raise HTTPException(status_code=403)
```

거의 모든 로직이 Boolean 판단에서 시작됩니다! 💡

## 1. Boolean 타입 기초

### Boolean 값 사용하기

Python에서 Boolean 타입은 `True`와 `False` 두 가지 값을 가집니다. **첫 글자가 대문자**라는 점에 주의하세요!

````python
# Boolean 값 할당
is_logged_in = True   # 로그인 됨
is_adult = False      # 미성년자

print(f"로그인 상태: {is_logged_in}")  # True
print(f"성인 여부: {is_adult}")        # False
print(f"type(is_logged_in) = {type(is_logged_in)}")  # <class 'bool'>

print("\n왜 Boolean이 필요한가?")
print("→ 프로그램이 '예/아니오' 질문에 답하고 분기하기 위해")
print("→ 실생활 예시:")
print("  - 로그인 검증: 아이디와 비밀번호가 맞는가?")
print("  - 나이 확인: 성인인가?")
print("  - 재고 확인: 상품이 있는가?")
print("  - 할인 대상: 조건을 만족하는가?")
````

**실행 결과:**
```
로그인 상태: True
성인 여부: False
type(is_logged_in) = <class 'bool'>

왜 Boolean이 필요한가?
→ 프로그램이 '예/아니오' 질문에 답하고 분기하기 위해
→ 실생활 예시:
  - 로그인 검증: 아이디와 비밀번호가 맞는가?
  - 나이 확인: 성인인가?
  - 재고 확인: 상품이 있는가?
  - 할인 대상: 조건을 만족하는가?
```

### 💡 알아두기

- Boolean 값은 `True`, `False` 정확히 대문자로 시작해야 합니다
- `type()` 함수로 확인하면 `<class 'bool'>`이 나옵니다
- Boolean 변수명은 보통 `is_`, `has_`, `can_` 등으로 시작하는 것이 관례입니다
  - 좋은 예: `is_admin`, `has_permission`, `can_delete`
  - 나쁜 예: `admin`, `permission`, `delete` (Boolean인지 알기 어려움)

## 📖 비교 연산자(Comparison Operators)란?

### 두 값을 비교해서 True/False 반환

비교 연산자는 **두 개의 값을 비교**해서 그 결과를 **Boolean 값(True/False)**으로 반환하는 연산자입니다.

### 비교 연산자의 종류

| 연산자 | 의미 | 예시 | 결과 |
|--------|------|------|------|
| `==` | 같다 | `5 == 5` | `True` |
| `!=` | 다르다 | `5 != 3` | `True` |
| `<` | 작다 | `3 < 5` | `True` |
| `>` | 크다 | `5 > 3` | `True` |
| `<=` | 작거나 같다 | `5 <= 5` | `True` |
| `>=` | 크거나 같다 | `5 >= 3` | `True` |

### ⚠️ 주의: = vs ==

- `=`: **할당 연산자** (변수에 값을 저장)
  ```python
  age = 25  # age 변수에 25를 저장
  ```

- `==`: **비교 연산자** (두 값이 같은지 비교)
  ```python
  age == 25  # age가 25와 같은가? → True/False
  ```

초보자가 가장 많이 하는 실수가 이 둘을 헷갈리는 것입니다! 🚨

### 실무에서는?

비교 연산자는 **조건을 체크**하는 모든 곳에서 사용됩니다:

```python
# 재고 확인
if stock > 0:
    print("구매 가능")

# 나이 제한
if age >= 19:
    print("성인 인증 완료")

# 비밀번호 확인
if input_password == correct_password:
    print("로그인 성공")
```

## 2. 비교 연산자 - 숫자 비교

### 숫자를 비교하여 Boolean 값 얻기

가장 기본적인 사용법은 숫자를 비교하는 것입니다. 나이, 가격, 점수 등을 비교할 때 사용합니다.

````python
# 숫자 비교
age = 25
print(f"나이: {age}")
print(f"age == 25: {age == 25}")  # True (같음)
print(f"age != 20: {age != 20}")  # True (다름)
print(f"age > 20: {age > 20}")    # True (큼)
print(f"age < 30: {age < 30}")    # True (작음)
print(f"age >= 25: {age >= 25}")  # True (크거나 같음)
print(f"age <= 20: {age <= 20}")  # False (작거나 같음)
````

**실행 결과:**
```
나이: 25
age == 25: True
age != 20: True
age > 20: True
age < 30: True
age >= 25: True
age <= 20: False
```

### 🔍 결과 해석

- `age == 25`: age가 25와 **정확히 같으므로** True
- `age != 20`: age가 20과 **다르므로** True
- `age > 20`: 25는 20보다 **크므로** True
- `age < 30`: 25는 30보다 **작으므로** True
- `age >= 25`: 25는 25보다 크거나 **같으므로** True (같아도 True!)
- `age <= 20`: 25는 20보다 크므로 "작거나 같다"는 **거짓**, 따라서 False

### 💡 실전 팁

- `>=`, `<=`는 **같을 때도 True**를 반환합니다 ("이상", "이하")
- `>`, `<`는 **같을 때는 False**를 반환합니다 ("초과", "미만")
- 한국 나이 계산 시 주의: 만 19세 **이상**은 `age >= 19`입니다!

## 3. 비교 연산자 - 문자열 비교

### 문자열도 비교할 수 있습니다!

문자열 비교는 주로 **로그인 검증**, **권한 체크** 등에서 사용됩니다.

````python
# 문자열 비교
username = "admin"
print(f"아이디: {username}")
print(f'username == "admin": {username == "admin"}')    # True
print(f'username == "user": {username == "user"}')      # False
print(f'username != "guest": {username != "guest"}')    # True

# 문자열은 사전순(알파벳순)으로 비교됨
print(f"\n문자열 크기 비교:")
print(f'"apple" < "banana": {"apple" < "banana"}')  # True
print(f'"zebra" > "apple": {"zebra" > "apple"}')    # True
````

**실행 결과:**
```
아이디: admin
username == "admin": True
username == "user": False
username != "guest": True

문자열 크기 비교:
"apple" < "banana": True
"zebra" > "apple": True
```

### 🔍 문자열 비교의 원리

1. **`==`, `!=`**: 문자가 완전히 같은지 다른지 비교
   - `"admin" == "admin"` → True (완전히 같음)
   - `"admin" == "Admin"` → False (대소문자 다름!)

2. **`<`, `>`, `<=`, `>=`**: 사전순(알파벳순) 비교
   - `"apple"` < `"banana"` → True (사전에서 apple이 먼저 나옴)
   - `"zebra"` > `"apple"` → True (사전에서 zebra가 나중에 나옴)

### ⚠️ 주의사항

- 문자열 비교는 **대소문자를 구분**합니다!
  ```python
  "Admin" == "admin"  # False!
  ```

- 대소문자를 무시하고 비교하려면 `lower()` 사용:
  ```python
  username.lower() == "admin".lower()  # 둘 다 소문자로 변환 후 비교
  ```

    # 저장된 정보와 비교

    else:

## 📖 논리 연산자(Logical Operators)란?

### 여러 조건을 조합하기

논리 연산자는 **여러 개의 Boolean 값을 조합**해서 하나의 Boolean 결과를 만듭니다.

실생활에서도 우리는 여러 조건을 동시에 체크합니다:
- "아이디**도** 맞**고** 비밀번호**도** 맞으면" 로그인 성공 → `and`
- "회원이**거나** 경로우대 대상이면" 할인 적용 → `or`
- "재고가 **없으면**" 품절 표시 → `not`

### 논리 연산자의 종류

| 연산자 | 의미 | 예시 | 언제 True? |
|--------|------|------|------------|
| `and` | 그리고 (둘 다) | `True and True` | **둘 다** True일 때 |
| `or` | 또는 (하나라도) | `True or False` | **하나라도** True일 때 |
| `not` | 아니다 (반대) | `not False` | Boolean 값을 **반대로** |

### and 연산자 진리표

| A | B | A and B |
|---|---|--------|
| True | True | **True** |
| True | False | False |
| False | True | False |
| False | False | False |

→ **둘 다 True여야** True!

### or 연산자 진리표

| A | B | A or B |
|---|---|--------|
| True | True | **True** |
| True | False | **True** |
| False | True | **True** |
| False | False | False |

→ **하나라도 True면** True!

### not 연산자

| A | not A |
|---|-------|
| True | **False** |
| False | **True** |

→ **반대로** 뒤집기!

## 4. 논리 연산자 - and (그리고)

### 둘 다 True여야 True

`and` 연산자는 **모든 조건이 참**이어야 최종 결과가 True입니다. 하나라도 False면 전체가 False가 됩니다.

로그인 시스템을 생각해보세요: 아이디**도** 맞아야 하고, 비밀번호**도** 맞아야 합니다!

````python
# and: 둘 다 True여야 True
print("[ and 연산자 - 둘 다 참이어야 참 ]")
has_id = True
has_password = True
print(f"아이디 있음: {has_id}, 비밀번호 있음: {has_password}")
print(f"로그인 가능: {has_id and has_password}")  # True

has_id = True
has_password = False
print(f"\n아이디 있음: {has_id}, 비밀번호 있음: {has_password}")
print(f"로그인 가능: {has_id and has_password}")  # False
````

**실행 결과:**
```
[ and 연산자 - 둘 다 참이어야 참 ]
아이디 있음: True, 비밀번호 있음: True
로그인 가능: True

아이디 있음: True, 비밀번호 있음: False
로그인 가능: False
```

### 🔍 and 연산의 원리

- 첫 번째 케이스: `True and True` → **True** (둘 다 참이므로)
- 두 번째 케이스: `True and False` → **False** (하나가 거짓이므로)

### 💡 실전 활용

```python
# 성인이면서 회원일 때만 구매 가능
if age >= 19 and is_member:
    print("구매 가능")

# 재고가 있고, 판매중이고, 할인 기간일 때
if stock > 0 and is_active and is_discount_period:
    print("할인가로 구매 가능")
```

**모든 조건**을 만족해야 할 때 `and`를 사용합니다!

## 5. 논리 연산자 - or (또는)

### 하나라도 True면 True

`or` 연산자는 **하나라도 참**이면 최종 결과가 True입니다. 모두 False일 때만 False가 됩니다.

할인 혜택을 생각해보세요: 회원이**거나** 경로우대 대상이면 할인을 받을 수 있습니다!

````python
# or: 하나라도 True면 True
print("[ or 연산자 - 하나라도 참이면 참 ]")
is_member = False
is_senior = True
print(f"회원: {is_member}, 경로우대: {is_senior}")
print(f"할인 대상: {is_member or is_senior}")  # True

is_member = False
is_senior = False
print(f"\n회원: {is_member}, 경로우대: {is_senior}")
print(f"할인 대상: {is_member or is_senior}")  # False
````

**실행 결과:**
```
[ or 연산자 - 하나라도 참이면 참 ]
회원: False, 경로우대: True
할인 대상: True

회원: False, 경로우대: False
할인 대상: False
```

### 🔍 or 연산의 원리

- 첫 번째 케이스: `False or True` → **True** (경로우대가 참이므로)
- 두 번째 케이스: `False or False` → **False** (둘 다 거짓이므로)

### 💡 실전 활용

```python
# VIP이거나 프리미엄 회원이면 접근 가능
if is_vip or is_premium:
    print("특별 콘텐츠 접근 가능")

# 주말이거나 공휴일이면 배송 불가
if is_weekend or is_holiday:
    print("평일에만 배송 가능")
```

**하나만 만족해도 될 때** `or`를 사용합니다!

## 6. 논리 연산자 - not (반대)

### Boolean 값을 반대로 뒤집기

`not` 연산자는 Boolean 값을 **반대로** 바꿉니다. True는 False로, False는 True로!

````python
# not: 반대로 뒤집기
print("[ not 연산자 - 반대로 ]")
is_available = True
print(f"재고 있음: {is_available}")
print(f"재고 없음: {not is_available}")  # False

is_available = False
print(f"\n재고 있음: {is_available}")
print(f"재고 없음: {not is_available}")  # True
````

**실행 결과:**
```
[ not 연산자 - 반대로 ]
재고 있음: True
재고 없음: False

재고 있음: False
재고 없음: True
```

### 🔍 not 연산의 원리

- `not True` → **False**
- `not False` → **True**

### 💡 실전 활용

```python
# 로그인 안 된 상태라면
if not is_logged_in:
    print("로그인이 필요합니다")

# 재고가 없다면
if not has_stock:
    print("품절")

# 성인이 아니라면
if not (age >= 19):
    print("미성년자")
```

"~이 아니라면"을 표현할 때 `not`을 사용합니다!

## 7. 실전 패턴 1 - 성인 검증

### 나이로 성인 여부 판단하기

한국에서는 만 19세 이상을 성인으로 간주합니다. 이를 Boolean으로 표현해봅시다.

````python
# 패턴 1: 성인 검증
age = 25
print(f"나이: {age}")
is_adult = age >= 19  # 비교 결과를 변수에 저장
print(f"성인인가? {is_adult}")  # True

age = 17
print(f"\n나이: {age}")
is_adult = age >= 19
print(f"성인인가? {is_adult}")  # False
````

**실행 결과:**
```
나이: 25
성인인가? True

나이: 17
성인인가? False
```

### 🔍 코드 분석

```python
is_adult = age >= 19
```

이 한 줄이 다음과 같이 동작합니다:
1. `age >= 19`를 계산 → 25 >= 19 → True
2. 계산 결과(True)를 `is_adult` 변수에 저장
3. 나중에 `if is_adult:` 같은 곳에서 사용 가능!

### 💡 왜 변수에 저장하나요?

```python
# 방법 1: 매번 계산 (비효율적)
if age >= 19:
    print("성인")
if age >= 19:
    print("주류 구매 가능")
if age >= 19:
    print("계약 가능")

# 방법 2: 한 번 계산 후 재사용 (효율적!)
is_adult = age >= 19
if is_adult:
    print("성인")
if is_adult:
    print("주류 구매 가능")
if is_adult:
    print("계약 가능")
```

코드가 **읽기 쉽고** **재사용**하기 좋습니다!

## 8. 실전 패턴 2 - 로그인 검증

### 아이디 AND 비밀번호 모두 확인

로그인은 아이디**도** 맞아야 하고 비밀번호**도** 맞아야 합니다. 하나라도 틀리면 실패!

````python
# 패턴 2: 로그인 검증 (아이디 AND 비밀번호)
input_id = "admin"
input_pw = "1234"
correct_id = "admin"
correct_pw = "1234"

print(f"입력 아이디: {input_id}, 입력 비밀번호: {input_pw}")
print(f"저장된 아이디: {correct_id}, 저장된 비밀번호: {correct_pw}")

# 둘 다 맞아야 로그인 성공
login_success = (input_id == correct_id) and (input_pw == correct_pw)
print(f"로그인 성공? {login_success}")  # True

# 아이디만 틀린 경우
input_id = "user"
login_success = (input_id == correct_id) and (input_pw == correct_pw)
print(f"\n아이디 틀림 → 로그인 성공? {login_success}")  # False
````

**실행 결과:**
```
입력 아이디: admin, 입력 비밀번호: 1234
저장된 아이디: admin, 저장된 비밀번호: 1234
로그인 성공? True

아이디 틀림 → 로그인 성공? False
```

### 🔍 코드 분석

```python
login_success = (input_id == correct_id) and (input_pw == correct_pw)
```

실행 과정:
1. `input_id == correct_id` 계산 → "admin" == "admin" → True
2. `input_pw == correct_pw` 계산 → "1234" == "1234" → True
3. `True and True` 계산 → True
4. 최종 결과(True)를 `login_success`에 저장

두 번째 케이스:
1. `input_id == correct_id` 계산 → "user" == "admin" → False
2. `input_pw == correct_pw` 계산 → "1234" == "1234" → True
3. `False and True` 계산 → False (하나라도 False면 False!)

### 💡 괄호를 사용하는 이유

```python
# 괄호 있음 (권장) - 읽기 쉬움
(input_id == correct_id) and (input_pw == correct_pw)

# 괄호 없어도 동작은 함 (권장하지 않음)
input_id == correct_id and input_pw == correct_pw
```

괄호를 사용하면 **어떤 부분을 먼저 계산하는지** 명확합니다!

## 9. 실전 패턴 3 - 범위 체크

### 점수가 0~100 사이인지 확인

점수는 0점 이상 100점 이하여야 유효합니다. Python은 이를 매우 간단하게 표현할 수 있습니다!

````python
# 패턴 3: 범위 체크 (점수가 0~100 사이인가?)
score = 85
print(f"점수: {score}")
is_valid = (score >= 0) and (score <= 100)
print(f"유효한 점수? {is_valid}")  # True

# 더 간단한 표현 (Python 특징!)
is_valid = 0 <= score <= 100
print(f"유효한 점수? (간단 표현) {is_valid}")  # True

score = 150
print(f"\n점수: {score}")
is_valid = 0 <= score <= 100
print(f"유효한 점수? {is_valid}")  # False
````

**실행 결과:**
```
점수: 85
유효한 점수? True
유효한 점수? (간단 표현) True

점수: 150
유효한 점수? False
```

### 🔍 두 가지 표현 방식

```python
# 방법 1: and 사용 (일반적)
(score >= 0) and (score <= 100)
# → "점수가 0 이상이고 AND 점수가 100 이하"

# 방법 2: 연쇄 비교 (Python 특징!)
0 <= score <= 100
# → "0 이상 score 이하 100"
```

두 방법 모두 동일하게 동작하지만, **방법 2가 더 간결하고 읽기 쉽습니다!**

### ⚠️ 다른 언어와의 차이

대부분의 프로그래밍 언어에서는 `0 <= score <= 100` 같은 표현을 지원하지 않습니다. Python만의 편리한 기능입니다! 🐍

### 💡 실전 활용

```python
# 나이 범위 체크
if 13 <= age <= 18:
    print("청소년")

# 가격 범위 체크
if 10000 <= price <= 50000:
    print("적정 가격")

# 온도 범위 체크
if -10 <= temperature <= 35:
    print("정상 범위")
```

## 10. 실전 패턴 4 - 할인 대상 판단

### 회원 OR 경로우대 (하나만 만족해도 OK)

할인 혜택은 회원이**거나** 65세 이상이면 받을 수 있습니다. 둘 중 하나만 해당해도 됩니다!

````python
# 패턴 4: 할인 대상 (회원 OR 경로우대)
is_member = False
age = 70

print(f"회원: {is_member}, 나이: {age}")
is_discount = is_member or (age >= 65)
print(f"할인 대상? {is_discount}")  # True (경로우대)

is_member = True
age = 30
print(f"\n회원: {is_member}, 나이: {age}")
is_discount = is_member or (age >= 65)
print(f"할인 대상? {is_discount}")  # True (회원)

is_member = False
age = 30
print(f"\n회원: {is_member}, 나이: {age}")
is_discount = is_member or (age >= 65)
print(f"할인 대상? {is_discount}")  # False
````

**실행 결과:**
```
회원: False, 나이: 70
할인 대상? True

회원: True, 나이: 30
할인 대상? True

회원: False, 나이: 30
할인 대상? False
```

### 🔍 케이스별 분석

**케이스 1**: 비회원이지만 70세
- `is_member or (age >= 65)`
- `False or True` → **True** ✓

**케이스 2**: 회원이지만 30세
- `is_member or (age >= 65)`
- `True or False` → **True** ✓

**케이스 3**: 비회원이고 30세
- `is_member or (age >= 65)`
- `False or False` → **False** ✗

### 💡 or의 특징

`or` 연산자는 **하나만 만족해도 True**이므로:
- 회원이면 → 나이 상관없이 할인 ✓
- 65세 이상이면 → 회원 아니어도 할인 ✓
- 둘 다 아니면 → 할인 없음 ✗

## 11. 실전 패턴 5 - 복잡한 조건 조합

### and + or + not을 모두 사용하기

실제 프로그램에서는 여러 조건을 복잡하게 조합해야 할 때가 많습니다. 차근차근 단계별로 나누어 생각하면 어렵지 않습니다!

````python
# 패턴 5: 복잡한 조건 조합 (and + or + not)
# 조건: (회원이거나 VIP이고) AND (재고가 있고) AND (판매중지가 아님)
is_member = True
is_vip = False
stock = 10
is_discontinued = False

print(f"회원: {is_member}, VIP: {is_vip}, 재고: {stock}, 판매중지: {is_discontinued}")

# 단계별로 계산
can_access = is_member or is_vip  # 접근 권한
has_stock = stock > 0              # 재고 있음
is_available = not is_discontinued # 판매 가능

print(f"접근 가능: {can_access}")        # True
print(f"재고 있음: {has_stock}")         # True
print(f"판매 가능: {is_available}")      # True

can_purchase = can_access and has_stock and is_available
print(f"구매 가능? {can_purchase}")  # True

# 한 줄로도 가능 (가독성은 떨어짐)
can_purchase = (is_member or is_vip) and (stock > 0) and (not is_discontinued)
print(f"구매 가능? (한 줄) {can_purchase}")  # True
````

**실행 결과:**
```
회원: True, VIP: False, 재고: 10, 판매중지: False
접근 가능: True
재고 있음: True
판매 가능: True
구매 가능? True
구매 가능? (한 줄) True
```

### 🔍 단계별 실행 과정

```python
# 1단계: 접근 권한 체크
can_access = is_member or is_vip
# → True or False → True

# 2단계: 재고 체크
has_stock = stock > 0
# → 10 > 0 → True

# 3단계: 판매 가능 체크
is_available = not is_discontinued
# → not False → True

# 4단계: 최종 판단
can_purchase = can_access and has_stock and is_available
# → True and True and True → True
```

### 💡 복잡한 조건을 다루는 팁

**❌ 나쁜 예 (한 줄에 다 쓰기)**
```python
# 읽기 어렵고 실수하기 쉬움
if (is_member or is_vip) and (stock > 0) and (not is_discontinued) and (price < budget):
    # ...
```

**✅ 좋은 예 (의미 있는 변수로 나누기)**
```python
# 읽기 쉽고 디버깅도 쉬움
can_access = is_member or is_vip
has_stock = stock > 0
is_available = not is_discontinued
is_affordable = price < budget

if can_access and has_stock and is_available and is_affordable:
    # ...
```

**복잡한 조건은 의미 있는 변수로 나누어 표현**하면 훨씬 이해하기 쉽습니다!

## 12. 실생활 시나리오 - 영화 티켓 구매

### 종합 실습: 여러 조건을 모두 활용하기

지금까지 배운 모든 것을 종합해서 **실제와 같은 시나리오**를 만들어봅시다!

````python
# 시나리오: 영화 티켓 구매 가능 여부
age = 17
has_money = True
is_rated_adult = True  # 청소년 관람불가

print(f"나이: {age}, 돈 있음: {has_money}, 청불: {is_rated_adult}")

# 조건 1: 성인 영화면 19세 이상이어야 함
# "청불이 아니거나, 19세 이상이어야 함"
age_check = (not is_rated_adult) or (age >= 19)
print(f"나이 조건 통과? {age_check}")  # False (17세라서)

# 조건 2: 돈이 있어야 함
money_check = has_money
print(f"돈 조건 통과? {money_check}")  # True

# 최종 구매 가능 여부
can_buy = age_check and money_check
print(f"티켓 구매 가능? {can_buy}")  # False


# 성인으로 변경
age = 20
print(f"\n나이: {age}로 변경")
age_check = (not is_rated_adult) or (age >= 19)
can_buy = age_check and money_check
print(f"티켓 구매 가능? {can_buy}")  # True
````

**실행 결과:**
```
나이: 17, 돈 있음: True, 청불: True
나이 조건 통과? False
돈 조건 통과? True
티켓 구매 가능? False

나이: 20로 변경
티켓 구매 가능? True
```

### 🔍 상세 분석

**17세일 때:**
```python
# 나이 조건
age_check = (not is_rated_adult) or (age >= 19)
# → (not True) or (17 >= 19)
# → False or False
# → False ✗

# 최종 판단
can_buy = age_check and money_check
# → False and True
# → False ✗ (나이가 부족해서 구매 불가)
```

**20세일 때:**
```python
# 나이 조건
age_check = (not is_rated_adult) or (age >= 19)
# → (not True) or (20 >= 19)
# → False or True
# → True ✓

# 최종 판단
can_buy = age_check and money_check
# → True and True
# → True ✓ (구매 가능!)
```

### 💡 조건 로직 설계 팁

복잡한 조건을 만들 때는 **자연어로 먼저 생각**하세요:

1. 자연어로 조건 작성
   - "청불이 **아니거나**, 19세 **이상**이면 OK"

2. Boolean 표현으로 변환
   - `(not is_rated_adult) or (age >= 19)`

3. 변수로 저장해서 의미 명확히
   - `age_check = (not is_rated_adult) or (age >= 19)`
