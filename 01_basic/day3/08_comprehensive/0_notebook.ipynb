### 2. 논리 연산자
```python
# and: 둘 다 True
if age >= 18 and has_license:
    print("운전 가능")

# or: 하나만 True
if is_weekend or is_holiday:
    print("쉬는 날")

# not: 반대
if not is_logged_in:
    print("로그인 필요")
```

### 4. for문
```python
# 리스트 순회
for item in items:
    print(item)

# enumerate (인덱스 + 값)
for i, item in enumerate(items):
    print(f"{i}: {item}")

# 딕셔너리 순회
for key, value in dict.items():
    print(f"{key}: {value}")
```

### 5. while문
```python
while condition:
    # 조건이 True인 동안 반복
    if some_condition:
        break  # 루프 탈출
    if other_condition:
        continue  # 다음 반복으로
```

**핵심 개념**: `while문` + `if문` + `딕셔너리`

### 요구사항

1. ✅ 메뉴 목록 출력
2. ✅ 사용자가 메뉴 번호 선택 → 주문 목록에 추가
3. ✅ 0 입력 → 주문 종료 + 총 금액 출력

### 사용할 개념

```python
# while문으로 반복
while True:
    # 입력 받기
    choice = int(input("선택: "))
    
    # 종료 조건
    if choice == 0:
        break
    
    # 딕셔너리에서 데이터 가져오기
    if choice in menu:
        item = menu[choice]
```

## 🔴 도전 Problem: 숫자 맞추기 게임 (10분)

### 문제 설명

컴퓨터가 선택한 1-100 사이의 숫자를 맞추는 게임을 만들어봅시다!

**핵심 개념**: `while문` + `if/elif/else` + `break` 종합 활용

### 게임 규칙

1. 🎲 컴퓨터가 1-100 사이의 랜덤 숫자 선택 (이미 제공됨)
2. 🎯 최대 7번의 기회
3. 💬 "UP" 또는 "DOWN" 힌트 제공
4. 🎉 정답 맞추면 시도 횟수 출력 + 게임 종료
5. 😢 7번 안에 못 맞추면 정답 공개

### 사용할 개념

```python
# 제한된 횟수 동안 반복
max_tries = 7
tries = 0

while tries < max_tries:
    guess = int(input("숫자: "))
    tries += 1
    
    if guess == answer:
        print("정답!")
        break  # 게임 종료
    elif guess < answer:
        print("UP!")
    else:
        print("DOWN!")
else:
    # while문이 break 없이 끝났을 때
    print(f"실패! 정답은 {answer}")
```

### 💡 핵심 팁

- **while-else 구문**: `while`문이 `break` 없이 정상 종료되면 `else` 블록 실행
- **시도 횟수 관리**: `tries += 1`로 증가, `max_tries - tries`로 남은 기회 계산
