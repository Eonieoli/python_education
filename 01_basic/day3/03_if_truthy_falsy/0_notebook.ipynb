## 📖 조건문이란 무엇인가?

### 프로그램은 선택의 연속입니다

우리가 살아가면서 매 순간 선택을 하듯이, 프로그램도 끊임없이 **"만약 ~라면 이렇게 하고, 그렇지 않으면 저렇게 해라"**라는 판단을 내려야 합니다.

예를 들어:
- 로그인할 때: "비밀번호가 맞으면 로그인 성공, 틀리면 실패"
- 쇼핑몰에서: "재고가 있으면 구매 가능, 없으면 품절"
- 성적 처리: "90점 이상이면 A, 80점 이상이면 B..."

이런 **조건에 따른 분기 처리**를 하는 것이 바로 **조건문(Conditional Statement)**입니다.

### 왜 조건문이 중요한가요?

조건문 없이는 프로그램이 항상 똑같이 동작할 수밖에 없습니다. 하지만 실제 세상은 다양한 상황에 대응해야 하죠.

**FastAPI에서도 조건문은 필수입니다**:
```python
# 사용자 권한 체크
if user.role == "admin":
    return all_data  # 관리자는 모든 데이터 접근
else:
    return user_data  # 일반 사용자는 자기 데이터만

# 입력값 검증
if age < 0:
    raise ValueError("나이는 0 이상이어야 합니다")
```

### Python의 조건문 구조

Python은 `if`, `elif`, `else` 세 가지 키워드로 조건문을 만듭니다:

```python
if 조건1:
    # 조건1이 참일 때 실행
elif 조건2:
    # 조건1은 거짓이고, 조건2가 참일 때 실행
elif 조건3:
    # 조건1, 2는 거짓이고, 조건3이 참일 때 실행
else:
    # 모든 조건이 거짓일 때 실행
```

**핵심 포인트**:
- `if`: 필수! 가장 먼저 체크할 조건
- `elif`: 선택! 추가 조건이 필요할 때 (여러 개 가능)
- `else`: 선택! 모든 조건이 거짓일 때
- **들여쓰기(indentation)**가 매우 중요! Python은 들여쓰기로 코드 블록을 구분합니다

## 1. if/elif/else 기본 구조

### 성적 등급을 판정해봅시다

학생의 점수를 받아서 A, B, C, D 등급을 매기는 프로그램을 만들어보겠습니다.

**등급 기준**:
- 90점 이상: A 등급
- 80점 이상 90점 미만: B 등급
- 70점 이상 80점 미만: C 등급
- 70점 미만: D 등급

**중요**: `elif`는 **위에서부터 순서대로** 체크합니다. 첫 번째 조건이 참이면 나머지는 체크하지 않습니다!

````python
# 성적에 따른 등급 판정
score = 85  # 학생의 점수

# 조건문으로 등급 판정
if score >= 90:  # 첫 번째 조건: 90점 이상인가?
    print("A 등급")
elif score >= 80:  # 두 번째 조건: 80점 이상인가? (이미 90점 이상은 아님)
    print("B 등급")  # 출력: B 등급 ✅
elif score >= 70:  # 세 번째 조건: 70점 이상인가?
    print("C 등급")
else:  # 모든 조건이 거짓 (70점 미만)
    print("D 등급")
````

**실행 결과**:
```
B 등급
```

### 🔍 코드 실행 흐름 분석

1. `score = 85`로 시작
2. `if score >= 90`: 85는 90 이상이 아니므로 **거짓** → 다음으로
3. `elif score >= 80`: 85는 80 이상이므로 **참** → "B 등급" 출력
4. 조건이 만족되었으므로 나머지 `elif`, `else`는 **실행되지 않음**

### 💡 왜 `elif score >= 80`에서 `score < 90`을 체크하지 않나요?

좋은 질문입니다! 이미 첫 번째 `if`에서 90 이상을 걸러냈기 때문에, `elif`에 도달했다는 것은 "90점 미만"이 확실합니다. 따라서 굳이 `score < 90`을 추가로 체크할 필요가 없습니다.

이것이 바로 **위에서부터 순서대로 체크하는 조건문의 효율성**입니다!

### ⚠️ 일반적인 실수

```python
# ❌ 잘못된 예시
if score >= 70:  # 70점 이상이면 C?
    print("C 등급")  # 85점도 C가 되어버림!
elif score >= 80:
    print("B 등급")  # 여기는 절대 실행 안 됨
```

조건의 **순서가 중요**합니다! 항상 **더 엄격한 조건을 먼저** 체크해야 합니다.

## 2. Truthy/Falsy - Python의 숨겨진 보석 💎

### 모든 값은 True 또는 False로 취급될 수 있습니다

이것은 Python의 **매우 강력하고 중요한 개념**입니다!

Python에서는 `True`와 `False`라는 Boolean 값 외에도, **모든 값을 조건문에서 참(Truthy) 또는 거짓(Falsy)으로 평가**할 수 있습니다.

### 왜 이게 중요한가요?

실무에서는 "데이터가 있는지 없는지", "리스트가 비어있는지", "사용자가 값을 입력했는지" 등을 매우 자주 체크합니다. Truthy/Falsy를 이해하면 이런 체크를 **훨씬 간결하고 파이썬답게** 할 수 있습니다!

```python
# ❌ 초보자 스타일 (길고 복잡)
if len(items) > 0:
    print("아이템이 있음")

# ✅ 프로 스타일 (간결하고 명확)
if items:
    print("아이템이 있음")
```

### False로 취급되는 값들 (Falsy Values)

Python에서 다음 값들은 조건문에서 **거짓(False)**으로 취급됩니다:

1. `False` - Boolean 거짓값
2. `None` - 값이 없음을 나타내는 특수한 값
3. `0` - 숫자 0 (정수)
4. `0.0` - 숫자 0 (실수)
5. `""` - 빈 문자열
6. `[]` - 빈 리스트
7. `{}` - 빈 딕셔너리
8. `()` - 빈 튜플
9. `set()` - 빈 집합

### True로 취급되는 값들 (Truthy Values)

**위의 Falsy 값들을 제외한 모든 값**은 참(True)으로 취급됩니다!

- 0이 아닌 모든 숫자: `1`, `-5`, `3.14`, ...
- 내용이 있는 문자열: `"hello"`, `" "` (공백도 내용!), ...
- 내용이 있는 컬렉션: `[1]`, `{"a": 1}`, ...

---

````python
# False로 취급되는 값들 (Falsy) - 하나하나 확인해봅시다!

# 1. 빈 문자열
if "":  # 빈 문자열은 False
    print("실행 안 됨")  # 이 줄은 실행되지 않습니다
else:
    print("빈 문자열은 Falsy입니다")  # 출력: 빈 문자열은 Falsy입니다

# 2. 숫자 0
if 0:  # 0은 False
    print("실행 안 됨")
else:
    print("0은 Falsy입니다")  # 출력: 0은 Falsy입니다

# 3. 빈 리스트
if []:  # 빈 리스트는 False
    print("실행 안 됨")
else:
    print("빈 리스트는 Falsy입니다")  # 출력: 빈 리스트는 Falsy입니다

# 4. 빈 딕셔너리
if {}:  # 빈 딕셔너리는 False
    print("실행 안 됨")
else:
    print("빈 딕셔너리는 Falsy입니다")  # 출력: 빈 딕셔너리는 Falsy입니다

# 5. None
if None:  # None은 False
    print("실행 안 됨")
else:
    print("None은 Falsy입니다")  # 출력: None은 Falsy입니다
````

**실행 결과**:
```
빈 문자열은 Falsy입니다
0은 Falsy입니다
빈 리스트는 Falsy입니다
빈 딕셔너리는 Falsy입니다
None은 Falsy입니다
```

### 💡 핵심 이해

Falsy 값들을 보면 공통점이 있습니다:
- **"없음"**, **"비어있음"**, **"0"**을 나타내는 값들입니다
- 즉, **"의미 있는 값이 없다"**는 것을 표현합니다

이것이 Python이 이 값들을 False로 취급하는 철학적 이유입니다!

---

````python
# True로 취급되는 값들 (Truthy) - 의미 있는 값들!

# 1. 문자열 (내용이 있으면 True)
if "hello":  # 문자열에 내용이 있으면 True
    print("문자열 있음")  # 출력: 문자열 있음

# 2. 0이 아닌 숫자
if 1:  # 0이 아닌 숫자는 True
    print("0이 아님")  # 출력: 0이 아님

if -5:  # 음수도 True!
    print("-5도 True입니다")  # 출력: -5도 True입니다

# 3. 내용이 있는 리스트
if [1, 2]:  # 리스트에 요소가 있으면 True
    print("리스트 있음")  # 출력: 리스트 있음

# 4. 내용이 있는 딕셔너리
if {"name": "Alice"}:  # 딕셔너리에 데이터가 있으면 True
    print("딕셔너리 있음")  # 출력: 딕셔너리 있음
````

**실행 결과**:
```
문자열 있음
0이 아님
-5도 True입니다
리스트 있음
딕셔너리 있음
```

## 3. Truthy/Falsy 실전 활용 - 장바구니 체크

### 쇼핑몰에서 장바구니가 비어있는지 확인하기

실무에서 가장 많이 사용하는 패턴 중 하나입니다!

**시나리오**: 사용자가 결제하기 버튼을 눌렀을 때, 장바구니가 비어있으면 "장바구니가 비어있습니다"라고 알려주고, 상품이 있으면 결제를 진행해야 합니다.

````python
# 실전 활용: 장바구니 체크
cart = []  # 빈 장바구니 (사용자가 아무것도 담지 않음)

# Truthy/Falsy를 활용한 간결한 체크
if cart:  # 빈 리스트는 Falsy → 조건이 거짓
    print(f"상품이 {len(cart)}개 있습니다. 결제를 진행합니다.")
else:
    print("장바구니가 비어있습니다")  # 출력: 장바구니가 비어있습니다 ✅
````

**실행 결과**:
```
장바구니가 비어있습니다
```

### 💡 초보자 vs 프로 비교

```python
# ❌ 초보자 스타일 (불필요하게 복잡)
if len(cart) == 0:
    print("장바구니가 비어있습니다")

# ✅ 프로 스타일 (간결하고 파이썬답게)
if not cart:  # 또는 if cart: 의 반대
    print("장바구니가 비어있습니다")
```

**두 번째 방법이 더 좋은 이유**:
1. 코드가 짧고 읽기 쉽습니다
2. "리스트가 비어있다"는 의도가 더 명확합니다
3. Python 커뮤니티의 표준 스타일입니다

### ⚠️ 주의: 공백도 내용입니다!

```python
if " ":  # 공백 문자도 문자열 내용이므로 True!
    print("공백도 Truthy입니다")  # 출력됨
```

완전히 비어있는 `""`와 공백이 있는 `" "`는 다릅니다!

## 4. in 연산자 활용

### "어떤 값이 포함되어 있는가?"를 체크하기

`in` 연산자는 **어떤 값이 컬렉션(리스트, 딕셔너리, 문자열 등)에 포함되어 있는지** 확인합니다.

**문법**: `값 in 컬렉션`
- 포함되어 있으면 `True` 반환
- 없으면 `False` 반환

### 리스트에서 값 찾기

````python
# 리스트에서 값 찾기
fruits = ["사과", "바나나", "딸기"]

# "사과"가 리스트에 있는지 확인
if "사과" in fruits:  # True → 조건 만족
    print("사과가 있습니다")  # 출력: 사과가 있습니다 ✅

# "오렌지"가 리스트에 있는지 확인
if "오렌지" in fruits:  # False → 조건 불만족
    print("오렌지가 있습니다")
else:
    print("오렌지가 없습니다")  # 출력: 오렌지가 없습니다
````

**실행 결과**:
```
사과가 있습니다
오렌지가 없습니다
```

### 딕셔너리에서 키 확인

딕셔너리에서 `in`을 사용하면 **키(key)**의 존재 여부를 확인합니다.

이것은 **매우 중요**합니다! 없는 키에 접근하면 KeyError가 발생하니까요.

````python
# 딕셔너리에서 키 확인
user = {"name": "홍길동", "age": 25}  # 이메일 정보가 없음

# "email" 키가 있는지 확인
if "email" in user:  # False → 없음
    print(f"이메일: {user['email']}")
else:
    print("이메일 없음")  # 출력: 이메일 없음 ✅

# "name" 키가 있는지 확인
if "name" in user:  # True → 있음
    print(f"이름: {user['name']}")  # 출력: 이름: 홍길동
````

**실행 결과**:
```
이메일 없음
이름: 홍길동
```

### 💡 왜 in으로 체크해야 하나요?

```python
# ❌ 위험한 방법 - KeyError 발생 가능
email = user["email"]  # KeyError! 프로그램 중단!

# ✅ 안전한 방법 1 - in 연산자
if "email" in user:
    email = user["email"]
else:
    email = "없음"

# ✅ 안전한 방법 2 - get() 메서드 (Day 2에서 배웠죠!)
email = user.get("email", "없음")
```

**FastAPI에서 실전 활용**:
```python
# 사용자가 선택적으로 보낼 수 있는 데이터 처리
if "bio" in user_data:
    user.bio = user_data["bio"]  # 있으면 업데이트
else:
    user.bio = "소개가 없습니다"  # 없으면 기본값
```

## 5. 실전 예제: 로그인 검증

### 실제 웹 서비스의 로그인 로직을 구현해봅시다

지금까지 배운 모든 것을 종합하는 실전 예제입니다!

**요구사항**:
1. 아이디가 입력되지 않았으면 "아이디를 입력하세요"
2. 비밀번호가 입력되지 않았으면 "비밀번호를 입력하세요"
3. 아이디와 비밀번호가 모두 맞으면 "로그인 성공!"
4. 하나라도 틀리면 "로그인 실패!"

여기서 **Truthy/Falsy**, **and 연산자**, **조건문**을 모두 사용합니다!

````python
# 실전 예제: 로그인 검증

# 시스템에 저장된 올바른 계정 정보
username = "admin"
password = "1234"

# 사용자가 입력한 정보 (빈 문자열 = 입력 안 함)
input_id = ""  # 아이디를 입력하지 않음!
input_pw = "1234"

# 로그인 검증 로직
if not input_id:  # 빈 문자열은 Falsy → not을 붙이면 True
    print("아이디를 입력하세요")  # 출력: 아이디를 입력하세요 ✅
elif not input_pw:  # 비밀번호가 비어있는지 체크
    print("비밀번호를 입력하세요")
elif input_id == username and input_pw == password:  # 둘 다 일치하는지
    print("로그인 성공!")
else:  # 입력은 했지만 틀린 경우
    print("로그인 실패!")
````

**실행 결과**:
```
아이디를 입력하세요
```

### 🔍 코드 분석

**1단계**: `if not input_id:`
- `input_id`는 빈 문자열 `""`
- 빈 문자열은 Falsy (거짓으로 취급)
- `not`을 붙이면 → `not False` = `True`
- 조건 만족! → "아이디를 입력하세요" 출력
- 나머지 `elif`, `else`는 실행되지 않음

### 💡 다양한 시나리오 테스트

같은 코드에서 입력값만 바꿔서 테스트해보세요:

```python
# 시나리오 1: 아이디만 입력 안 함
input_id = ""
input_pw = "1234"
# 결과: "아이디를 입력하세요"

# 시나리오 2: 비밀번호만 입력 안 함
input_id = "admin"
```

```python
@app.post("/login")
```

## 📌 핵심 정리

### 이번 시간에 배운 내용:

1. **조건문 (if/elif/else)**: 조건에 따라 다른 코드를 실행하는 핵심 구조
   - `if`: 첫 번째 조건 체크
   - `elif`: 추가 조건 (여러 개 가능)
   - `else`: 모든 조건이 거짓일 때
   - **순서가 중요!** 위에서부터 체크하고, 만족하면 나머지는 실행 안 함

2. **Truthy/Falsy**: Python의 강력한 특징 (프로처럼 코딩하려면 필수!)
   - **Falsy**: `False`, `None`, `0`, `""`, `[]`, `{}` (비어있거나 없는 것)
   - **Truthy**: 위의 Falsy를 제외한 모든 값
   - 활용: `if cart:` (장바구니에 뭔가 있나?), `if not name:` (이름 안 넣었나?)

3. **in 연산자**: 값이 컬렉션에 포함되어 있는지 확인
   - 리스트: `if "사과" in fruits:`
   - 딕셔너리: `if "email" in user:` (키 존재 확인)
   - KeyError 방지에 필수!

4. **실전 패턴**: 입력 검증, 데이터 존재 확인, 로그인 검증
   - 웹 개발에서 매일 사용하는 필수 패턴들

### ⚠️ 반드시 기억할 점:

1. **들여쓰기가 생명!** Python은 들여쓰기로 코드 블록을 구분합니다
   ```python
   if score >= 90:
       print("A")  # 4칸 들여쓰기 (또는 탭)
   ```

2. **조건의 순서가 중요!** 더 엄격한 조건을 먼저 체크
   ```python
   # ✅ 올바른 순서
   if score >= 90:  # 가장 엄격
   elif score >= 80:
   elif score >= 70:
   ```

3. **Truthy/Falsy는 프로의 무기!** 간결하고 파이썬다운 코드를 위해 꼭 익히세요
   ```python
   if items:  # ✅ 프로 스타일
   if len(items) > 0:  # ❌ 초보자 스타일
   ```

4. **in 연산자로 안전하게!** 딕셔너리 키 접근 전에 항상 체크
   ```python
   if "email" in user:  # ✅ 안전
       email = user["email"]
   ```
