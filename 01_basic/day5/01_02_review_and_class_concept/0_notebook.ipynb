## 🔄 Day 4 복습: 타입 힌트 복습

본격적으로 클래스를 배우기 전에, 어제 배운 **타입 힌트**를 간단히 복습해봅시다.

왜냐하면 오늘 배울 클래스에서도 **타입 힌트가 필수**로 사용되기 때문입니다! ✨

### 함수와 타입 힌트 기본

어제 배운 내용을 떠올려봅시다:

````python
# 기본 타입 힌트 복습
def greet(name: str) -> str:
    """이름을 받아서 인사말을 반환하는 함수"""
    return f"Hello, {name}"

# 함수 실행
result = greet("Alice")
print(result)  # 출력: Hello, Alice
````

**실행 결과:**
```
Hello, Alice
```

### 고급 타입 힌트 복습

`List`, `Dict`, `Optional` 같은 고급 타입도 복습해봅시다:

````python
from typing import List, Dict, Optional

def get_users() -> List[str]:
    """사용자 이름 리스트를 반환하는 함수"""
    return ["Alice", "Bob", "Charlie"]

# 함수 실행
users = get_users()
print(users)  # 출력: ['Alice', 'Bob', 'Charlie']
````

**실행 결과:**
```
['Alice', 'Bob', 'Charlie']
```

## 📖 클래스(Class)란 무엇인가?

### 🤔 왜 클래스가 필요할까요?

학생 정보를 관리한다고 생각해봅시다. 학생 한 명의 정보는:
- 이름: "홍길동"
- 나이: 20
- 학번: "2024001"
- 성적: [85, 90, 88]

지금까지 배운 방법으로는 이렇게 저장했을 겁니다:

```python
# 학생 1
student1_name = "홍길동"
student1_age = 20
student1_id = "2024001"
student1_scores = [85, 90, 88]

# 학생 2
student2_name = "김철수"
student2_age = 21
student2_id = "2024002"
student2_scores = [78, 82, 80]

# 학생 3, 4, 5... 😱
```

**문제점:**
- 변수가 너무 많아집니다! 😰
- 학생이 100명이면? 변수가 400개!
- 관련된 데이터인데 따로따로 관리됩니다
- 학생 정보를 처리하는 함수도 매번 따로 만들어야 합니다

딕셔너리로 개선할 수도 있습니다:

```python
student1 = {
    "name": "홍길동",
    "age": 20,
    "id": "2024001",
    "scores": [85, 90, 88]

```

하지만 여전히 문제가 있습니다:
- 평균 성적을 계산하는 함수를 어디에 둘까요?
- 성적을 추가하는 함수는?
- 데이터와 기능(함수)이 분리되어 있습니다

### ✨ 클래스가 해결책입니다!

**클래스(Class)**는:
- **데이터(변수)**와 **기능(함수)**를 **하나로 묶는** 방법입니다
- "학생"이라는 **설계도(틀)**를 만드는 것입니다
- 이 설계도로 학생 **객체**를 여러 개 만들 수 있습니다

### 🧁 붕어빵 비유

가장 많이 사용하는 비유입니다:

```
클래스(Class)    = 붕어빵 틀 (설계도)
객체(Object)     = 실제 붕어빵 (설계도로 만든 결과물)
인스턴스(Instance) = 객체와 같은 의미
```

- 붕어빵 틀은 **하나**지만, 붕어빵은 **여러 개** 만들 수 있습니다
- 각 붕어빵은 독립적입니다 (속 재료가 다를 수 있음)
- 하지만 모두 같은 틀에서 나왔기 때문에 기본 구조는 같습니다

### 🎯 FastAPI와 클래스

FastAPI를 배울 때 **Pydantic BaseModel**을 사용하게 됩니다.
이것도 **클래스**입니다!

```python
# FastAPI에서 이렇게 쓰게 됩니다 (미리보기)
from pydantic import BaseModel

class User(BaseModel):  # ← 클래스!
    name: str
    age: int
    email: str
```

지금 배우는 클래스 개념이 **FastAPI의 핵심**입니다! 💪

## 🏗️ 클래스의 구조

클래스는 크게 **3가지 요소**로 구성됩니다:

### 1. class 키워드

클래스를 정의할 때 사용하는 키워드입니다.

```python
class 클래스명:
    # 클래스 내용
```

**클래스 이름 규칙:**
- **대문자로 시작**합니다 (PascalCase)
- 예: `User`, `BankAccount`, `StudentInfo`
- 함수와 구분하기 위함입니다 (함수는 소문자로 시작)

### 2. `__init__` 메서드 (생성자)

객체를 **생성할 때 자동으로 실행**되는 특별한 메서드입니다.

```python
def __init__(self, name: str, age: int):
    self.name = name  # 인스턴스 변수
    self.age = age    # 인스턴스 변수
```

**역할:**
- 객체의 **초기 상태를 설정**합니다
- **인스턴스 변수를 초기화**합니다
- 생성자(Constructor)라고도 부릅니다

**왜 `__init__`일까요?**
- Python의 특별한 메서드(Magic Method)입니다
- 앞뒤에 `__` (언더스코어 2개)가 붙습니다
- "initialize"(초기화)의 줄임말입니다

### 3. self

`self`는 **자기 자신(객체 자신)**을 가리킵니다.

```python
def __init__(self, name: str):
    self.name = name  # 이 객체의 name 변수
```

**왜 필요할까요?**
- 여러 객체가 만들어질 때, "이 데이터는 누구의 것인가?"를 구분하기 위해
- `student1.name`과 `student2.name`은 다른 값입니다
- `self`가 각 객체를 구분해줍니다

**중요한 규칙:**
- 클래스의 **모든 메서드**의 첫 번째 매개변수는 `self`입니다
- 호출할 때는 `self`를 넘기지 **않습니다** (Python이 자동으로 넘겨줌)

```python
# 정의할 때
def greet(self):  # self가 있음
    return f"Hello, {self.name}"

# 호출할 때
user.greet()  # self는 안 넘김!
```

### 4. 인스턴스 변수 (Instance Variable)

**각 객체가 가지는 고유한 데이터**입니다.

```python
self.name = name  # 인스턴스 변수
self.age = age    # 인스턴스 변수
```

- `self.`를 붙여서 정의합니다
- 객체마다 **다른 값**을 가질 수 있습니다

### 5. 메서드 (Method)

클래스 안에 정의된 **함수**를 메서드라고 부릅니다.

```python
def greet(self) -> str:
    return f"Hello, {self.name}"
```

**메서드 vs 함수:**
- 메서드: 클래스 안에 있는 함수
- 함수: 클래스 밖에 있는 함수
- 메서드는 항상 첫 번째 매개변수가 `self`입니다

### 🎨 전체 구조 한눈에 보기

```python
class User:                           # ← 클래스 정의
    def __init__(self, name: str):   # ← 생성자
        self.name = name              # ← 인스턴스 변수
    
    def greet(self) -> str:          # ← 메서드
        return f"Hello, {self.name}" # ← self로 변수 접근
```

## 🎯 타입 힌트와 클래스

클래스를 작성할 때도 **타입 힌트는 필수**입니다!

### `__init__`에 타입 힌트 붙이기

```python
class User:
    def __init__(self, name: str, age: int):
        #            ↑ 타입 힌트   ↑ 타입 힌트
        self.name = name
        self.age = age
```

**주의사항:**
- `__init__`의 반환 타입은 **명시하지 않습니다**
- 왜냐하면 항상 `None`이기 때문입니다
- 객체를 반환하는 것처럼 보이지만, 실제로는 `None`입니다

### 메서드에 타입 힌트 붙이기

```python
def greet(self) -> str:
    #           ↑ 반환 타입
    return f"Hello, {self.name}"

def is_adult(self) -> bool:
    #             ↑ 반환 타입
    return self.age >= 18
```

**왜 중요한가요?**
- FastAPI의 Pydantic 모델에서 **필수**입니다
- IDE가 자동완성을 더 잘 해줍니다
- 코드 가독성이 높아집니다
- 실수를 미리 방지할 수 있습니다

오늘부터 클래스를 작성할 때:

2. 모든 메서드에 **반드시** 반환 타입 명시

## 1️⃣ 첫 번째 클래스 만들기

가장 간단한 `User` 클래스를 만들어봅시다.

### 무엇을 만들까요?

사용자 정보를 관리하는 클래스입니다:
- **이름(name)**: 사용자의 이름
- **나이(age)**: 사용자의 나이
- **인사(greet)**: 인사말을 반환하는 기능
- **성인 확인(is_adult)**: 성인인지 확인하는 기능

### 왜 클래스로 만들까요?

- 사용자 데이터(name, age)와 기능(greet, is_adult)을 **하나로 묶어서** 관리하기 위해
- 여러 사용자를 쉽게 만들기 위해
- FastAPI에서 Pydantic 모델과 비슷한 구조를 익히기 위해

````python
class User:
    """사용자 정보를 관리하는 클래스"""
    
    def __init__(self, name: str, age: int):
        """
        생성자 메서드: 객체가 생성될 때 자동으로 실행됩니다
        
        매개변수:
            name: 사용자 이름 (문자열)
            age: 사용자 나이 (정수)
        """
        # 인스턴스 변수 초기화
        # self.name은 "이 객체의 name 변수"를 의미합니다
        self.name = name  # 매개변수 name을 인스턴스 변수에 저장
        self.age = age    # 매개변수 age를 인스턴스 변수에 저장
    
    def greet(self) -> str:
        """
        인사 메시지를 반환하는 메서드
        
        반환값:
            str: 인사 메시지
        """
        # self.name을 사용해서 이 객체의 이름에 접근합니다
        return f"안녕하세요, 저는 {self.name}입니다."
    
    def is_adult(self) -> bool:
        """
        성인 여부를 확인하는 메서드
        
        반환값:
            bool: 18세 이상이면 True, 아니면 False
        """
        # self.age를 사용해서 이 객체의 나이에 접근합니다
        return self.age >= 18


# User 클래스로 객체(인스턴스) 생성하기
# User("홍길동", 25) 호출 시 자동으로 __init__이 실행됩니다
user1 = User("홍길동", 25)

# 인스턴스 변수에 접근하기 (점 표기법 사용)
print(user1.name)  # 출력: 홍길동
print(user1.age)   # 출력: 25

# 메서드 호출하기 (self는 Python이 자동으로 넘겨줌)
print(user1.greet())      # 출력: 안녕하세요, 저는 홍길동입니다.
print(user1.is_adult())   # 출력: True (25세는 성인)
````

**실행 결과:**
```
홍길동
25
안녕하세요, 저는 홍길동입니다.
True
```

### 🔍 코드 분석

**1단계: 클래스 정의**
```python
class User:
```
- `User`라는 이름의 클래스를 정의합니다
- 이것은 "사용자"라는 설계도를 만드는 것입니다

**2단계: 생성자 정의**
```python
def __init__(self, name: str, age: int):
    self.name = name
    self.age = age
```
- 객체가 생성될 때 자동으로 실행됩니다
- `name`과 `age`를 받아서 인스턴스 변수로 저장합니다
- 타입 힌트로 `name`은 문자열, `age`는 정수임을 명시합니다

**3단계: 메서드 정의**
```python
def greet(self) -> str:
    return f"안녕하세요, 저는 {self.name}입니다."
```
- `self.name`으로 이 객체의 이름에 접근합니다
- 반환 타입 `-> str`로 문자열을 반환함을 명시합니다

**4단계: 객체 생성**
```python
user1 = User("홍길동", 25)
```
- `User` 클래스로 새로운 객체를 만듭니다
- `"홍길동"`과 `25`가 `__init__`의 `name`, `age`로 전달됩니다
- `user1`이라는 변수에 생성된 객체를 저장합니다

**5단계: 사용**
```python
user1.name        # 속성 접근
user1.greet()     # 메서드 호출
```

## 2️⃣ 여러 인스턴스 생성하기

클래스의 진짜 힘은 **여러 객체를 쉽게 만들 수 있다**는 것입니다!

### 붕어빵 틀로 붕어빵 여러 개 만들기

같은 `User` 클래스로 여러 사용자를 만들어봅시다:

````python
# 같은 User 클래스로 여러 객체 생성
# 각 객체는 독립적입니다!
user1 = User("홍길동", 25)
user2 = User("김철수", 17)
user3 = User("이영희", 30)

# 각 객체는 자신만의 데이터를 가집니다
print(user1.name, user1.age)  # 출력: 홍길동 25
print(user2.name, user2.age)  # 출력: 김철수 17
print(user3.name, user3.age)  # 출력: 이영희 30

# 각 객체마다 메서드 결과도 다릅니다
print(user1.is_adult())  # 출력: True (25세는 성인)
print(user2.is_adult())  # 출력: False (17세는 미성년자)
print(user3.is_adult())  # 출력: True (30세는 성인)
````

**실행 결과:**
```
홍길동 25
김철수 17
이영희 30
True
False
True
```

### 🎯 핵심 개념: 독립성

- `user1`, `user2`, `user3`는 **모두 독립적인 객체**입니다
- 같은 `User` 클래스에서 만들어졌지만, **각자 다른 데이터**를 가집니다
- `user1.name`을 바꿔도 `user2.name`은 영향을 받지 않습니다

### 💡 리스트와 함께 사용하기

## 3️⃣ 실전 예제: 은행 계좌 클래스

이제 좀 더 실용적인 예제를 만들어봅시다!

### 무엇을 만들까요?

은행 계좌를 관리하는 `BankAccount` 클래스입니다:

**데이터:**
- 계좌 소유자 (owner)
- 잔액 (balance)

**기능:**
- 입금 (deposit)
- 출금 (withdraw)
- 잔액 조회 (get_balance)

### 실무에서는?

실제 은행 시스템이나 금융 앱에서 이런 클래스를 사용합니다:
- 각 사용자의 계좌를 객체로 관리
- 입출금 기록을 메서드로 처리
- 잔액 부족 검증 등의 비즈니스 로직 포함

````python
class BankAccount:
    """은행 계좌를 관리하는 클래스"""
    
    def __init__(self, owner: str, balance: int = 0):
        """
        생성자 메서드
        
        매개변수:
            owner: 계좌 소유자 (필수)
            balance: 초기 잔액 (기본값 0) ← 기본값 매개변수!
        """
        self.owner = owner      # 계좌 소유자 저장
        self.balance = balance  # 초기 잔액 저장 (기본값이 있으면 생략 가능)
    
    def deposit(self, amount: int) -> None:
        """
        입금 메서드
        
        매개변수:
            amount: 입금할 금액
        
        반환값:
            None (반환값 없음)
        """
        # 잔액에 입금액을 더합니다
        self.balance += amount
        # 입금 결과를 출력합니다
        print(f"{amount}원 입금 완료. 현재 잔액: {self.balance}원")
    
    def withdraw(self, amount: int) -> bool:
        """
        출금 메서드
        
        매개변수:
            amount: 출금할 금액
        
        반환값:
            bool: 출금 성공 시 True, 실패 시 False
        """
        # 잔액이 부족한지 확인 (비즈니스 로직!)
        if amount > self.balance:
            print("잔액이 부족합니다!")
            return False  # 출금 실패
        else:
            # 잔액에서 출금액을 뺍니다
            self.balance -= amount
            # 출금 결과를 출력합니다
            print(f"{amount}원 출금 완료. 현재 잔액: {self.balance}원")
            return True  # 출금 성공
    
    def get_balance(self) -> int:
        """
        현재 잔액 조회 메서드
        
        반환값:
            int: 현재 잔액
        """
        return self.balance


# 은행 계좌 생성하기
# owner="홍길동", balance=10000으로 초기화
account = BankAccount("홍길동", 10000)

# 계좌 정보 확인
print(f"계좌 소유자: {account.owner}")  
# 출력: 계좌 소유자: 홍길동

print(f"현재 잔액: {account.get_balance()}원")  
# 출력: 현재 잔액: 10000원

# 입금하기
account.deposit(5000)   
# 출력: 5000원 입금 완료. 현재 잔액: 15000원

# 출금하기 (성공)
account.withdraw(3000)  
# 출력: 3000원 출금 완료. 현재 잔액: 12000원

# 출금하기 (실패 - 잔액 부족)
account.withdraw(20000) 
# 출력: 잔액이 부족합니다!
````

**실행 결과:**
```
계좌 소유자: 홍길동
현재 잔액: 10000원
5000원 입금 완료. 현재 잔액: 15000원
3000원 출금 완료. 현재 잔액: 12000원
잔액이 부족합니다!
```

### 🔍 코드 분석

**1. 기본값 매개변수**
```python
def __init__(self, owner: str, balance: int = 0):
```
- `balance`에 기본값 `0`이 있습니다
- 생성할 때 `balance`를 생략하면 자동으로 `0`이 됩니다
- 예: `BankAccount("홍길동")` → `balance`는 `0`

**2. 비즈니스 로직**
```python
if amount > self.balance:
    print("잔액이 부족합니다!")
    return False
```
- 출금하기 전에 **잔액을 검증**합니다
- 이런 검증 로직을 "비즈니스 로직"이라고 부릅니다
- 실무에서 매우 중요합니다!

**3. 상태 변경**
```python
self.balance += amount  # 입금
self.balance -= amount  # 출금
```
- 메서드를 호출할 때마다 객체의 **상태(balance)가 변경**됩니다
- 이것이 객체지향 프로그래밍의 핵심입니다!

## 4️⃣ 실전 예제: 상품 관리 클래스

마지막으로 쇼핑몰의 상품을 관리하는 클래스를 만들어봅시다!

### 무엇을 만들까요?

`Product` 클래스:

**데이터:**
- 상품명 (name)
- 가격 (price)
- 재고 수량 (stock)

**기능:**
- 판매 (sell) - 재고 감소
- 재고 보충 (restock) - 재고 증가

### 실무에서는?

- 쇼핑몰 재고 관리 시스템
- 편의점 POS 시스템
- 창고 관리 시스템

등에서 이런 클래스를 사용합니다!

````python
class Product:
    """상품 정보를 관리하는 클래스"""
    
    def __init__(self, name: str, price: int, stock: int):
        """
        생성자 메서드
        
        매개변수:
            name: 상품명
            price: 가격 (원)
            stock: 재고 수량
        """
        self.name = name    # 상품명 저장
        self.price = price  # 가격 저장
        self.stock = stock  # 재고 수량 저장
    
    def sell(self, quantity: int) -> bool:
        """
        상품 판매 메서드
        
        매개변수:
            quantity: 판매할 수량
        
        반환값:
            bool: 판매 성공 시 True, 실패 시 False
        """
        # 재고가 충분한지 확인 (비즈니스 로직)
        if quantity > self.stock:
            # 재고가 부족하면 판매 불가
            print(f"재고 부족! (현재 재고: {self.stock}개)")
            return False
        else:
            # 재고에서 판매 수량만큼 차감
            self.stock -= quantity
            
            # 총 판매 금액 계산
            total_price = self.price * quantity
            
            # 판매 결과 출력
            print(f"{self.name} {quantity}개 판매 완료!")
            print(f"판매 금액: {total_price}원, 남은 재고: {self.stock}개")
            return True
    
    def restock(self, quantity: int) -> None:
        """
        재고 보충 메서드
        
        매개변수:
            quantity: 입고할 수량
        """
        # 재고에 입고 수량을 더합니다
        self.stock += quantity
        # 입고 결과 출력
        print(f"{quantity}개 입고 완료. 현재 재고: {self.stock}개")


# 상품 객체 생성
# 노트북: 가격 120만원, 재고 5개
laptop = Product("노트북", 1200000, 5)

# 상품 정보 확인
print(f"상품: {laptop.name}, 가격: {laptop.price}원")  
# 출력: 상품: 노트북, 가격: 1200000원

# 노트북 2개 판매
laptop.sell(2)
# 출력: 노트북 2개 판매 완료!
#       판매 금액: 2400000원, 남은 재고: 3개

# 노트북 10개 입고
laptop.restock(10)
# 출력: 10개 입고 완료. 현재 재고: 13개
````

**실행 결과:**
```
상품: 노트북, 가격: 1200000원
노트북 2개 판매 완료!
판매 금액: 2400000원, 남은 재고: 3개
10개 입고 완료. 현재 재고: 13개
```

## 📌 핵심 정리

이번 시간에 배운 내용을 정리해봅시다!

### 1. 클래스와 객체

```python
# 클래스 = 설계도
class User:
    def __init__(self, name: str):
        self.name = name

# 객체 = 설계도로 만든 실제 결과물
user1 = User("홍길동")  # 객체 1
user2 = User("김철수")  # 객체 2
```

- **클래스**: 붕어빵 틀 (설계도)
- **객체/인스턴스**: 실제 붕어빵 (만들어진 것)
- 하나의 클래스로 여러 객체를 만들 수 있습니다

### 2. `__init__` 메서드

```python
def __init__(self, name: str, age: int):
    self.name = name
    self.age = age
```

- 객체가 생성될 때 **자동으로 실행**됩니다
- **인스턴스 변수를 초기화**하는 역할입니다
- 생성자(Constructor)라고도 부릅니다

### 3. self

```python
def greet(self) -> str:
    return f"Hello, {self.name}"  # self로 자신의 변수 접근
```

- **자기 자신(객체)**을 가리킵니다
- 모든 메서드의 **첫 번째 매개변수**는 `self`입니다
- 호출할 때는 `self`를 넘기지 **않습니다** (Python이 자동으로 처리)

### 4. 타입 힌트

```python
class User:
    def __init__(self, name: str, age: int):  # ← 타입 힌트
        self.name = name
        self.age = age
    
    def greet(self) -> str:  # ← 반환 타입
        return f"Hello, {self.name}"
```

- 모든 매개변수에 **타입 힌트** 추가
- 모든 메서드에 **반환 타입** 명시
- FastAPI에서 **필수**입니다!

### 5. 인스턴스 변수 vs 메서드

```python
class User:
    def __init__(self, name: str):
        self.name = name  # ← 인스턴스 변수 (데이터)
    
    def greet(self) -> str:  # ← 메서드 (기능)
        return f"Hello, {self.name}"
```

- **인스턴스 변수**: 객체가 가지는 데이터
- **메서드**: 객체가 할 수 있는 기능
- 클래스는 데이터와 기능을 **하나로 묶습니다**
