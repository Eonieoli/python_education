## 📖 클래스 심화 - 실전 활용으로 한 단계 더!

지난 시간에 우리는 클래스의 기본 개념을 배웠습니다. 이번 시간에는 **실무에서 정말 많이 사용하는 패턴**들을 배워볼 거예요.

### 왜 여러 메서드가 필요할까요?

실제 프로그램은 단순하지 않습니다. 예를 들어 은행 계좌를 생각해보세요:

- 돈을 입금해야 하고 (`deposit`)
- 돈을 출금해야 하고 (`withdraw`)
- 잔액을 확인해야 하고 (`get_balance`)
- 거래 내역을 봐야 하고... (`show_history`)

**하나의 클래스에 여러 개의 메서드**가 필요한 이유가 바로 이것입니다!

### FastAPI에서는 어떻게 쓰일까요?

FastAPI로 쇼핑몰 API를 만든다고 생각해봅시다:

```python
# 장바구니 클래스
class ShoppingCart:
    def add_item(self, product, quantity):     # 상품 추가
        ...
    
    def remove_item(self, product_id):         # 상품 제거
        ...
    
    def get_total(self):                       # 총 금액 계산
        ...
    
    def apply_coupon(self, coupon_code):       # 쿠폰 적용
        ...
```

실무에서는 이렇게 **하나의 객체가 여러 가지 일**을 할 수 있어야 합니다. 오늘 배울 내용은 실제 프로젝트에서 바로 써먹을 수 있는 진짜 실력입니다! 💪

### 인스턴스 변수 vs 클래스 변수는 뭐가 다를까요?

이것도 중요한 개념입니다:

- **인스턴스 변수**: 각 객체마다 다른 값 (예: 사용자마다 다른 이름, 나이)
- **클래스 변수**: 모든 객체가 공유하는 값 (예: 전체 사용자 수)

실무 예시:
```python
class User:
    total_users = 0  # 클래스 변수 - 모든 사용자가 공유
    
    def __init__(self, name):
        self.name = name  # 인스턴스 변수 - 각자 다름
        User.total_users += 1  # 사용자가 생성될 때마다 증가
```

이제 본격적으로 실전 코드를 통해 배워봅시다! 🚀

## 1. 여러 메서드를 가진 클래스 - 은행 계좌 예제

### 실무에서 가장 많이 쓰는 패턴

하나의 클래스에 여러 메서드를 만들어서 **다양한 기능**을 구현할 수 있습니다. 은행 계좌를 예로 들어볼게요.

**이 코드가 보여주는 것**:
- 입금(`deposit`), 출금(`withdraw`), 잔액 조회(`get_balance`) 메서드
- 각 메서드가 `self.balance`를 수정하거나 읽음
- 출금 시 잔액 검사 (비즈니스 로직)
- 명확한 타입 힌트 (`-> None`, `-> bool`, `-> int`)

**실행 전 주목할 포인트**:
- `withdraw` 메서드는 성공/실패를 `bool`로 반환합니다
- 잔액이 부족하면 출금이 실패하고 `False`를 반환합니다
- 모든 메서드가 `self`를 통해 같은 `balance`를 공유합니다

````python
class BankAccount:
    """은행 계좌 클래스 - 여러 메서드 예제"""
    
    def __init__(self, owner: str, balance: int = 0):
        """계좌 초기화
        
        Args:
            owner: 예금주 이름
            balance: 초기 잔액 (기본값 0원)
        """
        self.owner = owner      # 예금주
        self.balance = balance  # 잔액
    
    def deposit(self, amount: int) -> None:
        """입금 메서드
        
        Args:
            amount: 입금할 금액
        """
        self.balance += amount  # 잔액에 입금액 더하기
        print(f"{amount}원 입금 완료. 잔액: {self.balance}원")
    
    def withdraw(self, amount: int) -> bool:
        """출금 메서드 - 성공 여부 반환
        
        Args:
            amount: 출금할 금액
            
        Returns:
            bool: 출금 성공 시 True, 잔액 부족 시 False
        """
        # 잔액 부족 검사 (비즈니스 로직!)
        if amount > self.balance:
            print(f"잔액 부족! 현재 잔액: {self.balance}원")
            return False  # 실패
        
        # 출금 진행
        self.balance -= amount  # 잔액에서 출금액 빼기
        print(f"{amount}원 출금 완료. 잔액: {self.balance}원")
        return True  # 성공
    
    def get_balance(self) -> int:
        """잔액 조회 메서드
        
        Returns:
            int: 현재 잔액
        """
        return self.balance

# 계좌 사용 예시
account = BankAccount("홍길동", 10000)  # 초기 잔액 10,000원으로 시작
account.deposit(5000)                    # 5,000원 입금
account.withdraw(3000)                   # 3,000원 출금 (성공)
account.withdraw(20000)                  # 20,000원 출금 시도 (실패!)
````

**실행 결과:**
```
5000원 입금 완료. 잔액: 15000원
3000원 출금 완료. 잔액: 12000원
잔액 부족! 현재 잔액: 12000원
```

**결과 해석:**
- 초기 10,000원 + 입금 5,000원 = 15,000원
- 15,000원 - 출금 3,000원 = 12,000원
- 12,000원인데 20,000원 출금 시도 → 실패!

### 💡 핵심 포인트

1. **메서드는 각자 역할이 있습니다**
   - `deposit`: 입금만 담당
   - `withdraw`: 출금만 담당 (+ 잔액 검사)
   - `get_balance`: 조회만 담당

2. **반환 타입이 다를 수 있습니다**
   - `-> None`: 반환값 없음 (그냥 동작만)
   - `-> bool`: True/False 반환 (성공/실패)
   - `-> int`: 숫자 반환 (잔액)

3. **모든 메서드가 `self`로 같은 데이터를 공유합니다**
   - `self.balance`는 모든 메서드에서 접근 가능
   - 한 메서드에서 수정하면 다른 메서드에서도 변경된 값 사용

### ⚠️ 일반적인 실수

```python
# 잘못된 예시
def withdraw(self, amount: int) -> bool:
    balance -= amount  # ❌ self를 빼먹으면 에러!
    
# 올바른 예시
def withdraw(self, amount: int) -> bool:
    self.balance -= amount  # ✅ self를 붙여야 인스턴스 변수 접근
```

## 2. 인스턴스 변수 vs 클래스 변수

### 둘의 차이를 확실히 알아야 합니다!

이 개념은 처음에는 헷갈릴 수 있지만, 한 번 이해하면 정말 유용합니다.

**인스턴스 변수**: `self.변수명` - 각 객체마다 **독립적인 값**
- 사용자마다 다른 이름, 이메일
- 계좌마다 다른 잔액
- 상품마다 다른 가격

**클래스 변수**: `클래스명.변수명` - 모든 객체가 **공유하는 값**
- 전체 사용자 수
- 상품의 전체 개수
- 회사의 이름

**FastAPI에서는 언제 쓸까요?**
```python
# 예: API 요청 횟수를 추적
class APIEndpoint:
    total_requests = 0  # 클래스 변수 - 전체 요청 수
    
    def __init__(self, path):
        self.path = path  # 인스턴스 변수 - 각 엔드포인트의 경로
        APIEndpoint.total_requests += 1
```

이제 코드로 확인해봅시다!

````python
class User:
    """사용자 클래스 - 클래스 변수 예제"""
    
    # 클래스 변수: 모든 인스턴스가 공유하는 변수
    # __init__ 밖에서 정의됩니다!
    total_users = 0
    
    def __init__(self, name: str, email: str):
        """사용자 초기화
        
        Args:
            name: 사용자 이름
            email: 이메일 주소
        """
        # 인스턴스 변수: 각 인스턴스마다 고유한 변수
        # self를 붙여서 정의합니다!
        self.name = name
        self.email = email
        
        # 클래스 변수 증가 (새 사용자가 생성될 때마다)
        # 주의: User.total_users로 접근!
        User.total_users += 1
    
    def get_info(self) -> str:
        """사용자 정보 반환
        
        Returns:
            str: "이름 (이메일)" 형식의 문자열
        """
        return f"{self.name} ({self.email})"

# 사용자 3명 생성
user1 = User("김철수", "kim@example.com")
user2 = User("이영희", "lee@example.com")
user3 = User("박민수", "park@example.com")

# 각 인스턴스는 고유한 정보를 가짐 (인스턴스 변수)
print(user1.get_info())  # 김철수의 정보
print(user2.get_info())  # 이영희의 정보

# 클래스 변수는 모든 인스턴스가 공유
print(f"\n전체 사용자 수: {User.total_users}명")  # 3명!
````

**실행 결과:**
```
김철수 (kim@example.com)
이영희 (lee@example.com)

전체 사용자 수: 3명
```

**결과 해석:**
- 각 사용자는 자기만의 이름과 이메일을 가집니다 (인스턴스 변수)
- 하지만 `total_users`는 모두가 공유하며, 사용자가 생성될 때마다 1씩 증가합니다

### 💡 핵심 포인트

1. **인스턴스 변수**
   ```python
   def __init__(self, name):
       self.name = name  # ✅ 각 객체마다 다른 값
   ```

2. **클래스 변수**
   ```python
   class User:
       total_users = 0  # ✅ 모든 객체가 공유
   ```

3. **접근 방법의 차이**
   ```python
   # 인스턴스 변수: self.변수명
   self.name = "홍길동"
   
   # 클래스 변수: 클래스명.변수명
   User.total_users += 1
   ```

### 📊 비교표

| 구분 | 인스턴스 변수 | 클래스 변수 |
|------|--------------|-------------|
| **정의 위치** | `__init__` 안 | 클래스 블록 최상단 |
| **접근 방법** | `self.변수명` | `클래스명.변수명` |
| **값의 특징** | 각 객체마다 다름 | 모든 객체가 공유 |
| **사용 예시** | 이름, 나이, 잔액 | 전체 개수, 설정값 |

### ⚠️ 일반적인 실수

```python
# 실수 1: 클래스 변수를 self로 접근
User.total_users += 1  # ✅ 올바름
self.total_users += 1  # ⚠️ 동작은 하지만 권장하지 않음

# 실수 2: 클래스 변수를 __init__ 안에 정의
class User:
    def __init__(self):
        total_users = 0  # ❌ 이건 그냥 지역 변수!
```

## 3. 실전 패턴: 온라인 쇼핑몰 장바구니

### 실무에서 정말 많이 사용하는 패턴!

이제 지금까지 배운 모든 것을 종합해서 **실전 프로젝트**에서 사용할 법한 코드를 만들어봅시다.

**온라인 쇼핑몰의 장바구니 클래스**입니다:
- 상품을 추가하고 (`add_item`)
- 총 금액을 계산하고 (`get_total`)
- 장바구니 내역을 보여줍니다 (`show_items`)

**이 코드의 핵심**:
- 리스트(`List`)와 딕셔너리(`Dict`)를 타입 힌트로 사용
- 실제 쇼핑몰 로직 구현
- 여러 메서드가 협력하는 구조

**FastAPI에서는?**
```python
# 이런 식으로 사용됩니다
@app.post("/cart/add")
def add_to_cart(item: CartItem):
    cart = ShoppingCart()
    cart.add_item(item.name, item.price, item.quantity)
    return {"total": cart.get_total()}
```

자, 코드를 봅시다!

````python
from typing import List, Dict

class ShoppingCart:
    """온라인 쇼핑몰 장바구니 클래스"""
    
    def __init__(self):
        """장바구니 초기화 - 빈 리스트로 시작"""
        # 타입 힌트: 딕셔너리들의 리스트
        # 각 딕셔너리는 {"name": str, "price": int, "quantity": int}
        self.items: List[Dict[str, int]] = []
    
    def add_item(self, name: str, price: int, quantity: int = 1) -> None:
        """상품 추가
        
        Args:
            name: 상품명
            price: 가격
            quantity: 수량 (기본값 1개)
        """
        # 상품 정보를 딕셔너리로 만들기
        item = {
            "name": name,
            "price": price,
            "quantity": quantity
        }
        # 장바구니에 추가
        self.items.append(item)
        print(f"{name} {quantity}개 추가됨")
    
    def get_total(self) -> int:
        """총 금액 계산
        
        Returns:
            int: 장바구니의 총 금액
        """
        total = 0
        # 모든 상품을 순회하며 금액 계산
        for item in self.items:
            # 가격 × 수량을 총액에 더하기
            total += item["price"] * item["quantity"]
        return total
    
    def show_items(self) -> None:
        """장바구니 내역 출력"""
        print("\n=== 장바구니 ===")
        
        # 각 상품의 정보를 보기 좋게 출력
        for item in self.items:
            # 소계 = 가격 × 수량
            subtotal = item["price"] * item["quantity"]
            print(f"{item['name']}: {item['price']}원 x {item['quantity']}개 = {subtotal}원")
        
        # 총 금액 출력
        print(f"총 금액: {self.get_total()}원\n")

# 장바구니 사용 예시
cart = ShoppingCart()
cart.add_item("노트북", 1500000)      # 노트북 1개
cart.add_item("마우스", 30000, 2)     # 마우스 2개
cart.add_item("키보드", 80000)        # 키보드 1개
cart.show_items()                      # 장바구니 내역 확인
````

**실행 결과:**
```
노트북 1개 추가됨
마우스 2개 추가됨
키보드 1개 추가됨

=== 장바구니 ===
노트북: 1500000원 x 1개 = 1500000원
마우스: 30000원 x 2개 = 60000원
키보드: 80000원 x 1개 = 80000원
총 금액: 1640000원
```

**결과 해석:**
- 노트북 150만원 + 마우스 6만원(3만원×2) + 키보드 8만원 = 총 164만원
- `get_total()` 메서드가 자동으로 계산해줍니다

### 💡 핵심 포인트

1. **타입 힌트의 중요성**
   ```python
   # 명확한 타입 정의
   self.items: List[Dict[str, int]] = []
   # "딕셔너리들의 리스트야!" 라고 명확히 알려줌
   ```

2. **여러 메서드가 협력합니다**
   - `add_item()`: 상품을 `self.items`에 추가
   - `get_total()`: `self.items`를 읽어서 총액 계산
   - `show_items()`: `self.items`와 `get_total()`을 사용해 출력

3. **실무 패턴**
   - 데이터는 딕셔너리로 저장 (JSON과 호환!)
   - 리스트로 여러 개 관리
   - 타입 힌트로 명확성 확보

### 🚀 실무 확장 아이디어

이 클래스를 실무에서는 이렇게 확장할 수 있습니다:

```python
class ShoppingCart:
    # ... 기존 메서드들 ...
    
    def remove_item(self, name: str) -> bool:
        """상품 제거"""
        ...
    
    def apply_discount(self, discount_rate: float) -> int:
        """할인 적용"""
        ...
    
    def to_json(self) -> dict:
        """JSON으로 변환 (FastAPI 응답용)"""
        return {
            "items": self.items,
            "total": self.get_total()
        }
```

### ⚠️ 주의사항
