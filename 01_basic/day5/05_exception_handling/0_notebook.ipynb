## 📖 예외 처리(Exception Handling)란 무엇인가?

### 프로그램에서 "오류"는 피할 수 없습니다

여러분이 아무리 완벽하게 코드를 작성해도, 실행 중에 **예상치 못한 상황**은 반드시 발생합니다:

- 사용자가 숫자를 입력해야 하는데 문자를 입력했을 때
- 존재하지 않는 파일을 열려고 할 때
- 0으로 나누기를 시도할 때
- 네트워크 연결이 끊어졌을 때
- 데이터베이스 연결이 실패했을 때

이런 상황에서 **프로그램이 갑자기 멈추면** 어떻게 될까요? 🤔

### 예외 처리가 없다면?

```python
# 예외 처리가 없는 코드
result = 10 / 0  # ❌ ZeroDivisionError 발생!
print("이 줄은 절대 실행되지 않습니다")  # 프로그램 종료
```

프로그램이 **즉시 종료**되고, 사용자는 당황하게 됩니다. 😱

### 예외 처리를 하면?

```python
# 예외 처리가 있는 코드
try:
    result = 10 / 0
except ZeroDivisionError:
    print("0으로 나눌 수 없습니다. 다시 시도해주세요.")  # ✅ 친절한 안내

print("프로그램이 계속 실행됩니다!")  # 정상 실행
```

프로그램이 **안전하게 계속 실행**되고, 사용자에게 **친절한 안내**를 제공합니다. ✨

## 🌐 실무에서 예외 처리가 중요한 이유

### 1. 사용자 경험(UX)

웹 서비스에서 오류가 발생했을 때:

**예외 처리 없음:**
```
500 Internal Server Error
(화면이 하얗게 나옴)
```

**예외 처리 있음:**
```json
{
  "error": "이메일 형식이 올바르지 않습니다",
  "message": "올바른 이메일 주소를 입력해주세요"
}
```

어느 쪽이 더 좋은 경험일까요?

### 2. FastAPI에서의 예외 처리

FastAPI는 예외 처리를 **매우 중요하게** 다룹니다:

```python
from fastapi import HTTPException

@app.get("/users/{user_id}")
def get_user(user_id: int):
    user = find_user(user_id)  # 사용자 검색
    
    if user is None:
        # ✅ 404 에러를 명확하게 반환
        raise HTTPException(
            status_code=404,
            detail="사용자를 찾을 수 없습니다"
        )
    
    return user
```

이렇게 하면 클라이언트(프론트엔드)가 **정확한 에러 정보**를 받아서 처리할 수 있습니다!

### 3. 디버깅과 유지보수

예외 처리를 제대로 하면:

- **어디서** 오류가 발생했는지 정확히 알 수 있습니다
- **왜** 오류가 발생했는지 로그로 기록할 수 있습니다
- **어떻게** 해결해야 하는지 빠르게 파악할 수 있습니다

## 🔑 핵심 개념들

### Exception(예외)이란?

프로그램 실행 중에 발생하는 **오류 상황**을 Python 객체로 표현한 것입니다.

**주요 예외 종류:**

| 예외 이름 | 발생 상황 | 예시 |
|----------|---------|-----|
| `ZeroDivisionError` | 0으로 나누기 | `10 / 0` |
| `ValueError` | 잘못된 값 | `int("abc")` |
| `FileNotFoundError` | 파일 없음 | `open("없는파일.txt")` |
| `KeyError` | 딕셔너리 키 없음 | `dict["없는키"]` |
| `IndexError` | 리스트 인덱스 범위 초과 | `list[999]` |
| `TypeError` | 타입 오류 | `"2" + 2` |

### try-except의 동작 방식

```python
try:
    # 오류가 발생할 수 있는 코드
    위험한_작업()
except 예외종류:
    # 오류가 발생했을 때 실행할 코드
    안전하게_처리()
```

**실행 순서:**
1. `try` 블록의 코드를 실행합니다
2. 오류가 **없으면** → `except` 블록을 건너뛰고 계속 진행
3. 오류가 **발생하면** → 즉시 `except` 블록으로 이동하여 실행

### 💡 기억할 점

- 예외 처리는 **방어적 프로그래밍**의 핵심입니다
- 모든 예외를 막을 필요는 없지만, **예상 가능한 오류**는 꼭 처리하세요
- FastAPI를 배우면 `HTTPException`을 많이 사용하게 됩니다
- 실무에서는 예외를 **로그로 기록**하는 것이 매우 중요합니다

## 1️⃣ 기본 try-except: 안전한 나눗셈

### 0으로 나누기는 프로그램의 대표적인 오류입니다

수학적으로 0으로 나누는 것은 **정의되지 않은 연산**입니다. Python에서 이를 시도하면 `ZeroDivisionError`가 발생합니다.

### 예외 처리 없이 실행하면?

```python
# 이렇게 하면 프로그램이 멈춥니다!
result = 10 / 0  # ❌ ZeroDivisionError!
```

### try-except로 안전하게 처리하기

이제 `try-except`를 사용해서 **안전한 나눗셈 함수**를 만들어봅시다:

````python
# 기본 try-except 예제

def divide(a: int, b: int) -> float:
    """두 숫자를 나누는 안전한 함수
    
    Args:
        a: 나누어질 수 (분자)
        b: 나누는 수 (분모)
        
    Returns:
        나눗셈 결과. 0으로 나누면 0.0 반환
    """
    try:
        # 나눗셈 시도
        result = a / b
        return result
    except ZeroDivisionError:
        # 0으로 나누면 이 블록이 실행됩니다
        print("❌ 오류: 0으로 나눌 수 없습니다")
        return 0.0  # 안전한 기본값 반환

# 테스트
print("=== 기본 try-except ===")
print(divide(10, 2))  # 정상 케이스
print(divide(10, 0))  # 예외 케이스 - 프로그램이 멈추지 않습니다!
````

**실행 결과:**
```
=== 기본 try-except ===
5.0
❌ 오류: 0으로 나눌 수 없습니다
0.0
```

### 🔍 코드 분석

**정상 케이스 (`divide(10, 2)`):**
1. `try` 블록 진입
2. `result = 10 / 2` → 성공! `result = 5.0`
3. `return result` → `5.0` 반환
4. `except` 블록은 **실행되지 않음**

**예외 케이스 (`divide(10, 0)`):**
1. `try` 블록 진입
2. `result = 10 / 0` → ❌ `ZeroDivisionError` 발생!
3. 즉시 `except ZeroDivisionError` 블록으로 이동
4. 에러 메시지 출력
5. `return 0.0` → 안전한 값 반환

### 💡 왜 이렇게 작성할까요?

- **사용자 경험**: "프로그램 오류"보다 "0으로 나눌 수 없습니다"가 훨씬 친절합니다
- **프로그램 안정성**: 한 번의 오류로 전체 프로그램이 멈추지 않습니다
- **기본값 제공**: `0.0`을 반환해서 후속 코드가 계속 실행될 수 있습니다

### ⚠️ 주의사항

- `except` 뒤에는 **구체적인 예외 이름**을 써야 합니다 (`ZeroDivisionError`)
- `except:` 만 쓰면 **모든 예외**를 잡아버리므로 권장하지 않습니다
- 예외를 잡았다고 해서 **문제가 해결된 것은 아닙니다** (적절한 조치가 필요)

## 2️⃣ Exception as e: 에러 정보 활용하기

### 에러가 발생했을 때, 구체적으로 "무엇이" 잘못되었는지 알 수 있다면?

예외 객체는 **오류에 대한 상세 정보**를 담고 있습니다. `as e`를 사용하면 이 정보를 가져올 수 있습니다!

### 사용자가 "abc"를 숫자로 변환하려 한다면?

```python
int("abc")  # ❌ ValueError: invalid literal for int() with base 10: 'abc'
```

이 에러 메시지에는 **무엇이 잘못되었는지**에 대한 중요한 정보가 들어있습니다!

````python
# Exception as e 예제

def process_data(data: str) -> int:
    """문자열을 정수로 안전하게 변환하는 함수
    
    Args:
        data: 변환할 문자열 (예: "123")
        
    Returns:
        변환된 정수. 변환 실패 시 0 반환
    """
    try:
        # 문자열을 정수로 변환 시도
        number = int(data)
        return number
    except ValueError as e:
        # 예외 객체를 'e'라는 변수로 받아옵니다
        # 'e'에는 에러에 대한 상세 정보가 들어있습니다
        print(f"❌ 변환 실패: {e}")
        return 0

# 테스트
print("=== Exception as e ===")
print(process_data("123"))   # 정상: "123" → 123
print(process_data("abc"))   # 오류: "abc"는 숫자가 아님
print(process_data("12.5"))  # 오류: 소수점이 있음 (int()는 정수만 가능)
````

**실행 결과:**
```
=== Exception as e ===
123
❌ 변환 실패: invalid literal for int() with base 10: 'abc'
0
❌ 변환 실패: invalid literal for int() with base 10: '12.5'
0
```

### 🔍 `as e`가 없었다면?

```python
except ValueError:
    print("❌ 변환 실패했습니다")  # 무엇이 문제인지 모름
```

**vs**

```python
except ValueError as e:
    print(f"❌ 변환 실패: {e}")  # 구체적인 에러 원인을 알 수 있음
```

### 💡 실무에서 `as e`를 사용하는 이유

**1. 디버깅이 쉬워집니다:**
```python
except ValueError as e:
    logger.error(f"데이터 변환 오류: {e}")  # 로그에 상세 정보 기록
```

**2. 사용자에게 친절한 안내:**
```python
except ValueError as e:
    return {
        "error": "입력값 오류",
        "detail": str(e),  # 구체적인 오류 내용 전달
        "input": data
    }
```

**3. FastAPI에서의 활용:**
```python
except ValueError as e:
    raise HTTPException(
        status_code=400,
        detail=f"잘못된 데이터: {e}"
    )
```

### 📝 변수 이름은 꼭 `e`여야 할까요?

아닙니다! 다른 이름도 가능합니다:

```python
except ValueError as error:     # ✅ OK
except ValueError as err:       # ✅ OK
except ValueError as exception: # ✅ OK
```

하지만 대부분 **`e`** 또는 **`error`**를 사용합니다. (관례)

## 3️⃣ finally: 반드시 실행되어야 하는 코드

### 파일을 열었으면 반드시 닫아야 합니다!

파일을 열고(`open`) 사용한 후에는 **반드시** 닫아야(`close`) 합니다. 그렇지 않으면:

- 💾 메모리 누수 (Memory Leak)
- 🔒 파일 잠금 (다른 프로그램이 접근 불가)
- 📉 시스템 리소스 낭비

### 문제 상황: 에러가 발생하면 close()가 실행 안 됨

```python
f = open("file.txt", 'r')
content = f.read()  # 여기서 에러 발생하면?
f.close()  # ❌ 이 줄은 실행 안 됨!
```

### finally가 해결책입니다!

`finally` 블록은 **try 블록에서 무슨 일이 일어나든** 상관없이 **항상 실행**됩니다.

````python
# finally 예제

def read_file(filename: str) -> str:
    """파일을 안전하게 읽는 함수
    
    Args:
        filename: 읽을 파일 이름
        
    Returns:
        파일 내용. 파일이 없으면 빈 문자열
    """
    f = None  # 파일 객체를 미리 선언 (finally에서 사용하기 위해)
    
    try:
        # 파일 열기 시도
        f = open(filename, 'r', encoding='utf-8')
        content = f.read()
        return content
    
    except FileNotFoundError:
        # 파일이 없을 때
        print(f"❌ 파일을 찾을 수 없습니다: {filename}")
        return ""
    
    finally:
        # ✅ 항상 실행됩니다! (성공이든 실패든)
        # 파일이 열렸다면 반드시 닫습니다
        if f is not None:
            f.close()
            print("✅ 파일을 안전하게 닫았습니다")

# 테스트 (실제로 파일이 없어도 finally는 실행됨)
print("=== finally ===")
result = read_file("없는파일.txt")
print(f"읽은 내용: '{result}'")
````

**실행 결과:**
```
=== finally ===
❌ 파일을 찾을 수 없습니다: 없는파일.txt
✅ 파일을 안전하게 닫았습니다
읽은 내용: ''
```

### 🔍 실행 흐름 분석

**파일이 없는 경우:**
1. `try` 블록 진입
2. `f = open(...)` → ❌ `FileNotFoundError` 발생!
3. `except FileNotFoundError` 블록 실행
4. **`finally` 블록 실행** ← 항상!
5. `return ""` 실행

**파일이 있는 경우:**
1. `try` 블록 진입
2. `f = open(...)` → ✅ 성공
3. `content = f.read()` → 내용 읽기
4. **`finally` 블록 실행** ← 항상!
5. `return content` 실행

### 💡 finally의 3가지 보장

1. **성공해도** 실행됩니다
2. **실패해도** 실행됩니다
3. **`return`이 있어도** 실행됩니다!

```python
try:
    return "성공"  # return이 있지만...
finally:
    print("이것도 실행됩니다!")  # ✅ return 전에 실행!
```

### 🌟 더 간편한 방법: with문 (참고)

Python에는 `finally`를 자동으로 처리해주는 `with`문이 있습니다:

```python
# with문을 사용하면 자동으로 close()가 호출됩니다
with open("file.txt", 'r') as f:
    content = f.read()
# 여기서 자동으로 f.close()가 호출됨!
```

### ⚠️ 실무에서 finally를 사용하는 경우

- 파일 닫기
- 데이터베이스 연결 닫기
- 네트워크 연결 닫기
- 잠금(Lock) 해제
- 로그 기록

**공통점:** 반드시 정리(cleanup)해야 하는 리소스!

## 4️⃣ raise: 의도적으로 예외 발생시키기

### 때로는 우리가 직접 "오류"를 만들어야 합니다

지금까지는 Python이 자동으로 발생시키는 예외를 **잡는** 방법을 배웠습니다. 이제는 우리가 **직접** 예외를 발생시키는 방법을 배워봅시다!

### 왜 일부러 오류를 만들까요? 🤔

**예시 1: 입력값 검증**
```python
# 나이는 음수일 수 없습니다
age = -5  # ❌ 이건 명백히 잘못된 값!
```

**예시 2: 비즈니스 규칙**
```python
# 잔액이 부족하면 출금할 수 없습니다
if 출금액 > 잔액:
    # 오류를 발생시켜서 잘못된 동작을 막아야 합니다!
```

`raise`를 사용하면 이런 상황에서 **명확한 오류 메시지**와 함께 프로그램을 중단시킬 수 있습니다.

````python
# raise 예제

def validate_age(age: int) -> None:
    """나이가 유효한지 검증하는 함수
    
    Args:
        age: 검증할 나이
        
    Raises:
        ValueError: 나이가 유효하지 않을 때
    """
    if age < 0:
        # ValueError를 직접 발생시킵니다!
        # 메시지를 같이 전달할 수 있습니다
        raise ValueError("나이는 0 이상이어야 합니다")
    
    if age > 150:
        raise ValueError("나이가 너무 많습니다")
    
    # 모든 검증을 통과하면 정상 메시지 출력
    print(f"✅ 유효한 나이: {age}세")

# 테스트
print("=== raise로 예외 발생 ===")

try:
    validate_age(25)   # ✅ 정상
    validate_age(-5)   # ❌ ValueError 발생!
except ValueError as e:
    print(f"❌ 검증 실패: {e}")

try:
    validate_age(200)  # ❌ ValueError 발생!
except ValueError as e:
    print(f"❌ 검증 실패: {e}")
````

**실행 결과:**
```
=== raise로 예외 발생 ===
✅ 유효한 나이: 25세
❌ 검증 실패: 나이는 0 이상이어야 합니다
❌ 검증 실패: 나이가 너무 많습니다
```

### 🔍 raise의 동작 방식

```python
raise ValueError("에러 메시지")
```

이 코드를 실행하면:
1. `ValueError` 예외 객체를 생성합니다
2. "에러 메시지"를 예외 객체에 담습니다
3. 예외를 **던집니다(throw)** → 프로그램 실행이 즉시 중단!
4. 가장 가까운 `except` 블록을 찾아갑니다

### 💡 어떤 예외를 발생시켜야 할까요?

|-----|-----------|-----|

| 잘못된 타입 | `TypeError` | `raise TypeError("문자열이어야 합니다")` |

## 5️⃣ 클래스 메서드에서 예외 처리하기

### 실전에서는 클래스 메서드에서 예외 처리를 많이 합니다!

은행 계좌를 생각해봅시다. 계좌에서 돈을 인출할 때:

- ✅ 잔액이 충분하면 → 인출 성공
- ❌ 잔액이 부족하면 → 오류 발생!

이런 비즈니스 로직을 **클래스 메서드**에서 `raise`로 구현할 수 있습니다.

````python
# 클래스 메서드 예외 처리 예제

class BankAccount:
    """은행 계좌를 나타내는 클래스"""
    
    def __init__(self, balance: int):
        """초기 잔액으로 계좌 생성
        
        Args:
            balance: 초기 잔액
        """
        self.balance = balance
    
    def withdraw(self, amount: int) -> None:
        """계좌에서 돈을 인출하는 메서드
        
        Args:
            amount: 인출할 금액
            
        Raises:
            ValueError: 잔액이 부족할 때
        """
        # 잔액 검증
        if amount > self.balance:
            # 잔액 부족 시 예외 발생
            raise ValueError(
                f"잔액 부족! (잔액: {self.balance}원, 요청: {amount}원)"
            )
        
        # 검증 통과 → 인출 실행
        self.balance -= amount
        print(f"✅ {amount}원 인출 완료 (잔액: {self.balance}원)")
    
    def deposit(self, amount: int) -> None:
        """계좌에 돈을 입금하는 메서드
        
        Args:
            amount: 입금할 금액
            
        Raises:
            ValueError: 입금액이 0 이하일 때
        """
        # 입금액 검증
        if amount <= 0:
            raise ValueError("입금액은 0보다 커야 합니다")
        
        # 검증 통과 → 입금 실행
        self.balance += amount
        print(f"✅ {amount}원 입금 완료 (잔액: {self.balance}원)")

# 테스트
print("=== 클래스 메서드 예외 처리 ===")
account = BankAccount(10000)  # 초기 잔액 10,000원

try:
    account.withdraw(3000)   # ✅ 3,000원 인출
    account.deposit(5000)    # ✅ 5,000원 입금
    account.withdraw(20000)  # ❌ 20,000원 인출 시도 → 잔액 부족!
except ValueError as e:
    print(f"❌ 거래 실패: {e}")

print(f"\n최종 잔액: {account.balance}원")
````

**실행 결과:**
```
=== 클래스 메서드 예외 처리 ===
✅ 3000원 인출 완료 (잔액: 7000원)
✅ 5000원 입금 완료 (잔액: 12000원)
❌ 거래 실패: 잔액 부족! (잔액: 12000원, 요청: 20000원)

최종 잔액: 12000원
```

### 🔍 클래스 메서드의 예외 처리 패턴

**1단계: 입력값 검증**
```python
if amount > self.balance:  # 조건 확인
    raise ValueError("...")
```

**2단계: 정상 로직 실행**
```python
self.balance -= amount  # 검증 통과 후 실행
```

**3단계: 호출하는 쪽에서 예외 처리**
```python
try:
    account.withdraw(20000)
except ValueError as e:
    print(f"거래 실패: {e}")
```

### 💡 왜 이렇게 설계할까요?

**잘못된 상태를 막습니다:**
```python
# 예외 처리가 없다면?
self.balance -= 20000  # balance = -8000 ← 말이 안 되는 상태!
```

**명확한 책임 분리:**
- `withdraw()` 메서드: "잔액이 부족하면 오류를 발생시킨다"
- 호출하는 코드: "오류를 적절히 처리한다"

FastAPI에서도 똑같은 패턴을 사용합니다:

### 📝 docstring의 Raises 섹션

클래스 메서드가 예외를 발생시킬 수 있다면, **docstring에 명시**하세요:


이렇게 하면 다른 개발자(또는 미래의 나)가 **어떤 예외를 처리해야 하는지** 쉽게 알 수 있습니다!

## 📌 핵심 정리

이번 시간에 배운 예외 처리의 핵심 내용:

### 1. 기본 try-except
```python
try:
    위험한_작업()
except 예외타입:
    안전하게_처리()
```
- 예상 가능한 오류를 잡아서 프로그램이 멈추지 않게 합니다
- **구체적인** 예외 타입을 지정하세요

### 2. Exception as e
```python
except ValueError as e:
    print(f"오류: {e}")  # 상세 정보 활용
```
- 예외 객체를 변수로 받아서 에러 정보를 활용합니다
- 디버깅과 로깅에 필수적입니다

### 3. finally
```python
try:
    파일_열기()
finally:
    파일_닫기()  # 항상 실행!
```
- 성공하든 실패하든 **반드시** 실행되어야 하는 코드
- 리소스 정리에 필수적입니다

### 4. raise
```python
if 잘못된_입력:
    raise ValueError("명확한 에러 메시지")
```
- 입력값 검증, 비즈니스 규칙 위반 시 사용
- FastAPI의 `HTTPException`도 같은 원리!

### 5. 클래스 메서드 예외 처리
```python
class MyClass:
    def my_method(self):
        if 조건:
            raise ValueError("...")
```
- 메서드에서 검증 로직을 담당
- 호출하는 쪽에서 `try-except`로 처리
