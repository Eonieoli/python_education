<details>
<summary>정답 보기</summary>

```python
def divide(a: float, b: float) -> float:
    try:
        return a / b
    except ZeroDivisionError:
        print("0으로 나눌 수 없습니다")
        return 0.0
```

### 🎯 학습 포인트

#### 1. BaseModel 패턴 (Pydantic 미리보기)

FastAPI는 Pydantic의 BaseModel을 사용합니다:

```python
# 실제 FastAPI/Pydantic 코드
from pydantic import BaseModel

class UserCreate(BaseModel):
    email: str
    password: str

class UserResponse(BaseModel):
    email: str
    id: int
    # 비밀번호 없음! (보안)
```

💡 **왜 모델을 분리하나요?**
- **UserCreate**: 회원가입 요청 (비밀번호 포함)
- **UserResponse**: API 응답 (비밀번호 제외)
- **보안**: 응답에 민감한 정보 포함 안 됨

#### 2. 상속 구조로 코드 재사용

```python
class BaseModel:              # 최상위 부모
    def to_dict(self): ...    # 모든 모델이 사용

class UserBase(BaseModel):    # 공통 필드
    email: str

class UserCreate(UserBase):   # 요청용
    password: str

class UserResponse(UserBase): # 응답용
    id: int
    # password 없음!
```

💡 **상속의 장점:**
- email 필드는 한 번만 정의
- UserBase를 수정하면 모든 자식 클래스에 반영
- DRY 원칙 (Don't Repeat Yourself)

#### 3. 검증 로직의 중앙 관리

```python
# validators.py
def validate_password(password: str) -> str:
    """비밀번호 검증 - 모든 곳에서 동일한 규칙 적용!"""
    if len(password) < 8:
        raise ValueError("비밀번호는 8자 이상이어야 합니다")
    if not any(c.isupper() for c in password):
        raise ValueError("비밀번호는 대문자를 포함해야 합니다")
    return password
```

💡 **중앙 관리의 이점:**
- 모든 User 객체가 동일한 검증 규칙 적용
- 규칙 변경 시 한 곳만 수정
- 테스트 용이

#### 4. 실제 FastAPI 사용 예시

```python
# 실제 FastAPI 코드 (참고용)
from fastapi import FastAPI
from pydantic import BaseModel

app = FastAPI()

@app.post("/users", response_model=UserResponse)
def create_user(user: UserCreate):
    # user는 UserCreate 타입 (검증 자동!)
    # 응답은 UserResponse 타입 (비밀번호 자동 제외!)
    return UserResponse(email=user.email, id=1)
```

💡 **FastAPI의 자동 기능:**
- 요청 데이터 자동 검증
- 타입 자동 변환
- 응답 데이터 자동 필터링
- API 문서 자동 생성
