## 📚 Python(파이썬)이란?

**Python(파이썬)**은 1991년 네덜란드의 프로그래머 귀도 반 로섬(Guido van Rossum)이 만든 **고급 프로그래밍 언어**입니다.

### 이름의 유래

많은 분들이 "뱀(Python)" 때문에 지어진 이름이라고 생각하시지만, 사실은 영국의 코미디 그룹 "Monty Python"에서 따온 이름입니다. 귀도는 프로그래밍이 재미있고 즐거워야 한다고 생각했기 때문에, 자신이 좋아하던 코미디 그룹의 이름을 붙였다고 합니다. 🎭

## ✨ Python의 특징

### 1. 읽기 쉬운 코드 (Readable)

Python은 **영어 문장처럼 읽히는** 코드를 지향합니다. 마치 책을 읽듯이 코드를 이해할 수 있습니다.

```python
# 다른 언어들
if (x > 0) {
    System.out.println("양수입니다");
}

# Python
if x > 0:
    print("양수입니다")
```

훨씬 간결하고 읽기 쉽죠? 이것이 Python의 철학입니다.

### 2. 배우기 쉬운 언어 (Easy to Learn)

- **문법이 단순**해서 프로그래밍 초보자도 빠르게 배울 수 있습니다
- 복잡한 기호나 규칙이 적습니다
- 들여쓰기로 코드 블록을 구분하기 때문에 **자연스럽게 깔끔한 코드**를 작성하게 됩니다

### 3. 강력한 라이브러리 (Powerful Libraries)

Python은 **"배터리가 포함된(Batteries Included)"** 언어라고 불립니다. 이미 만들어진 수많은 도구(라이브러리)들이 있어서, 바퀴를 다시 발명할 필요가 없습니다.

- 웹 개발: **Django, FastAPI, Flask**
- 데이터 분석: **Pandas, NumPy**
- 인공지능: **TensorFlow, PyTorch**
- 자동화: **Selenium, Beautiful Soup**

### 4. 인터프리터 언어 (Interpreted Language)

Python은 코드를 **한 줄씩 실행**합니다. 컴파일 과정 없이 바로 실행할 수 있어서 개발 속도가 빠릅니다.

### 5. 플랫폼 독립적 (Cross-Platform)

Windows, Mac, Linux 어디에서나 동일하게 실행됩니다. 한 번 작성한 코드를 어디서든 사용할 수 있습니다! 🌍

## 🎯 Python의 철학 - The Zen of Python

Python 커뮤니티에는 **"The Zen of Python"**이라는 19가지 철학이 있습니다. 그 중 가장 유명한 몇 가지를 소개합니다:

> **"아름다운 것이 추한 것보다 낫다."**  
> Beautiful is better than ugly.

> **"명시적인 것이 암시적인 것보다 낫다."**  
> Explicit is better than implicit.

> **"단순한 것이 복잡한 것보다 낫다."**  
> Simple is better than complex.

> **"가독성은 중요하다."**  
> Readability counts.

이러한 철학 덕분에 Python 코드는 **아름답고, 명확하고, 단순**합니다.

💡 **Tip**: Python 인터프리터에서 `import this`를 실행하면 전체 19가지 철학을 볼 수 있습니다!

## 1. Hello, World! - 첫 번째 Python 프로그램

### print() 함수란?

`print()`는 Python에서 가장 많이 사용하는 **내장 함수**(Built-in Function)입니다.
- **역할**: 괄호 안의 내용을 화면에 출력합니다
- **문법**: `print(출력할내용)`
- **특징**: 문자열은 따옴표로 감싸야 합니다

### 실행해보기

아래 코드를 실행하면 화면에 "Hello, World!"가 출력됩니다:

````python
# print() 함수는 화면에 글자를 출력합니다
print("Hello, World!")  # 출력: Hello, World!
````

**실행 결과:**
```
Hello, World!
```

### 코드 해설

- `print`: 출력 함수의 이름
- `()`: 함수를 호출할 때 사용하는 괄호
- `"Hello, World!"`: 출력할 문자열 (큰따옴표로 감쌈)
- `# 출력: Hello, World!`: 주석 (코드 설명, 실행되지 않음)

### 한글도 출력할 수 있나요?

물론입니다! Python은 **유니코드를 완벽하게 지원**하기 때문에 한글, 중국어, 일본어, 이모지 등 모든 문자를 출력할 수 있습니다:

````python
# 한글도 출력할 수 있습니다
print("안녕하세요, Python!")  # 출력: 안녕하세요, Python!
````

**실행 결과:**
```
안녕하세요, Python!
```

### 💡 알아두기

- Python은 **대소문자를 구분**합니다: `print`는 작동하지만, `Print`나 `PRINT`는 오류가 납니다
- 문자열은 **작은따옴표(`'`)** 또는 **큰따옴표(`"`)**로 감쌀 수 있습니다
  - `print('Hello')` ✅
  - `print("Hello")` ✅
  - `print(Hello)` ❌ (따옴표 없으면 오류!)

## 2. 주석(Comment)이란?

### 주석의 역할

**주석(Comment)**은 코드에 대한 **설명이나 메모**를 남기는 방법입니다.

**주석의 특징:**
- Python 인터프리터가 **실행하지 않습니다** (무시됨)
- 코드의 동작에 영향을 주지 않습니다
- 다른 사람(또는 미래의 나)이 코드를 이해하기 쉽게 해줍니다

### 왜 주석이 필요한가요?

1. **코드 설명**: 복잡한 로직을 설명
2. **메모**: 나중에 수정할 부분 표시
3. **임시 비활성화**: 코드를 삭제하지 않고 일시적으로 실행 안 되게 함
4. **협업**: 팀원들에게 코드 의도 전달

### 한 줄 주석: # (샵, 해시)

`#` 기호 뒤의 모든 내용은 주석으로 처리됩니다:

````python
# 이것은 주석입니다
# 주석은 코드 설명을 위한 메모입니다
# Python은 주석을 실행하지 않습니다

print("주석 테스트")  # 이 줄은 실행됩니다
# print("이 줄은 실행되지 않습니다")  # 앞에 #이 있으면 주석
````

**실행 결과:**
```
주석 테스트
```

### 코드 해설

- 1-3줄: 모두 주석이므로 실행되지 않음
- 5줄: `print("주석 테스트")`는 실행됨, `# 이 줄은 실행됩니다`는 주석
- 6줄: 전체가 주석 처리되어 `print("이 줄은 실행되지 않습니다")`도 실행 안 됨

### 💡 주석 작성 팁

**좋은 주석:**
```python
# 사용자의 나이를 입력받아 성인 여부를 판단합니다
age = int(input("나이를 입력하세요: "))
```

**불필요한 주석:**
```python
# age 변수에 18을 할당합니다  # 너무 당연한 설명
age = 18
```

> 📝 **원칙**: 코드만 봐도 이해할 수 있다면 주석이 필요 없습니다. **왜(Why)**와 **무엇을(What)** 설명하되, **어떻게(How)**는 코드로 설명하세요.

## 3. 여러 줄 주석 (Multi-line Comment)

### 여러 줄을 한 번에 주석 처리하기

긴 설명이나 문서를 작성할 때는 **여러 줄 주석**을 사용합니다.

**방법 1: # 여러 번 사용**
```python
# 이것은
# 여러 줄
# 주석입니다
```

**방법 2: 큰따옴표 3개 (""")**

더 편리한 방법은 **큰따옴표 3개**(`"""`)로 여러 줄을 감싸는 것입니다:

````python
"""
이것은 여러 줄 주석입니다.
큰따옴표 3개(\""")로 감싸면
여러 줄을 한번에 주석으로 만들 수 있습니다.

프로그램 설명이나
긴 메모를 남길 때 유용합니다.
"""

print("여러 줄 주석 테스트 완료!")
````

**실행 결과:**
```
여러 줄 주석 테스트 완료!
```

### 코드 해설

- `"""` ~ `"""` 사이의 모든 내용은 실행되지 않습니다
- 줄바꿈도 자유롭게 할 수 있습니다
- 마지막 `print()` 함수만 실행됩니다

### 언제 사용하나요?

**여러 줄 주석은 주로:**
1. **파일 맨 위**: 프로그램 전체 설명
   ```python
   """
   파일명: calculator.py
   작성자: 홍길동
   작성일: 2024-01-15
   설명: 간단한 계산기 프로그램
   """
   ```

2. **함수 설명** (Docstring):
   ```python
   def add(a, b):
       """
       두 숫자를 더하는 함수
       
       매개변수:
           a (int): 첫 번째 숫자
           b (int): 두 번째 숫자
       
       반환값:
           int: 두 숫자의 합
       """
       return a + b
   ```

3. **임시로 코드 비활성화**:
   ```python
   """
   # 나중에 다시 사용할 코드
   user = get_user()
   print(user.name)
   """
   ```

### 🎯 작은따옴표 vs 큰따옴표

작은따옴표 3개(`'''`)도 사용 가능합니다:

```python
'''
이것도 여러 줄 주석입니다
작은따옴표 3개로도 가능합니다
'''
```

**관례**: Python 커뮤니티에서는 **큰따옴표 3개**(`"""`)를 더 많이 사용합니다.

## 4. 간단한 계산 테스트

### Python은 강력한 계산기입니다!

Python을 사용하면 복잡한 계산도 쉽게 할 수 있습니다. `print()` 함수 안에 계산식을 넣으면 **결과값**이 출력됩니다.

### 기본 연산자

| 연산자 | 의미 | 예시 |
|--------|------|------|
| `+` | 더하기 | `10 + 5` → `15` |
| `-` | 빼기 | `20 - 8` → `12` |
| `*` | 곱하기 | `3 * 4` → `12` |
| `/` | 나누기 | `15 / 3` → `5.0` |

### 실행해보기

````python
# Python으로 계산도 할 수 있습니다
print(10 + 5)  # 출력: 15
print(20 - 8)  # 출력: 12
print(3 * 4)   # 출력: 12
print(15 / 3)  # 출력: 5.0
````

**실행 결과:**
```
15
12
12
5.0
```

### 코드 해설

- `print(10 + 5)`: 10과 5를 더한 **결과인 15**를 출력
- `print(20 - 8)`: 20에서 8을 뺀 **결과인 12**를 출력
- `print(3 * 4)`: 3과 4를 곱한 **결과인 12**를 출력
- `print(15 / 3)`: 15를 3으로 나눈 **결과인 5.0**을 출력

### 🤔 왜 15 / 3 = 5.0 인가요?

나누기 연산자(`/`)는 항상 **소수점이 있는 숫자(float)**를 반환합니다.
- `15 / 3` → `5.0` (float)
- `10 / 2` → `5.0` (float)
- `7 / 2` → `3.5` (float)

> 💡 **Tip**: 정수 나누기를 하고 싶다면 `//` 연산자를 사용하세요!
> ```python
> print(15 // 3)  # 출력: 5 (정수)
> print(7 // 2)   # 출력: 3 (소수점 버림)
> ```

### 💡 더 많은 연산자들

다음 교시부터 더 자세히 배우겠지만, 미리 맛보기로 소개합니다:

```python
print(7 // 2)   # 몫: 3
print(7 % 2)    # 나머지: 1
print(2 ** 3)   # 거듭제곱: 8 (2의 3승)
```

**실행 결과:**
```
3
1
8
```
