## 📖 f-string이란 무엇인가요?

### 문자열 포매팅, 왜 필요할까요?

프로그램을 만들다 보면 **변수의 값을 문자열에 넣어서 출력**해야 하는 경우가 정말 많습니다.

예를 들어볼까요? 쇼핑몰에서 사용자에게 이런 메시지를 보여주고 싶다고 생각해봅시다:

> "안녕하세요, 홍길동님! 주문하신 노트북 1,200,000원이 배송 준비 중입니다."

이 문장에서 **변하는 부분**이 보이시나요?
- 사용자 이름: "홍길동" → 변수로 저장
- 상품명: "노트북" → 변수로 저장
- 가격: "1,200,000" → 변수로 저장

**옛날 방식**으로 하면 이렇게 해야 했습니다:

```python
name = "홍길동"
product = "노트북"
price = 1200000

# 방법 1: 쉼표로 연결 (띄어쓰기 조절이 어려움)
print("안녕하세요,", name, "님! 주문하신", product, price, "원이...")

# 방법 2: + 연산자 (타입 변환 필요, 복잡함)
print("안녕하세요, " + name + "님! 주문하신 " + product + " " + str(price) + "원이...")

# 방법 3: format() 메서드 (좀 나아졌지만 여전히 번거로움)
print("안녕하세요, {}님! 주문하신 {} {:,}원이...".format(name, product, price))
```

😫 복잡하죠? 특히 변수가 많아지면 헷갈리기 쉽습니다!

### ✨ f-string의 등장!

**f-string**은 Python 3.6부터 등장한 **가장 현대적이고 편리한** 문자열 포매팅 방법입니다.

```python
name = "홍길동"
product = "노트북"
price = 1200000

# f-string: 변수를 직접 문자열 안에!
print(f"안녕하세요, {name}님! 주문하신 {product} {price:,}원이 배송 준비 중입니다.")
```

**장점**:
1. 👀 **가독성 최고**: 문자열을 읽는 그대로 이해할 수 있습니다
2. ⚡ **간결함**: 타입 변환 걱정 없이 바로 사용
3. 💪 **강력함**: 계산식, 함수 호출도 바로 가능
4. 🎯 **직관적**: 어디에 무엇이 들어가는지 한눈에 보임

### FastAPI에서는?

FastAPI로 API를 만들 때, **에러 메시지**, **로그 기록**, **응답 메시지** 등을 만들 때 f-string을 정말 많이 사용합니다!

```python
# 실무 예시: 사용자에게 에러 메시지 반환
user_id = 123
raise HTTPException(
    status_code=404,
    detail=f"사용자 ID {user_id}를 찾을 수 없습니다."  # f-string 사용!
)

# 로그 기록
logger.info(f"사용자 {username}이(가) {timestamp}에 로그인했습니다.")

# API 응답
return {"message": f"환영합니다, {user.name}님!"}
```

### 기본 문법

f-string은 정말 간단합니다:

1. 문자열 앞에 **`f`** 붙이기: `f"..."`
2. 변수를 넣을 곳에 **중괄호** `{}` 사용
3. 중괄호 안에 **변수명** 또는 **표현식** 작성

```python
name = "Alice"
age = 25

# f 붙이고, {}로 변수 감싸기
print(f"이름: {name}, 나이: {age}")
```

> ⚠️ **주의**: `f`를 빼먹으면 그냥 일반 문자열로 인식됩니다!
>
> ```python
> # f 없으면 → {name}이 그대로 출력됨
> print("이름: {name}")  # 출력: 이름: {name}
> 
> # f 있어야 → 변수 값으로 대체됨
> print(f"이름: {name}")  # 출력: 이름: Alice
> ```

자, 이제 f-string의 다양한 기능을 하나씩 살펴볼까요? 🚀

## 1. f-string 기본 - 변수 삽입하기

### 가장 기본적인 사용법

f-string의 핵심은 **변수를 중괄호 `{}`로 감싸서** 문자열 안에 직접 넣는 것입니다.

**문법**: `f"텍스트 {변수명} 텍스트"`

Python이 실행될 때, `{변수명}` 부분이 **변수의 실제 값**으로 자동 치환됩니다.

````python
# 변수를 문자열에 삽입할 때 f-string 사용
name = "홍길동"
age = 25

# f-string: f"..." 형식으로 작성, {변수}로 삽입
message = f"안녕하세요, {name}님!"
print(message)  # 출력: 안녕하세요, 홍길동님!

# 여러 변수를 함께 사용
print(f"이름: {name}, 나이: {age}세")  # 출력: 이름: 홍길동, 나이: 25세
````

**실행 결과**:
```
안녕하세요, 홍길동님!
이름: 홍길동, 나이: 25세
```

### 💡 핵심 포인트

- `{name}` → 변수 `name`의 값인 `"홍길동"`으로 치환됨
- `{age}` → 변수 `age`의 값인 `25`로 치환됨
- **타입 자동 변환**: `age`는 정수(int)지만 문자열에 들어갈 때 자동으로 문자열로 변환됩니다!
- 하나의 f-string에 **여러 변수**를 자유롭게 사용 가능

### 기존 방식과 비교해볼까요?

````python
# 기존 방식들 (참고용)
name = "홍길동"
age = 25

# 방법 1: 쉼표 사용 (띄어쓰기 조절이 어려움)
print("이름:", name, ", 나이:", age, "세")  # 쉼표 때문에 공백이 이상함

# 방법 2: + 연산자 (타입 변환 필요, 복잡함)
print("이름: " + name + ", 나이: " + str(age) + "세")  # str(age) 필요!

# 방법 3: f-string (가장 깔끔!)
print(f"이름: {name}, 나이: {age}세")  # 타입 변환 자동, 간결함
````

**실행 결과**:
```
이름: 홍길동 , 나이: 25 세
이름: 홍길동, 나이: 25세
이름: 홍길동, 나이: 25세
```

첫 번째 줄을 보세요! 쉼표 때문에 공백이 어색하죠? 두 번째는 `str(age)`를 써야 해서 번거롭고요.

f-string이 왜 좋은지 이제 확실히 느껴지시나요? 😊

## 2. 표현식 포함하기

### f-string의 진정한 파워!

f-string의 중괄호 `{}` 안에는 **변수**만 들어갈 수 있는 게 아닙니다!

**모든 Python 표현식**을 넣을 수 있습니다:
- ✅ 계산식: `{x + y}`
- ✅ 함수 호출: `{len(text)}`
- ✅ 메서드 호출: `{name.upper()}`
- ✅ 조건식: `{"성인" if age >= 18 else "미성년자"}`

이게 바로 f-string이 **강력한** 이유입니다! 💪

````python
# f-string 안에서 계산식도 사용 가능
x = 10
y = 20
print(f"{x} + {y} = {x + y}")  # 출력: 10 + 20 = 30

# 문자열 메서드도 직접 사용 가능
name = "alice"
print(f"대문자: {name.upper()}")  # 출력: 대문자: ALICE

# 복잡한 표현식
price = 15000
quantity = 3
print(f"총 금액: {price * quantity}원")  # 출력: 총 금액: 45000원
````

**실행 결과**:
```
10 + 20 = 30
대문자: ALICE
총 금액: 45000원
```

### 어떻게 작동하나요?

Python은 f-string을 실행할 때:
1. 중괄호 `{}` 안의 표현식을 **먼저 계산**합니다
2. 계산 결과를 **문자열로 변환**합니다
3. 그 자리에 **치환**합니다

예를 들어:
```python
f"{x + y}"  # 1. x + y 계산 → 30
            # 2. 30을 "30"으로 변환
            # 3. 결과: "30"

f"{name.upper()}"  # 1. name.upper() 실행 → "ALICE"
                   # 2. 이미 문자열
                   # 3. 결과: "ALICE"
```

### 💡 실전 팁

표현식이 너무 복잡하면 **가독성**이 떨어질 수 있습니다.

```python
# ❌ 너무 복잡 (한눈에 안 들어옴)
print(f"결과: {sum([i**2 for i in range(10) if i % 2 == 0])}")

# ✅ 변수로 분리 (가독성 좋음)
even_squares_sum = sum([i**2 for i in range(10) if i % 2 == 0])
print(f"결과: {even_squares_sum}")
```

**원칙**: 중괄호 안의 표현식은 **한 줄로 읽기 쉬운 정도**까지만!

## 3. 변수와 텍스트 혼합 - 실전 예제

### 실무에서 가장 많이 쓰는 패턴

실제 프로그램에서는 **여러 변수와 텍스트**를 조합해서 출력하는 경우가 정말 많습니다.

특히 **쇼핑몰**, **재고 관리**, **보고서** 같은 곳에서 자주 사용하죠.

한번 쇼핑몰 상품 정보를 출력해볼까요?

````python
# 실전 예제: 상품 정보 출력
product = "노트북"
price = 1200000
stock = 5

print(f"상품명: {product}")
print(f"가격: {price:,}원")  # :, → 천단위 콤마 추가!
print(f"재고: {stock}개")
print(f"{product}의 총 재고 가치: {price * stock:,}원")
# 출력: 노트북의 총 재고 가치: 6,000,000원
````

**실행 결과**:
```
상품명: 노트북
가격: 1,200,000원
재고: 5개
노트북의 총 재고 가치: 6,000,000원
```

### ✨ 새로운 포매팅: 천단위 콤마 `{변수:,}`

여기서 처음 등장한 **`:,`**를 주목하세요!

```python
{price:,}  # 1200000 → 1,200,000 (천단위 콤마)
```

**문법**: `{변수:포맷지정자}`
- **콜론(`:`)** 뒤에 포맷 지정자를 붙입니다
- **`,`**: 천단위마다 콤마를 추가

### 왜 천단위 콤마가 중요한가요?

```python
# 콤마 없으면? (읽기 어려움)
print(f"가격: {price}원")  # 출력: 가격: 1200000원

# 콤마 있으면? (한눈에 파악!)
print(f"가격: {price:,}원")  # 출력: 가격: 1,200,000원
```

1200000 vs 1,200,000 - 어느 쪽이 "120만 원"이라고 빠르게 읽히나요? 당연히 콤마가 있는 쪽이죠!

### 💼 실무 활용

FastAPI에서 금액을 반환할 때:

```python
# API 응답 메시지

    "formatted": f"{total_sales:,}"
```

## 4. 소수점 포매팅

### 소수를 깔끔하게 표시하기

계산 결과에 **소수점**이 나오는 경우가 많죠? 특히 **평균**, **비율**, **통계** 계산할 때요.

문제는, Python의 나눗셈 결과가 이렇게 나온다는 거예요:

```python
average = (85 + 92 + 88) / 3
print(average)  # 출력: 88.33333333333333
```

😱 소수점이 너무 길어요! 우리가 원하는 건 **88.33**인데 말이죠.

이럴 때 **소수점 포매팅**을 사용합니다!

````python
# 소수점 자리수 지정: {변수:.2f} → 소수점 2자리
pi = 3.141592653589793
print(f"원주율: {pi:.2f}")  # 출력: 원주율: 3.14
print(f"원주율: {pi:.4f}")  # 출력: 원주율: 3.1416
````

**실행 결과**:
```
원주율: 3.14
원주율: 3.1416
```

### 📐 문법 분해

`{pi:.2f}`를 하나씩 뜯어볼까요?

- `{pi` : 변수 이름
- `:` : "포맷 지정 시작"
- `.2` : 소수점 **2자리**까지
- `f` : **f**loat (부동소수점) 표현

**패턴**: `{변수:.자릿수f}`

### 자동 반올림!

소수점 포매팅은 자동으로 **반올림**됩니다:

```python
value = 3.14159

{value:.2f}  # 3.14159 → 3.14 (반올림)
{value:.3f}  # 3.14159 → 3.142 (반올림)
{value:.1f}  # 3.14159 → 3.1 (반올림)
```

### 실전 예제: 성적 평균 계산

````python
# 실전 예제: 평균 계산
korean = 85
english = 92
math = 88
average = (korean + english + math) / 3

print(f"국어: {korean}점")
print(f"영어: {english}점")
print(f"수학: {math}점")
print(f"평균: {average:.2f}점")  # 출력: 평균: 88.33점
````

**실행 결과**:
```
국어: 85점
영어: 92점
수학: 88점
평균: 88.33점
```

### 💡 다양한 소수점 자릿수

필요에 따라 자릿수를 조절할 수 있습니다:

```python
value = 12.3456789

# 정수만 (소수점 0자리)
f"{value:.0f}"  # → "12"

# 소수점 1자리
f"{value:.1f}"  # → "12.3"

# 소수점 2자리 (가장 많이 씀!)
f"{value:.2f}"  # → "12.35"

# 소수점 5자리
f"{value:.5f}"  # → "12.34568"
```

### ⚠️ 주의사항

1. **정수에는 `.f` 사용 불가**
   ```python
   num = 10  # 정수
   f"{num:.2f}"  # ✅ 가능: "10.00"
   ```

2. **문자열에는 사용 불가**
   ```python
   text = "hello"
   f"{text:.2f}"  # ❌ 에러! 문자열은 .f 못 씀
   ```

### 🎯 실무에서는?

## 5. 실전 종합 예제: 쇼핑몰 영수증

### 지금까지 배운 것을 모두 활용해봅시다!

자, 이제 **진짜 실무 수준**의 예제를 만들어볼 시간입니다! 🎉

쇼핑몰 영수증을 만들어보겠습니다. 이 예제에는 지금까지 배운 **모든 기술**이 들어갑니다:
- ✅ 변수 삽입
- ✅ 표현식 계산
- ✅ 천단위 콤마
- ✅ 소수점 포매팅
- ✅ 정렬 (보너스!)

천천히 따라오세요! 각 부분을 자세히 설명하겠습니다. 😊

````python
# 상품 정보
item1 = "티셔츠"
price1 = 29000
qty1 = 2

item2 = "청바지"
price2 = 59000
qty2 = 1

# 소계 계산
subtotal1 = price1 * qty1  # 티셔츠: 29,000 × 2 = 58,000
subtotal2 = price2 * qty2  # 청바지: 59,000 × 1 = 59,000
total = subtotal1 + subtotal2  # 총합: 117,000

# 할인 적용 (10%)
discount_rate = 0.1  # 10% = 0.1
discount_amount = total * discount_rate  # 할인액: 11,700
final_price = total - discount_amount  # 최종: 105,300

# 영수증 출력
print("=" * 40)
print("       쇼핑몰 영수증")
print("=" * 40)
print(f"{item1:10s} {price1:>8,}원 x {qty1}개 = {subtotal1:>10,}원")
print(f"{item2:10s} {price2:>8,}원 x {qty2}개 = {subtotal2:>10,}원")
print("-" * 40)
print(f"{'소계':20s} {total:>10,}원")
print(f"{'할인 (10%)':20s} -{discount_amount:>9,.0f}원")
print("=" * 40)
print(f"{'최종 금액':20s} {final_price:>10,.0f}원")
print("=" * 40)
````

**실행 결과**:
```
========================================
       쇼핑몰 영수증
========================================
티셔츠        29,000원 x 2개 =     58,000원
청바지        59,000원 x 1개 =     59,000원
----------------------------------------
소계                          117,000원
할인 (10%)                    -11,700원
========================================
최종 금액                     105,300원
========================================
```

### 🔍 코드 분석 - 새로운 포매팅 기법

이 예제에는 **고급 포매팅**이 사용되었습니다. 하나씩 뜯어볼까요?

#### 1. 문자열 정렬과 공간 확보

```python
{item1:10s}
```

- `:10s` = 문자열(**s**tring)을 **10자리** 공간에 배치
- "티셔츠"는 3글자 → 나머지 7칸은 공백으로 채움
- 왼쪽 정렬이 기본

```python
"티셔츠      "  # 10자리 (티셔츠 + 공백 7칸)
```

#### 2. 숫자 오른쪽 정렬

```python
{price1:>8,}
```

- `>` = **오른쪽** 정렬
- `8` = **8자리** 공간
- `,` = **천단위** 콤마

```python
29000 → " 29,000"  # 8자리 (공백 2 + "29,000")
```

#### 3. 여러 포맷 조합

```python
{discount_amount:>9,.0f}
```

- `>9` = 오른쪽 정렬, 9자리
- `,` = 천단위 콤마
- `.0f` = 소수점 0자리 (정수로 표시)

```python
11700.0 → "  11,700"  # 9자리
```

### 📊 정렬의 중요성

정렬이 없으면 이렇게 됩니다:

```
티셔츠 29,000원 x 2개 = 58,000원
청바지 59,000원 x 1개 = 59,000원
```

정렬이 있으면:

```
티셔츠        29,000원 x 2개 =     58,000원
청바지        59,000원 x 1개 =     59,000원
```

훨씬 깔끔하고 읽기 쉽죠! 📈

## 📌 핵심 정리

이번 시간에 배운 내용을 정리해봅시다!

### 1. f-string 기본 문법

```python
name = "홍길동"
age = 25

# f를 붙이고, {}로 변수 감싸기
print(f"이름: {name}, 나이: {age}세")
```

**핵심**: `f"...{변수}..."` - 문자열 앞에 `f`, 변수는 중괄호로!

### 2. 표현식 사용

```python
# 계산식
f"{x + y}"

# 메서드 호출
f"{name.upper()}"

# 복잡한 표현식
f"{price * quantity}"
```

**핵심**: 중괄호 안에는 **모든 Python 표현식** 가능!

### 3. 천단위 콤마

```python
price = 1200000
print(f"{price:,}원")  # 출력: 1,200,000원
```

**핵심**: `{변수:,}` - 콜론 뒤에 콤마!

### 4. 소수점 포매팅

```python
average = 88.333333
print(f"{average:.2f}")  # 출력: 88.33
```

**핵심**: `{변수:.자릿수f}` - 소수점 자릿수 지정!

### 5. 조합 사용

```python
# 천단위 콤마 + 소수점
f"{value:,.2f}"

# 정렬 + 천단위
f"{price:>10,}"

# 정렬 + 콤마 + 소수점
f"{amount:>12,.2f}"
```

**핵심**: 여러 포맷을 `:` 뒤에 조합 가능!

## 💡 기억할 점

### ✅ f-string을 사용해야 하는 이유

1. **가독성**: 코드를 읽는 그대로 이해 가능
2. **간결성**: 타입 변환 자동, 코드 짧음
3. **강력함**: 표현식, 메서드 호출 모두 가능
4. **현대적**: Python 3.6+ 표준, FastAPI에서 필수

### ⚠️ 주의사항

1. **`f` 빼먹지 않기**: `f"..."` - f 없으면 일반 문자열!
2. **중괄호 정확히**: `{변수}` - 띄어쓰기 주의
3. **복잡한 표현식은 변수로**: 가독성 우선
4. **포맷 문법 외우지 말고**: 필요할 때 찾아보기

### 🚀 실무 팁

- **금액**: `{금액:,}` - 항상 천단위 콤마
- **비율/평균**: `{값:.2f}` - 보통 소수점 2자리
- **정렬**: 영수증, 보고서에서 `{값:>10,}` 많이 사용
- **로그**: `f"[{시간}] {사용자}: {메시지}"` 패턴 자주 사용
