## 🟡 응용 Problem: 쇼핑몰 영수증 (15분)

### 문제 설명

온라인 쇼핑몰에서 상품 구매 후 **영수증**을 출력하세요!

### 사용할 개념

- ✅ `input()`: 상품 정보 입력
- ✅ `int()`: 가격/수량 변환
- ✅ 사칙연산: 소계, 할인, 최종 금액 계산
- ✅ `f-string` + `:.0f`: 소수점 제거 포매팅

### 계산 과정

1. 소계 = 가격 × 수량
2. 할인 금액 = 소계 × 0.1 (10% 할인)
3. 최종 금액 = 소계 - 할인 금액

### 예상 출력

```
========================================
           쇼핑몰 영수증
========================================
상품명: 무선 이어폰
단가: 50000원
수량: 2개
소계: 100000원
할인 (10%): -10000원
========================================
최종 결제 금액: 90000원
========================================
```

## 🔴 도전 Problem: 데이터 처리 프로그램 (10분)

### 문제 설명

CSV 형식의 사용자 데이터를 입력받아 **정제**하고 출력하세요!

### 사용할 개념 (종합!)

- ✅ `split()`: 데이터 분리
- ✅ `strip()`: 공백 제거
- ✅ `lower()`, `upper()`: 대소문자 변환
- ✅ 문자열 슬라이싱: 첫 글자 추출
- ✅ `int()`: 형변환
- ✅ `f-string`: 포매팅

### 입력 예시

```
hong gildong,25,HONG@EXAMPLE.COM
```

### 처리 과정

1. 쉼표로 분리: `["hong gildong", "25", "HONG@EXAMPLE.COM"]`
2. 이름 정제: "Hong Gildong" (각 단어 첫 글자 대문자)
3. 나이 정제: 25 (숫자로 변환)
4. 이메일 정제: "hong@example.com" (소문자)

### 예상 출력

```
========================================
        데이터 처리 결과
========================================
원본: hong gildong,25,HONG@EXAMPLE.COM
----------------------------------------
이름: Hong Gildong
나이: 25세
이메일: hong@example.com
========================================
```
