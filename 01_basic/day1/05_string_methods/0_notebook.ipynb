## 📖 메서드(Method)란 무엇인가?

### 메서드의 개념

**메서드(Method)**는 특정 자료형(타입)에 **딸려있는 함수**입니다. 

지금까지 우리는 `print()`, `len()`, `type()` 같은 **함수**를 사용했습니다:

```python
len("hello")  # 함수: 괄호 안에 값을 넣음
```

하지만 **메서드**는 조금 다릅니다. 점(`.`)을 사용해서 호출합니다:

```python
"hello".upper()  # 메서드: 문자열 뒤에 . 찍고 호출
```

### 왜 메서드가 필요할까요?

각 자료형(문자열, 리스트, 딕셔너리 등)은 **자신만의 특별한 기능**이 필요합니다.

- 문자열은 **대소문자 변환**, **분리**, **교체** 같은 기능이 필요
- 리스트는 **추가**, **삭제**, **정렬** 같은 기능이 필요
- 딕셔너리는 **키 조회**, **값 추가** 같은 기능이 필요

이런 기능들을 각 자료형에 **메서드**로 붙여놓은 것입니다! 🎯

### FastAPI에서는?

FastAPI로 웹 API를 만들 때, **사용자 입력 데이터를 정제**하는 일이 정말 많습니다:

```python
# 사용자가 이메일을 대문자로 입력했다면?
user_email = "ADMIN@COMPANY.COM"

# 소문자로 변환해서 저장 (이메일은 대소문자 구분 없으니까)
user_email = user_email.lower()  # "admin@company.com"
```

```python
# CSV 형식으로 여러 태그가 들어왔다면?
tags = "python,fastapi,backend,api"

# split()으로 분리해서 리스트로 만들기
tag_list = tags.split(",")  # ['python', 'fastapi', 'backend', 'api']
```

실무에서 **메서드를 엄청 많이 사용**합니다! 💪

### 문자열의 불변성 복습

> ⚠️ **중요**: 문자열 메서드는 **원본을 변경하지 않습니다**!

문자열은 **불변(immutable)**이기 때문에, 메서드를 사용하면 **새로운 문자열**을 반환합니다:

```python
name = "python"
upper_name = name.upper()  # 새 문자열 "PYTHON" 생성

print(name)        # "python" (원본 그대로)
print(upper_name)  # "PYTHON" (새로운 문자열)
```

따라서 **메서드의 결과를 변수에 저장**해야 합니다!

## 1. upper()와 lower() - 대소문자 변환

### 언제 사용하나요?

- **이메일 주소**: 대소문자 구분 없이 저장 (`Admin@site.com` → `admin@site.com`)
- **사용자 이름**: 일관된 형식으로 저장
- **검색어**: 대소문자 구분 없이 검색 (`Python` vs `python`)
- **비교 연산**: 대소문자 상관없이 비교할 때

### 메서드는 점(.)으로 호출합니다

문자열 뒤에 `.메서드이름()`을 붙여서 사용합니다.

````python
# 메서드는 .을 사용해서 호출합니다
name = "Python Programming"

# 모두 대문자로 변환: upper()
# 원본 문자열의 모든 알파벳을 대문자로 바꾼 "새로운 문자열"을 반환
upper_name = name.upper()
print(upper_name)  # 출력: PYTHON PROGRAMMING

# 모두 소문자로 변환: lower()
# 원본 문자열의 모든 알파벳을 소문자로 바꾼 "새로운 문자열"을 반환
lower_name = name.lower()
print(lower_name)  # 출력: python programming

# 원본은 변경되지 않습니다 (불변성)
print(name)  # 출력: Python Programming (그대로!)
````

**실행 결과:**
```
PYTHON PROGRAMMING
python programming
Python Programming
```

**결과 해석:**
- `upper()`는 모든 문자를 대문자로 변환한 **새 문자열** 반환
- `lower()`는 모든 문자를 소문자로 변환한 **새 문자열** 반환
- 원본 `name`은 여전히 `"Python Programming"` (불변성!)

### 💡 보충 설명

**자주 하는 실수:**
```python
# ❌ 잘못된 사용 (결과를 저장 안 함)
email = "ADMIN@SITE.COM"
email.lower()  # 이렇게만 하면 결과가 사라짐!
print(email)   # 여전히 "ADMIN@SITE.COM"

# ✅ 올바른 사용 (결과를 변수에 저장)
email = "ADMIN@SITE.COM"
email = email.lower()  # 결과를 다시 email에 저장
print(email)   # "admin@site.com"
```

**실전 팁:**
- 영어가 아닌 문자(한글, 숫자, 특수문자)는 그대로 유지됩니다
- `upper()`와 `lower()`는 괄호 `()`가 비어있습니다 (매개변수 없음)

## 2. strip() - 공백 제거

### 왜 공백 제거가 필요한가요?

사용자가 입력한 데이터에는 **의도치 않은 공백**이 정말 많습니다:

- 실수로 스페이스바를 여러 번 누름
- 복사-붙여넣기 할 때 공백이 포함됨
- 엔터키를 눌러서 줄바꿈이 들어감

이런 공백을 제거하지 않으면:
```python
email1 = "user@example.com"
email2 = "  user@example.com  "  # 공백 포함

print(email1 == email2)  # False! (다른 문자열로 인식)
```

같은 이메일인데 로그인이 안 되는 참사가! 😱

### strip()은 양쪽 공백을 제거합니다

````python
# 사용자 입력에는 공백이 많이 들어옵니다
# (실제로 사용자가 "   hello world   "를 입력했다고 가정)
user_input = "   hello world   "
print(user_input)  # 출력:    hello world   (공백이 보임)

# 양쪽 공백 제거: strip()
# 문자열 왼쪽과 오른쪽의 공백(스페이스, 탭, 엔터 등)을 모두 제거
# 중간에 있는 공백은 그대로 유지됩니다!
cleaned = user_input.strip()
print(cleaned)  # 출력: hello world (양쪽 공백 사라짐)

# 실전 예제: 이메일 주소 정제
email = "  user@example.com  "
clean_email = email.strip()
print(clean_email)  # 출력: user@example.com
````

**실행 결과:**
```
   hello world   
hello world
user@example.com
```

**결과 해석:**
- 원본에는 앞뒤로 공백이 여러 개 있었음
- `strip()`은 양쪽 공백만 제거 (중간 공백 "hello world"는 유지)
- 깔끔하게 정제된 문자열 완성! ✨

### 💡 보충 설명

**strip()이 제거하는 것들:**
- 스페이스(` `)
- 탭(`\t`)
- 엔터/줄바꿈(`\n`)
- 기타 공백 문자들

**주의사항:**
```python
# strip()은 양쪽만 제거합니다
text = "  hello   world  "
print(text.strip())  # "hello   world" (중간 공백 3개는 유지)
```

**실무 팁:**
- 사용자 입력을 받을 때는 **항상 strip() 사용**을 습관화하세요
- FastAPI에서 이메일, 비밀번호 같은 중요 정보는 반드시 strip() 처리!

**왼쪽/오른쪽만 제거하는 메서드도 있습니다 (참고):**
- `lstrip()`: 왼쪽만 제거 (left strip)
- `rstrip()`: 오른쪽만 제거 (right strip)

→ 하지만 대부분 `strip()`만 사용합니다!

## 3. split() - 문자열 분리

### 왜 문자열 분리가 필요한가요?

실무에서는 **구분자로 구분된 데이터**를 정말 많이 다룹니다:

- **CSV 파일**: `"사과,바나나,딸기"` → 각각 분리
- **쿼리 파라미터**: `"tag1,tag2,tag3"` → 리스트로 변환
- **파일 경로**: `"C:\Users\Documents\file.txt"` → 폴더별로 분리
- **문장**: `"Hello Python World"` → 단어별로 분리

### split()은 문자열을 "리스트"로 변환합니다

> 📌 **참고**: 리스트(List)는 다음 시간에 자세히 배웁니다. 지금은 "여러 값을 저장하는 자료형"이라고만 알아두세요!

### 기준 문자를 지정해서 분리합니다

````python
# 쉼표로 구분된 데이터 (CSV 형식)
# CSV = Comma-Separated Values (쉼표로 구분된 값)
data = "사과,바나나,딸기"

# 쉼표를 기준으로 분리 (리스트로 반환)
# split(",")는 쉼표(,)를 찾아서 그 부분을 잘라냅니다
# 결과는 "리스트"라는 자료형으로 반환됩니다 (대괄호 []로 표시)
parts = data.split(",")
print(parts)  # 출력: ['사과', '바나나', '딸기']

# 공백을 기준으로 분리
sentence = "Hello Python World"
words = sentence.split(" ")  # 공백(" ")을 기준으로 분리
print(words)  # 출력: ['Hello', 'Python', 'World']

# split() 괄호 안이 비어있으면 공백 기준
# 공백이 여러 개여도 하나로 처리합니다 (똑똑함!)
words2 = sentence.split()  # 기본값: 모든 공백 문자 기준
print(words2)  # 출력: ['Hello', 'Python', 'World']
````

**실행 결과:**
```
['사과', '바나나', '딸기']
['Hello', 'Python', 'World']
['Hello', 'Python', 'World']
```

**결과 해석:**
- `split(",")`: 쉼표를 기준으로 3개로 분리 → 리스트 3개 요소
- `split(" ")`: 공백을 기준으로 3개로 분리
- `split()`: 기본값은 모든 공백 문자 기준 (가장 많이 사용!)

### 💡 보충 설명

**split()의 두 가지 사용법:**

```python
# 1. 구분자 지정: split("구분자")
"apple-banana-cherry".split("-")  # 하이픈 기준
# ['apple', 'banana', 'cherry']

"2024/01/15".split("/")  # 슬래시 기준
# ['2024', '01', '15']

# 2. 구분자 없음: split() - 모든 공백 기준 (권장!)
"Hello    World".split()  # 공백 여러 개도 OK
# ['Hello', 'World']
```

**FastAPI 실전 예제:**
```python
# 사용자가 쿼리 파라미터로 태그를 보냈다면?
# GET /api/posts?tags=python,fastapi,backend

tags_str = "python,fastapi,backend"  # 문자열로 받음
tags_list = tags_str.split(",")       # 리스트로 변환
# ['python', 'fastapi', 'backend']
```

**주의사항:**
- 구분자가 없으면 분리되지 않습니다:
  ```python
  "hello".split(",")  # ['hello'] (쉼표가 없으니 그대로)
  ```

## 4. replace() - 문자열 교체

### 언제 사용하나요?

- **금지어 필터링**: 욕설 → "***"
- **형식 변경**: 전화번호 `010-1234-5678` → `01012345678`
- **경로 변경**: `C:\Users\` → `C:/Users/`
- **데이터 정제**: 불필요한 문자 제거
- **템플릿 치환**: `"안녕하세요 {이름}님"` → `"안녕하세요 홍길동님"`

### replace(찾을_문자열, 바꿀_문자열)

````python
# 특정 문자열을 다른 문자열로 바꿉니다
message = "Hello World"

# World를 Python으로 교체
# replace("찾을 문자열", "바꿀 문자열")
# 원본에서 "World"를 찾아서 "Python"으로 바꾼 새 문자열 반환
new_message = message.replace("World", "Python")
print(new_message)  # 출력: Hello Python

# 공백을 빈 문자열로 교체 (= 공백 제거)
# 모든 공백(" ")을 찾아서 빈 문자열("")로 바꿈
no_space = message.replace(" ", "")
print(no_space)  # 출력: HelloWorld

# 실전 예제: 전화번호 형식 변경
phone = "010-1234-5678"
# 하이픈(-)을 모두 찾아서 빈 문자열("")로 교체 (= 하이픈 제거)
phone_no_dash = phone.replace("-", "")
print(phone_no_dash)  # 출력: 01012345678
````

**실행 결과:**
```
Hello Python
HelloWorld
01012345678
```

**결과 해석:**
- `replace("World", "Python")`: "World"를 찾아서 "Python"으로 교체
- `replace(" ", "")`: 공백을 빈 문자열로 교체 = 공백 삭제
- `replace("-", "")`: 하이픈을 모두 제거

### 💡 보충 설명

**replace()의 특징:**

1. **모든 occurrence를 교체합니다:**
   ```python
   text = "apple apple apple"
   result = text.replace("apple", "banana")
   print(result)  # "banana banana banana" (3개 모두 교체)
   ```

2. **찾는 문자열이 없으면 원본 그대로:**
   ```python
   "hello".replace("x", "y")  # "hello" (x가 없으니 그대로)
   ```

3. **대소문자를 구분합니다:**
   ```python
   "Hello".replace("hello", "Hi")  # "Hello" (대소문자 다름!)
   "Hello".replace("Hello", "Hi")  # "Hi" (정확히 일치)
   ```

**실무 팁:**

```python
# 금지어 필터링 (커뮤니티 사이트)
comment = "이건 정말 짱이에요!"
filtered = comment.replace("짱", "좋은")
# "이건 정말 좋은이에요!"

# 파일 경로 변환 (Windows → Unix)
path = "C:\\Users\\Documents\\file.txt"
unix_path = path.replace("\\", "/")
# "C:/Users/Documents/file.txt"
```

**주의사항:**
- 교체는 **순서대로** 진행됩니다
- 교체 후 다시 교체되지 않습니다:
  ```python
  "AAA".replace("AA", "B")  # "BA" (첫 AA만 교체)
  ```

## 5. 실전 예제 - 복합 활용 (메서드 체이닝)

### 여러 메서드를 연결해서 사용하기

실무에서는 **여러 가지 정제 작업**을 한꺼번에 해야 할 때가 많습니다:

1. 공백 제거
2. 소문자 변환
3. 특수문자 제거
4. ...

이럴 때 **메서드 체이닝(Method Chaining)**을 사용합니다!

### 메서드 체이닝이란?

메서드를 `.`으로 계속 연결해서 한 줄로 작성하는 것:

```python
result = text.strip().lower().replace(" ", "_")
#            1단계    2단계     3단계
```

왼쪽부터 순서대로 실행됩니다! 🔗

````python
# 사용자 데이터 정제
# 실제로 사용자가 이렇게 입력했다고 가정
user_data = "  HONG GIL DONG  "

# 방법 1: 단계별로 처리
# 1단계: 공백 제거
step1 = user_data.strip()
print(step1)  # 출력: HONG GIL DONG

# 2단계: 소문자 변환
step2 = step1.lower()
print(step2)  # 출력: hong gil dong

# 방법 2: 한 줄로 연결 (메서드 체이닝)
# strip()의 결과에 바로 lower()를 적용
# 더 간결하고 깔끔합니다!
result = user_data.strip().lower()
print(result)  # 출력: hong gil dong

print()  # 빈 줄 (가독성)

# CSV 데이터 처리 예제
# 쉼표로 구분된 회원 정보
csv_line = "김철수,30,서울,developer"

# 쉼표로 분리 (리스트로 변환)
data_parts = csv_line.split(",")
print(data_parts)  # 출력: ['김철수', '30', '서울', 'developer']

# 직업을 대문자로 변경하고 싶다면
# 리스트의 4번째 요소(인덱스 3)를 추출
job = data_parts[3]
job_upper = job.upper()
print(job_upper)  # 출력: DEVELOPER
````

**실행 결과:**
```

hong gil dong
hong gil dong

['김철수', '30', '서울', 'developer']
DEVELOPER
```

**결과 해석:**
- 단계별 처리와 메서드 체이닝의 결과는 동일
- 메서드 체이닝이 더 간결하고 프로페셔널함
- CSV 데이터를 split()으로 분리 → 리스트로 변환
- 리스트의 특정 요소에 메서드 적용 가능

## 📌 핵심 정리

이번 시간에 배운 내용:

1. **메서드(Method)**: 자료형에 딸려있는 함수, `.`으로 호출
2. **`upper()`, `lower()`**: 대소문자 변환 (이메일, 검색어 처리)
3. **`strip()`**: 양쪽 공백 제거 (사용자 입력 정제 필수!)
4. **`split()`**: 문자열을 리스트로 분리 (CSV, 태그 처리)
5. **`replace()`**: 문자열 교체 (금지어, 형식 변환)
6. **메서드 체이닝**: 여러 메서드를 `.`으로 연결 (간결한 코드)
