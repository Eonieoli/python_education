## 📖 사용자 입력과 형변환이란?

프로그램을 만들다 보면 **사용자로부터 정보를 입력받아야** 하는 경우가 많습니다.

예를 들어:
- 로그인할 때 아이디와 비밀번호 입력
- 쇼핑몰에서 상품 개수 입력
- 계산기 프로그램에서 숫자 입력

Python에서는 `input()` 함수를 사용하여 사용자로부터 데이터를 입력받을 수 있습니다.

### 하지만 한 가지 중요한 특징이 있습니다!

> **`input()` 함수는 사용자가 무엇을 입력하든 항상 문자열(str)로 반환합니다!**

숫자 25를 입력해도 `"25"`(문자열)로 저장됩니다. 그래서 숫자 계산을 하려면 **형변환(Type Conversion)**이 필요합니다.

### 형변환이란?

**형변환**은 데이터의 타입을 다른 타입으로 바꾸는 것입니다.

- `int()`: 문자열이나 실수를 **정수(integer)**로 변환
- `float()`: 문자열이나 정수를 **실수(float)**로 변환
- `str()`: 숫자를 **문자열(string)**로 변환

### FastAPI에서는?

웹 API를 만들 때도 똑같은 상황이 발생합니다!

- 사용자가 웹 브라우저에서 "25"를 입력하면 → **문자열 "25"**로 서버에 전달됩니다
- 이것을 숫자로 계산하려면 → **형변환**이 필요합니다
- FastAPI는 Pydantic 모델로 **자동 형변환**을 해주지만, 그 원리를 이해하는 것이 중요합니다!

```python
# FastAPI에서 이런 식으로 사용됩니다 (미리보기)
from pydantic import BaseModel

class User(BaseModel):
    name: str      # 문자열
    age: int       # 정수 (자동 형변환!)
    height: float  # 실수 (자동 형변환!)
```

## 1. input() 기본 - 항상 문자열 반환!

### input() 함수는 무엇을 하나요?

`input()` 함수는 프로그램 실행을 멈추고 **사용자가 무언가를 입력할 때까지 기다립니다**. 사용자가 엔터(Enter)를 누르면 입력한 내용을 **문자열로 반환**합니다.

**기본 문법:**
```python
변수 = input("안내 메시지")
```

### 핵심 포인트 🔑

- `input()`의 괄호 안에 있는 문자열은 **사용자에게 보여줄 안내 메시지**입니다
- 사용자가 입력한 값은 **항상 문자열(str) 타입**으로 반환됩니다
- 숫자를 입력해도 문자열로 저장됩니다!

실제로 확인해볼까요?

````python
# 사용자로부터 이름을 입력받습니다
# input() 함수는 프로그램을 멈추고 입력을 기다립니다
name = input("이름을 입력하세요: ")

# 입력받은 이름을 f-string으로 출력합니다
print(f"안녕하세요, {name}님!")  
# 만약 "홍길동"을 입력했다면 출력: 안녕하세요, 홍길동님!
````

**예상 실행 결과:**
```
이름을 입력하세요: 홍길동
안녕하세요, 홍길동님!
```

이름처럼 **문자를 입력받을 때는 문제가 없습니다**. 하지만 **숫자를 입력받을 때는 주의**해야 합니다!

````python
# input()의 특징: 항상 문자열(str)로 반환됩니다!
# 숫자 25를 입력해도 문자열 "25"로 저장됩니다
age_str = input("나이를 입력하세요: ")

# type() 함수로 데이터 타입을 확인해봅시다
print(type(age_str))  
# 출력: <class 'str'> ← 숫자를 입력해도 문자열(str)입니다!

# 입력받은 값을 출력합니다
print(f"입력값: {age_str}")  
# 만약 25를 입력했다면 출력: 입력값: 25
````

**예상 실행 결과:**
```
나이를 입력하세요: 25
<class 'str'>
입력값: 25
```

### 💡 중요한 개념

사용자가 숫자 `25`를 입력했지만, `type(age_str)`의 결과는 `<class 'str'>`입니다.

즉, **`"25"`(문자열)**로 저장되었습니다. **`25`(숫자)**가 아닙니다!

이것이 왜 문제가 될까요? 다음 섹션에서 알아봅시다.

## 2. 형변환의 필요성

### 문제 상황: 문자열로 저장된 숫자

사용자가 "25"를 입력했습니다. 우리는 이 나이에 10을 더하고 싶습니다. (10년 후 나이를 계산)

하지만 `input()`으로 받은 값은 **문자열 "25"**입니다. 문자열끼리 `+` 연산을 하면 어떻게 될까요?

````python
# 숫자 25를 입력받았지만 문자열 "25"로 저장됨
age_str = "25"  # input("나이를 입력하세요: ")와 동일한 상황

# 데이터 타입 확인
print(type(age_str))  
# 출력: <class 'str'> ← 문자열입니다!

# 문자열끼리 + 연산을 하면 어떻게 될까요?
print(age_str + "10")  
# 출력: 2510 ← 연결(concatenation)이 됩니다!
# 이건 우리가 원하는 결과가 아닙니다!
````

**실행 결과:**
```
<class 'str'>
2510
```

### 문제 분석 🔍

- `"25" + "10"` → `"2510"` (문자열 연결)
- 우리가 원한 것: `25 + 10 = 35` (숫자 덧셈)

**문자열끼리 `+` 연산을 하면 연결(concatenation)이 됩니다!**

### 해결 방법: 형변환!

숫자로 계산하려면 **문자열을 정수(int)로 변환**해야 합니다.

````python
# 문자열을 정수로 변환하는 int() 함수
age_str = "25"
age_int = int(age_str)  # "25" (문자열) → 25 (정수)

# 이제 숫자 계산이 가능합니다!
print(age_int + 10)  
# 출력: 35 ← 이게 우리가 원하는 결과입니다!
````

**실행 결과:**
```
35
```

### ✅ 해결!

- `int("25")` → `25` (정수로 변환)
- `25 + 10` → `35` (정확한 계산!)

이것이 바로 **형변환의 필요성**입니다!

## 3. int() - 문자열을 정수로 변환

### int() 함수란?

`int()` 함수는 **다른 타입의 데이터를 정수(integer)로 변환**하는 함수입니다.

**주요 용도:**
- 문자열로 된 숫자를 실제 정수로 변환: `int("100")` → `100`
- 실수를 정수로 변환 (소수점 버림): `int(3.14)` → `3`

### 문자열 → 정수 변환

````python
# 문자열 "100"을 정수 100으로 변환
num_str = "100"  # 문자열
num_int = int(num_str)  # int() 함수로 정수로 변환

# 데이터 타입 확인
print(type(num_int))  
# 출력: <class 'int'> ← 정수가 되었습니다!

# 이제 숫자 계산이 가능합니다
print(num_int + 50)  
# 출력: 150 ← 100 + 50 = 150
````

**실행 결과:**
```
<class 'int'>
150
```

### 실수 → 정수 변환 (소수점 버림)

````python
# 실수를 정수로 변환하면 소수점 이하가 버려집니다
pi = 3.14  # 실수
pi_int = int(pi)  # 소수점 이하 버림

print(pi_int)  
# 출력: 3 ← 소수점 이하 .14가 버려졌습니다!
````

**실행 결과:**
```
3
```

### ⚠️ 주의사항

`int()` 함수는 **소수점 이하를 반올림하지 않고 버립니다!**

- `int(3.9)` → `3` (4가 아닙니다!)
- `int(9.99)` → `9` (10이 아닙니다!)

반올림을 원한다면 `round()` 함수를 사용해야 합니다.

## 4. float() - 문자열을 실수로 변환

### float() 함수란?

`float()` 함수는 **다른 타입의 데이터를 실수(floating point number)로 변환**하는 함수입니다.

**주요 용도:**
- 문자열로 된 소수를 실제 실수로 변환: `float("15.99")` → `15.99`
- 정수를 실수로 변환: `float(10)` → `10.0`

### 언제 float()를 사용하나요?

- **소수점이 있는 숫자**를 다룰 때 (가격, 평균, 비율 등)
- **정확한 계산**이 필요할 때

### 문자열 → 실수 변환

````python
# 문자열 "15.99"를 실수 15.99로 변환
price_str = "15.99"  # 문자열
price_float = float(price_str)  # float() 함수로 실수로 변환

# 데이터 타입 확인
print(type(price_float))  
# 출력: <class 'float'> ← 실수가 되었습니다!

# 이제 소수점 계산이 가능합니다
print(price_float * 2)  
# 출력: 31.98 ← 15.99 × 2 = 31.98
````

**실행 결과:**
```
<class 'float'>
31.98
```

### 정수 → 실수 변환

````python
# 정수를 실수로 변환
count = 10  # 정수
count_float = float(count)  # 실수로 변환

print(count_float)  
# 출력: 10.0 ← 소수점이 추가되었습니다!
````

**실행 결과:**
```
10.0
```

### 💡 실무에서는?

쇼핑몰 API를 만든다고 생각해봅시다:

```python
# 사용자가 입력한 가격 (문자열로 전달됨)
price_str = "15.99"

# 실수로 변환하여 계산
price = float(price_str)
quantity = 3
total = price * quantity  # 15.99 × 3 = 47.97
```

## 5. str() - 숫자를 문자열로 변환

### str() 함수란?

`str()` 함수는 **다른 타입의 데이터를 문자열(string)로 변환**하는 함수입니다.

**주요 용도:**
- 숫자를 문자열로 변환하여 **문자열 연결**을 할 때
- 출력 형식을 맞출 때

### 정수 → 문자열 변환

````python
# 정수 25를 문자열 "25"로 변환
age = 25  # 정수
age_str = str(age)  # str() 함수로 문자열로 변환

# 데이터 타입 확인
print(type(age_str))  
# 출력: <class 'str'> ← 문자열이 되었습니다!

# 이제 문자열 연결(+)이 가능합니다
print("나이: " + age_str + "세")  
# 출력: 나이: 25세
````

**실행 결과:**
```
<class 'str'>
나이: 25세
```

### 💡 하지만 더 좋은 방법이 있습니다!

사실 Python에서는 **f-string**을 사용하면 `str()` 변환 없이도 숫자와 문자열을 함께 출력할 수 있습니다.

````python
# f-string을 사용하면 자동으로 형변환이 됩니다!
age = 25
print(f"나이: {age}세")  
# 출력: 나이: 25세
# str() 변환 없이도 동일한 결과!
````

**실행 결과:**
```
나이: 25세
```

### 그럼 str()은 언제 사용하나요?

- 명시적으로 문자열로 변환이 필요할 때
- 데이터를 파일에 저장할 때
- 특정 형식의 문자열을 만들 때

## 6. 형변환 주의점 - 변환 불가능한 경우

### ⚠️ 형변환은 항상 성공하는 것이 아닙니다!

적절하지 않은 데이터를 형변환하려고 하면 **에러(Error)**가 발생합니다.

### 사례 1: 숫자가 아닌 문자열을 int()로 변환

문자열 `"hello"`는 숫자가 아니므로 정수로 변환할 수 없습니다.

````python
# 주의! 이 코드는 에러를 발생시킵니다
# wrong = int("hello")  # ValueError 발생!

# 에러 메시지:
# ValueError: invalid literal for int() with base 10: 'hello'
# 의미: "hello"는 정수로 변환할 수 없는 값입니다

print("위의 코드는 주석 처리되어 있어서 에러가 발생하지 않습니다.")
````

### 사례 2: 소수점이 있는 문자열을 int()로 직접 변환

문자열 `"3.14"`를 `int()`로 바로 변환할 수 없습니다.

````python
# 주의! 이 코드는 에러를 발생시킵니다
# wrong = int("3.14")  # ValueError 발생!

# 에러 메시지:
# ValueError: invalid literal for int() with base 10: '3.14'
# 의미: "3.14"는 소수점이 있어서 정수로 직접 변환할 수 없습니다

print("위의 코드는 주석 처리되어 있어서 에러가 발생하지 않습니다.")
````

### ✅ 올바른 방법: 2단계 변환

소수점이 있는 문자열을 정수로 만들려면:
1. 먼저 `float()`로 실수로 변환
2. 그 다음 `int()`로 정수로 변환

````python
# 올바른 방법: float()으로 먼저 변환 후 int()
# 1단계: "3.14" (문자열) → 3.14 (실수)
# 2단계: 3.14 (실수) → 3 (정수, 소수점 버림)
correct = int(float("3.14"))

print(correct)  
# 출력: 3
````

**실행 결과:**
```
3
```

### 💡 정리: 형변환 가능 여부

**int() 변환 가능:**
- `int("123")` → `123` ✅
- `int(3.14)` → `3` ✅
- `int(float("3.14"))` → `3` ✅

**int() 변환 불가능:**
- `int("hello")` → ❌ ValueError
- `int("3.14")` → ❌ ValueError (바로 변환 불가)

**float() 변환 가능:**
- `float("3.14")` → `3.14` ✅
- `float("123")` → `123.0` ✅
- `float(10)` → `10.0` ✅

**float() 변환 불가능:**
- `float("hello")` → ❌ ValueError

## 7. 실전 예제: 사용자 나이 입력받아 계산

### 종합 예제: 태어난 연도로 나이 계산하기

지금까지 배운 내용을 모두 활용하여 실전 프로그램을 만들어봅시다!

**프로그램 요구사항:**
1. 사용자로부터 태어난 연도를 입력받기 (`input()`)
2. 문자열을 정수로 변환하기 (`int()`)
3. 현재 연도에서 빼서 나이 계산하기
4. 결과를 보기 좋게 출력하기 (`f-string`)

**핵심 개념:**
- `input()` → 항상 문자열 반환
- `int()` → 문자열을 정수로 변환
- 산술 연산 → 나이 계산
- `f-string` → 결과 출력

````python
# === 나이 계산 프로그램 ===

print("\n=== 나이 계산 프로그램 ===")

# 1단계: 사용자로부터 태어난 연도를 입력받습니다
# input()은 항상 문자열을 반환합니다!
birth_year_str = input("태어난 연도를 입력하세요 (예: 1995): ")

# 2단계: 문자열을 정수로 변환합니다
# "1995" (문자열) → 1995 (정수)
birth_year = int(birth_year_str)

# 3단계: 현재 연도를 설정합니다
current_year = 2026

# 4단계: 나이를 계산합니다
# 현재 연도 - 태어난 연도 = 나이
age = current_year - birth_year

# 5단계: 결과를 출력합니다
print(f"당신은 {age}세입니다.")
````

**예상 실행 결과:**
```
=== 나이 계산 프로그램 ===
태어난 연도를 입력하세요 (예: 1995): 1995
당신은 31세입니다.
```

### 🔍 코드 분석

1. **`input("태어난 연도를 입력하세요...")`**:
   - 사용자가 `1995`를 입력하면
   - `"1995"` (문자열)로 `birth_year_str`에 저장됩니다

2. **`int(birth_year_str)`**:
   - `"1995"` (문자열) → `1995` (정수)로 변환
   - 이제 계산이 가능합니다!

3. **`current_year - birth_year`**:
   - `2026 - 1995 = 31`
   - 정수끼리의 정확한 뺄셈 연산

4. **`f"당신은 {age}세입니다."`**:
   - f-string으로 변수 `age`를 문자열에 포함
   - 자동으로 숫자가 문자열로 변환되어 출력됩니다

### 💡 만약 형변환을 안 했다면?

```python
birth_year_str = "1995"  # input()의 결과
age = 2026 - birth_year_str  # ❌ TypeError 발생!
# 에러: unsupported operand type(s) for -: 'int' and 'str'
# 정수와 문자열은 뺄 수 없습니다!
```

형변환은 **필수**입니다!

## 📌 핵심 정리

이번 시간에 배운 내용:

1. **`input()` 함수**: 사용자로부터 입력을 받지만, **항상 문자열(str)로 반환**합니다
2. **`int()` 함수**: 문자열이나 실수를 정수로 변환 (소수점 버림)
3. **`float()` 함수**: 문자열이나 정수를 실수로 변환
4. **`str()` 함수**: 숫자를 문자열로 변환
5. **형변환 주의점**: 변환할 수 없는 값은 에러를 발생시킴

### 실무 적용 🚀

**FastAPI에서 이렇게 사용됩니다:**

```python
from pydantic import BaseModel

class User(BaseModel):
    name: str      # 문자열
    age: int       # 정수 (문자열로 들어와도 자동 변환!)
    height: float  # 실수 (문자열로 들어와도 자동 변환!)

# 사용자가 웹에서 {"name": "Alice", "age": "25", "height": "165.5"} 전송
# Pydantic이 자동으로 형변환: age는 int로, height는 float로!
```

### 기억할 점 💡

- **`input()`은 항상 문자열!** 숫자 계산을 하려면 반드시 `int()` 또는 `float()`로 변환하세요
- **형변환은 에러가 날 수 있습니다!** 변환 가능한 값인지 항상 확인하세요
- **f-string은 자동 변환!** 출력할 때는 형변환 없이도 숫자를 문자열에 포함할 수 있습니다
