## 📖 변수(Variable)란 무엇인가?

### 변수는 데이터를 담는 상자입니다

프로그래밍에서 **변수(Variable)**는 데이터를 저장하는 메모리 공간입니다. 마치 라벨이 붙은 상자에 물건을 넣어두는 것처럼, 변수에는 이름을 붙여서 데이터를 저장하고 나중에 다시 꺼내 쓸 수 있습니다.

### 왜 변수가 필요할까요?

같은 값을 여러 번 사용하거나, 계산 결과를 저장해두었다가 나중에 활용해야 할 때 변수를 사용합니다.

**변수 없이 코드를 작성하면:**
```python
print(15000 * 3)
print(15000 * 3 * 0.9)
print(15000 * 3 - 15000 * 3 * 0.1)
```

같은 값을 계속 반복해서 입력해야 하고, 나중에 가격이 바뀌면 모든 곳을 수정해야 합니다. 😰

**변수를 사용하면:**
```python
price = 15000
quantity = 3
total = price * quantity
discounted = total * 0.9
```

코드가 훨씬 읽기 쉽고, 수정도 간편합니다! ✨

### 메모리에서는 어떻게 저장될까요?

변수를 만들면 컴퓨터는 메모리(RAM)에 공간을 확보하고, 그 주소에 값을 저장합니다.

```python
age = 25
```

이 코드가 실행되면:
1. 메모리에 공간 확보
2. 그 공간에 `25`라는 값을 저장
3. 그 공간에 `age`라는 이름표를 붙임

나중에 `age`를 호출하면 Python은 그 이름표가 붙은 메모리 공간으로 가서 값을 가져옵니다.

### FastAPI에서는?

FastAPI로 웹 API를 만들 때도 변수를 사용해서 데이터를 저장하고 처리합니다.

```python
@app.get("/users/{user_id}")
def get_user(user_id: int):
    # 변수에 사용자 정보 저장
    name = "홍길동"
    age = 25
    email = "hong@example.com"
    
    # 변수를 사용해서 응답 생성
    return {
        "id": user_id,
        "name": name,
        "age": age,
        "email": email
    }
```

## 📊 자료형(Data Type)이란?

### 데이터의 종류를 구분합니다

Python에서 모든 데이터는 **자료형(Data Type)**을 가지고 있습니다. 자료형은 데이터가 어떤 종류인지, 어떤 연산이 가능한지를 결정합니다.

### 주요 숫자형 자료형

1. **int (정수)**: 소수점이 없는 숫자
   - 예: `10`, `-5`, `0`, `1000`
   - 사용 예: 나이, 학생 수, 주문 개수

2. **float (실수)**: 소수점이 있는 숫자
   - 예: `3.14`, `-0.5`, `175.5`
   - 사용 예: 가격, 키, 온도, 평균 점수

### 왜 구분해야 할까요?

- **메모리 효율**: int가 float보다 메모리를 적게 사용
- **정확도**: float는 소수점 계산에서 오차가 발생할 수 있음
- **연산 결과**: 같은 연산도 타입에 따라 결과가 다를 수 있음

```python
10 / 3    # 결과: 3.3333... (float)
10 // 3   # 결과: 3 (int)
```

### type() 함수

`type()` 함수는 변수나 값의 자료형을 확인할 때 사용합니다. 디버깅할 때 매우 유용합니다!

```python
type(25)      # <class 'int'>
type(3.14)    # <class 'float'>
type("hello") # <class 'str'>
```

### ⚠️ 주의사항

- 나눗셈(`/`)은 항상 float를 반환합니다
- int와 float를 함께 연산하면 결과는 float입니다
- 변수명은 숫자로 시작할 수 없습니다 (❌ `1st_place`, ✅ `first_place`)

## 1️⃣ 변수 선언과 할당

### 변수를 만들고 값을 저장하는 방법

Python에서 변수를 만드는 것은 아주 간단합니다. `=` (등호) 기호를 사용해서 값을 할당하면 됩니다.

**문법:**
```python
변수명 = 값
```

**실행 포인트:**
- 변수명은 의미있게 작성하세요 (❌ `a`, `b`, `x` → ✅ `age`, `price`, `total`)
- 여러 단어는 `_`(언더스코어)로 연결합니다: `student_count`, `total_price`
- 같은 줄에 여러 변수를 동시에 선언할 수 있습니다

````python
# 변수는 데이터를 저장하는 상자입니다
age = 25  # age라는 변수에 25를 저장
print("나이:", age)  # 변수에 저장된 값을 출력

# 변수명은 의미있게 작성합니다
student_count = 30  # 학생 수를 저장하는 변수
print("학생 수:", student_count)

# 여러 변수를 동시에 선언할 수도 있습니다 (튜플 언패킹)
name, height = "홍길동", 175  # name에 "홍길동", height에 175 저장
print("이름:", name, "키:", height)
````

**예상 출력:**
```
나이: 25
학생 수: 30
이름: 홍길동 키: 175
```

**코드 해석:**
- 첫 번째 줄: `age`라는 변수에 정수 `25`를 저장
- 다섯 번째 줄: `student_count`에 `30` 저장
- 아홉 번째 줄: `name`과 `height`에 동시에 값을 할당 (이것을 "언패킹"이라고 합니다)

### 💡 팁: 좋은 변수명 작성하기

**나쁜 예:**
```python
a = 25
x = 30
t = 15000
```

**좋은 예:**
```python

student_count = 30
product_price = 15000
```

좋은 변수명은 코드를 읽는 사람(미래의 나 포함!)이 쉽게 이해할 수 있게 해줍니다.

## 2️⃣ type() 함수로 타입 확인하기

### 변수의 자료형을 확인하는 방법

`type()` 함수는 변수나 값이 어떤 자료형인지 알려줍니다. 프로그래밍하다가 "이 변수가 숫자인가? 문자열인가?" 헷갈릴 때 유용합니다.

**문법:**
```python
type(변수명)
type(값)
```

**실행 포인트:**
- `<class 'int'>`: 정수형
- `<class 'float'>`: 실수형
- 나중에 배울 타입들: `str`(문자열), `list`(리스트), `dict`(딕셔너리)

````python
# type() 함수는 변수의 자료형을 알려줍니다
age = 25  # 정수(int)를 저장
print(type(age))  # age 변수의 타입 확인

height = 175.5  # 실수(float)를 저장
print(type(height))  # height 변수의 타입 확인

# int는 정수, float는 실수(소수점이 있는 숫자)입니다
score = 100  # 정수 - 소수점이 없는 숫자
temperature = 36.5  # 실수 - 소수점이 있는 숫자
````

**예상 출력:**
```
<class 'int'>
<class 'float'>
```

**코드 해석:**
- `age`는 `25`를 저장하므로 정수(int) 타입
- `height`는 `175.5`를 저장하므로 실수(float) 타입
- Python은 소수점의 유무로 자동으로 타입을 결정합니다

### 🔍 깊이 알아보기: int vs float

**같은 숫자라도 표현 방식에 따라 타입이 달라집니다:**
```python
x = 10      # int
y = 10.0    # float (소수점이 있으면 float)
```

**실무에서는 언제 뭘 쓸까요?**
- **int 사용**: 개수, 순번, 나이, ID 등 정확한 정수값
- **float 사용**: 가격, 키, 무게, 온도, 평균 등 소수점이 필요한 값

## 3️⃣ 사칙 연산

### Python으로 계산하기

Python은 계산기처럼 사용할 수 있습니다. 기본 사칙 연산은 물론, 몫과 나머지, 거듭제곱까지 간단하게 계산할 수 있습니다.

**연산자 종류:**
- `+`: 덧셈
- `-`: 뺄셈
- `*`: 곱셈
- `/`: 나눗셈 (항상 float 반환)
- `//`: 몫 (정수 나눗셈)
- `%`: 나머지
- `**`: 거듭제곱

**실행 포인트:**
- `/` 연산은 항상 float를 반환합니다 (10 / 2 = 5.0)
- `//`는 소수점을 버리고 몫만 반환합니다
- `%`는 홀짝 판별, 배수 확인 등에 유용합니다

````python
# 기본 사칙 연산
a = 10  # 첫 번째 숫자
b = 3   # 두 번째 숫자

print(a + b)  # 덧셈: 10 + 3 = 13
print(a - b)  # 뺄셈: 10 - 3 = 7
print(a * b)  # 곱셈: 10 × 3 = 30
print(a / b)  # 나눗셈: 10 ÷ 3 = 3.333... (float 반환)

# 특수 연산자
print(a // b)  # 몫: 10을 3으로 나눈 몫 = 3
print(a % b)   # 나머지: 10을 3으로 나눈 나머지 = 1
print(a ** b)  # 거듭제곱: 10의 3제곱 = 1000
````

**예상 출력:**
```
13
7
30
3.3333333333333335
3
1
1000
```

**코드 해석:**
- `/` 연산의 결과는 항상 float입니다 (3.333...)
- `//` 연산은 소수점을 버리고 몫만 반환합니다 (3)
- `%` 연산은 나머지만 반환합니다 (10 = 3×3 + 1, 나머지는 1)
- `**` 연산은 거듭제곱입니다 (10³ = 1000)

### 💡 실무 활용 예시

**몫과 나머지는 언제 쓸까요?**

1. **홀수/짝수 판별:**
```python

    print("짝수")
else:

```

2. **페이지 계산:**

## 4️⃣ 복합 연산자

### 계산과 할당을 동시에!

복합 연산자는 변수의 값을 변경할 때 코드를 간결하게 만들어줍니다. 특히 카운터나 누적 계산에 매우 유용합니다.

**주요 복합 연산자:**
- `+=`: 더한 후 저장
- `-=`: 뺀 후 저장
- `*=`: 곱한 후 저장
- `/=`: 나눈 후 저장

**실행 포인트:**
- `count += 1`은 `count = count + 1`과 완전히 동일합니다
- 코드가 짧아지고 읽기 쉬워집니다
- 실무에서 매우 자주 사용됩니다

````python
# 복합 연산자는 계산과 할당을 동시에 합니다
count = 10  # 초기값 설정
print("초기값:", count)

count += 5  # count = count + 5 와 같음 (10 + 5 = 15)
print("5 더한 후:", count)

count -= 3  # count = count - 3 와 같음 (15 - 3 = 12)
print("3 뺀 후:", count)

count *= 2  # count = count * 2 와 같음 (12 × 2 = 24)
print("2 곱한 후:", count)

count /= 4  # count = count / 4 와 같음 (24 ÷ 4 = 6.0)
print("4 나눈 후:", count)
````

**예상 출력:**
```
초기값: 10
5 더한 후: 15
3 뺀 후: 12
2 곱한 후: 24
4 나눈 후: 6.0
```

**코드 해석:**
- `count += 5`: count에 5를 더한 값을 다시 count에 저장
- 각 연산마다 count의 값이 누적되어 변경됩니다
- 마지막 나눗셈 결과는 float(6.0)입니다

### 🔍 왜 복합 연산자를 쓸까요?

**일반 방식:**

**복합 연산자 방식:**

훨씬 간결하고 읽기 쉽죠? 실무에서는 거의 항상 복합 연산자를 사용합니다.

## 5️⃣ 실전 예제: 상품 가격 계산

### 지금까지 배운 것을 모두 활용해봅시다!

실제 쇼핑몰 시스템을 만든다고 생각해보세요. 상품 가격, 수량, 할인율을 계산하는 프로그램입니다.

**이 예제에서 사용하는 개념:**
- ✅ 변수 선언과 할당
- ✅ 사칙 연산
- ✅ 복합 연산자는 아니지만, 계산 결과를 변수에 저장

**실행 포인트:**
- 변수명이 의미있게 작성되어 있어서 코드가 읽기 쉽습니다
- 단계별로 계산하면 복잡한 계산도 쉽게 할 수 있습니다
- 실무에서도 이렇게 단계별로 나눠서 계산합니다

````python
# 상품 정보 설정
price = 15000      # 상품 가격 (원)
quantity = 3       # 구매 수량 (개)
discount = 0.1     # 할인율 (10% = 0.1)

# 1단계: 정가 계산 (가격 × 수량)
total = price * quantity
print("정가:", total, "원")  # 15000 × 3 = 45000원

# 2단계: 할인가 계산
# 할인가 = 정가 × (1 - 할인율)
# 10% 할인 = 90%만 지불 = 1 - 0.1 = 0.9
discounted = total * (1 - discount)
print("할인가:", discounted, "원")  # 45000 × 0.9 = 40500원

# 3단계: 절약한 금액 계산
saved = total - discounted
print("절약한 금액:", saved, "원")  # 45000 - 40500 = 4500원
````

**예상 출력:**
```
정가: 45000 원
할인가: 40500.0 원
절약한 금액: 4500.0 원
```

**코드 해석:**
1. **정가 계산**: 15,000원 × 3개 = 45,000원
2. **할인가 계산**: 45,000원 × (1 - 0.1) = 45,000 × 0.9 = 40,500원
3. **절약 금액**: 45,000원 - 40,500원 = 4,500원

### 💡 왜 (1 - discount)를 사용할까요?

할인을 계산하는 방법은 두 가지입니다:

**방법 1: 할인 금액을 빼기**
```python
discount_amount = total * discount  # 할인 금액 = 45000 × 0.1 = 4500
final_price = total - discount_amount  # 최종 가격 = 45000 - 4500 = 40500
```

**방법 2: (1 - 할인율) 곱하기** ← 우리가 사용한 방법
```python
final_price = total * (1 - discount)  # 45000 × 0.9 = 40500
```

두 번째 방법이 더 간단하고 실무에서 자주 사용됩니다!

### 🔍 실무에서는?

FastAPI로 쇼핑몰 API를 만들 때 이런 계산 로직을 사용합니다:

```python
@app.post("/calculate-price")
def calculate_price(price: float, quantity: int, discount: float = 0.0):
    total = price * quantity
    discounted = total * (1 - discount)
    saved = total - discounted
    
    return {
        "original_price": total,
        "discounted_price": discounted,
        "saved_amount": saved
    }
```

지금 배운 내용이 바로 실무에서 사용되는 코드입니다! 🎉
