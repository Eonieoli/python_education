## 📖 키워드 인수(Keyword Arguments)란 무엇인가?

### 지금까지 배운 "위치 인수"의 문제점

지금까지 우리는 함수를 호출할 때 **순서대로** 값을 전달했습니다:

```python
def introduce(name, age, city):
    return f"{name}님은 {age}세이고 {city}에 살고 있습니다"

introduce("홍길동", 25, "서울")  # 순서를 지켜야 함!
```

하지만 이 방식에는 몇 가지 문제가 있습니다:

1. **순서를 외워야 합니다** - `introduce(25, "홍길동", "서울")`처럼 실수하기 쉽습니다
2. **코드 가독성이 떨어집니다** - 무엇이 무엇인지 한눈에 알기 어렵습니다
3. **매개변수가 많아지면 헷갈립니다** - 매개변수가 5개, 10개가 되면?

### 키워드 인수의 등장! ✨

**키워드 인수(Keyword Arguments)**는 이런 문제를 해결합니다. 매개변수의 **이름을 명시**하면서 값을 전달하는 방식입니다:

```python
introduce(name="홍길동", age=25, city="서울")  # 이름을 명시!
introduce(city="부산", name="김철수", age=30)  # 순서를 바꿔도 됨!
```

### 장점이 뭔가요?

1. **가독성 향상**: 코드만 봐도 무슨 값인지 바로 알 수 있습니다
2. **순서 자유**: 순서를 바꿔도 됩니다 (이름으로 매칭하니까!)
3. **실수 방지**: 순서를 잘못 쓸 걱정이 없습니다
4. **유지보수 용이**: 나중에 코드를 봐도 이해하기 쉽습니다

### FastAPI에서는?

FastAPI로 API를 만들 때 **쿼리 파라미터(Query Parameter)**를 받는 방식이 바로 키워드 인수 패턴입니다:

```python
# FastAPI 예시 (나중에 배울 내용)
@app.get("/search")
def search_products(keyword: str, min_price: int = None, max_price: int = None):
    # /search?keyword=노트북&min_price=500000&max_price=1500000
    # 이런 URL의 파라미터들이 키워드 인수로 전달됩니다!
    ...
```

## 💡 None이란 무엇인가?

### "값이 없음"을 표현하는 특별한 값

Python에는 **None**이라는 특별한 값이 있습니다. 이것은 "값이 없다", "아무것도 아니다"를 의미합니다.

```python
x = None  # "x에는 아무 값도 없어요"라는 의미
```

### 왜 필요한가요?

실무에서는 이런 상황이 자주 발생합니다:

- 사용자가 **선택적으로** 정보를 입력할 수 있습니다
- 데이터베이스에서 **값을 찾지 못했을 때** 표현해야 합니다
- API에서 **필수가 아닌 매개변수**를 받아야 합니다

예를 들어볼까요?

```python
# 회원가입 시나리오
# - 이름, 이메일은 필수
# - 자기소개는 선택 (안 써도 됨)

def create_user(name, email, bio=None):  # bio는 선택!
    if bio is None:  # 값이 없으면
        bio = "소개가 없습니다"  # 기본값 설정
    return {"name": name, "email": email, "bio": bio}

# 자기소개 없이 회원가입
user1 = create_user("홍길동", "hong@example.com")
# bio가 None으로 전달되고, 함수 내부에서 기본값으로 변경됨

# 자기소개 작성하고 회원가입
user2 = create_user("김철수", "kim@example.com", bio="Python 개발자")
```

### None vs 빈 문자열 vs 0

초보자들이 자주 헷갈리는 부분인데, 이 셋은 **모두 다릅니다**:

```python
value1 = None   # 값이 없음
value2 = ""     # 빈 문자열 (값은 있는데 비어있음)
value3 = 0      # 숫자 0 (값이 있고, 그 값이 0)

# 이 셋은 서로 다른 의미입니다!
```

### FastAPI에서의 None 활용

FastAPI에서는 **선택적 쿼리 파라미터**를 받을 때 None을 기본값으로 사용합니다:

```python
# FastAPI 예시
@app.get("/products")
def search_products(
    keyword: str,           # 필수 매개변수
    min_price: int = None,  # 선택적 매개변수
    max_price: int = None   # 선택적 매개변수
):
    # /products?keyword=노트북  ← 이것도 가능
    # /products?keyword=노트북&min_price=500000  ← 이것도 가능
    ...
```

### ⚠️ 중요한 주의사항

None을 체크할 때는 반드시 `is` 또는 `is not`을 사용하세요:

```python
# ✅ 올바른 방법
if bio is None:
    ...

if bio is not None:
    ...

# ❌ 잘못된 방법 (작동은 하지만 권장하지 않음)
if bio == None:
    ...
```

왜냐하면 None은 특별한 값이고, Python에서는 단 하나만 존재하기 때문입니다. `is`는 "정확히 같은 객체인가?"를 확인하는 연산자입니다.

## 1. 키워드 인수 기본 사용법

### 함수를 호출할 때 매개변수 이름을 명시하기

가장 기본적인 키워드 인수 사용법을 살펴보겠습니다. 우리가 만든 함수의 매개변수 이름을 그대로 사용하면 됩니다!

````python
def introduce(name, age, city):
    """키워드 인수로 호출할 수 있는 함수"""
    return f"{name}님은 {age}세이고 {city}에 살고 있습니다"

# 위치 인수로 호출 (기존 방식)
# - 순서대로 값을 전달해야 합니다
# - 순서를 틀리면 엉뚱한 결과가 나옵니다
result1 = introduce("홍길동", 25, "서울")
print(result1)  # 출력: 홍길동님은 25세이고 서울에 살고 있습니다

# 키워드 인수로 호출 (새로운 방식!)
# - 매개변수 이름을 명시하면서 값을 전달합니다
# - 순서를 바꿔도 됩니다! (이름으로 매칭하니까)
result2 = introduce(city="부산", name="김철수", age=30)
print(result2)  # 출력: 김철수님은 30세이고 부산에 살고 있습니다
````

**실행 결과:**
```
홍길동님은 25세이고 서울에 살고 있습니다
김철수님은 30세이고 부산에 살고 있습니다
```

### 💡 핵심 포인트

1. **위치 인수**: 순서대로 전달 → `introduce("홍길동", 25, "서울")`
2. **키워드 인수**: 이름을 명시하며 전달 → `introduce(name="홍길동", age=25, city="서울")`
3. **순서 자유**: 키워드로 호출하면 순서 상관없음 → `introduce(city="부산", age=30, name="김철수")`

### 🔍 언제 사용하나요?

- 매개변수가 3개 이상일 때 (가독성 향상)
- 비슷한 타입의 매개변수가 여러 개일 때 (혼동 방지)
- 나중에 코드를 읽을 사람을 위해 (유지보수)

## 2. 위치 인수와 키워드 인수 혼합하기

### 실무에서 가장 많이 사용하는 패턴

위치 인수와 키워드 인수를 **함께** 사용할 수 있습니다. 보통 다음과 같은 패턴으로 사용합니다:

1. **필수 매개변수** → 위치 인수로 전달
2. **선택적 매개변수** → 키워드 인수로 전달

이렇게 하면 코드가 더 명확해집니다!

````python
def create_profile(name, age, email="미입력", phone="미입력"):
    """일부는 위치, 일부는 키워드 인수로 호출"""
    return f"이름: {name}, 나이: {age}, 이메일: {email}, 전화번호: {phone}"

# 필수 매개변수는 위치로, 선택 매개변수는 키워드로
# - name, age는 필수이므로 순서대로 전달
# - email만 선택적으로 제공 (phone은 기본값 사용)
result3 = create_profile("이영희", 28, email="young@example.com")
print(result3)  # 출력: 이름: 이영희, 나이: 28, 이메일: young@example.com, 전화번호: 미입력

# 모두 키워드로 호출 (가장 명확!)
# - 코드만 봐도 무슨 값인지 바로 알 수 있습니다
# - 실무에서 권장하는 방식입니다
result4 = create_profile(
    name="박민수", 
    age=35, 
    email="park@example.com", 
    phone="010-1234-5678"
)
print(result4)  # 출력: 이름: 박민수, 나이: 35, 이메일: park@example.com, 전화번호: 010-1234-5678
````

**실행 결과:**
```
이름: 이영희, 나이: 28, 이메일: young@example.com, 전화번호: 미입력
이름: 박민수, 나이: 35, 이메일: park@example.com, 전화번호: 010-1234-5678
```

### ⚠️ 중요한 규칙

위치 인수와 키워드 인수를 혼합할 때는 **반드시 위치 인수가 먼저** 와야 합니다:

```python
# ✅ 올바른 방법
create_profile("홍길동", 25, email="hong@example.com")  # 위치 먼저, 키워드 나중

# ❌ 잘못된 방법 (에러 발생!)
create_profile(name="홍길동", 25)  # 키워드 다음에 위치 인수는 안 됨!
```

### 💡 실무 팁

- **필수 정보 2-3개**: 위치 인수로 간단하게
- **선택 정보**: 키워드 인수로 명확하게
- **모두 키워드**: 매개변수가 5개 이상이면 모두 키워드로 쓰는 게 좋습니다

## 3. None 기본값 패턴 - 선택적 매개변수

### "있어도 되고 없어도 되는" 정보 처리하기

실무에서 가장 자주 사용하는 패턴입니다! 사용자가 **선택적으로** 제공할 수 있는 정보를 처리할 때 사용합니다.

**핵심 패턴**:
1. 매개변수의 기본값을 `None`으로 설정
2. 함수 내부에서 `if bio is None:` 체크
3. None이면 적절한 기본값으로 설정

````python
def create_user(name, bio=None):
    """bio가 None이면 기본값을 설정하는 패턴"""
    # None 체크 후 기본값 설정
    # - bio가 None이면 (사용자가 입력하지 않았으면)
    # - 기본 메시지로 대체합니다
    if bio is None:
        bio = "소개가 없습니다"
    
    return {
        "name": name,
        "bio": bio
    }

# bio를 제공하지 않으면 None이 전달됨
# - create_user("홍길동") 호출
# - bio 매개변수가 없으므로 bio=None으로 설정됨
# - 함수 내부에서 "소개가 없습니다"로 변경됨
user1 = create_user("홍길동")
print(user1)  # 출력: {'name': '홍길동', 'bio': '소개가 없습니다'}

# bio를 제공하면 그 값이 사용됨
# - create_user("김철수", bio="Python 개발자입니다") 호출
# - bio에 실제 값이 있으므로 None이 아님
# - if문을 건너뛰고 원래 값 사용
user2 = create_user("김철수", bio="Python 개발자입니다")
print(user2)  # 출력: {'name': '김철수', 'bio': 'Python 개발자입니다'}
````

**실행 결과:**
```
{'name': '홍길동', 'bio': '소개가 없습니다'}
{'name': '김철수', 'bio': 'Python 개발자입니다'}
```

### 🎯 이 패턴이 중요한 이유

1. **유연성**: 사용자가 선택적으로 정보를 제공할 수 있습니다
2. **기본값 제공**: None일 때 적절한 기본값을 설정합니다
3. **FastAPI 핵심**: API에서 선택적 매개변수를 받을 때 사용합니다

### 💡 왜 빈 문자열 ""이 아니라 None인가요?

```python
# ❌ 빈 문자열을 기본값으로 사용하면?
def create_user(name, bio=""):
    # 사용자가 정말 빈 문자열을 입력한 건지,
    # 아니면 입력하지 않은 건지 구분할 수 없습니다!
    if bio == "":
        bio = "소개가 없습니다"

# ✅ None을 사용하면?
def create_user(name, bio=None):
    # "입력하지 않음"을 명확하게 표현할 수 있습니다
    if bio is None:
        bio = "소개가 없습니다"
```

None은 "값이 없다"는 것을 명확하게 표현하는 Python의 공식적인 방법입니다!

## 4. None 처리 - "값 없음" 표현하기

### 데이터를 찾지 못했을 때 None 반환하기

데이터베이스에서 사용자를 찾지 못했을 때, 또는 검색 결과가 없을 때 어떻게 표현할까요?
바로 **None**을 반환하면 됩니다! 이것이 Python의 표준 패턴입니다.

````python
def get_user_by_id(user_id, default=None):
    """사용자를 찾지 못하면 None 또는 default를 반환"""
    # 간단한 사용자 "데이터베이스" (딕셔너리로 시뮬레이션)
    # - 실제로는 진짜 데이터베이스에서 가져옵니다
    # - 여기서는 학습을 위해 딕셔너리 사용
    users = {
        1: "홍길동",
        2: "김철수",
        3: "이영희"
    }
    
    # 사용자가 없으면 default 반환
    # - user_id가 users 딕셔너리에 없으면
    # - default 값을 반환합니다 (기본값은 None)
    if user_id not in users:
        return default
    
    # 사용자가 있으면 이름 반환
    return users[user_id]

# 사용자를 찾으면 이름 반환
# - user_id=1은 딕셔너리에 있습니다
# - "홍길동"을 반환합니다
result5 = get_user_by_id(1)
print(f"사용자 1: {result5}")  # 출력: 사용자 1: 홍길동

# 사용자를 못 찾으면 None 반환
# - user_id=999는 딕셔너리에 없습니다
# - default 매개변수가 None이므로 None을 반환합니다
result6 = get_user_by_id(999)
print(f"사용자 999: {result6}")  # 출력: 사용자 999: None

# 사용자를 못 찾으면 기본값 반환
# - user_id=999는 딕셔너리에 없습니다
# - default="알 수 없는 사용자"를 제공했으므로 그 값을 반환합니다
result7 = get_user_by_id(999, default="알 수 없는 사용자")
print(f"사용자 999: {result7}")  # 출력: 사용자 999: 알 수 없는 사용자
````

**실행 결과:**
```
사용자 1: 홍길동
사용자 999: None
사용자 999: 알 수 없는 사용자
```

### 🎯 None 반환 패턴의 의미

이 패턴은 실무에서 **매우 자주** 사용됩니다:

```python
# 데이터베이스 조회
user = get_user_by_id(user_id)

if user is None:
    # 사용자를 찾지 못했을 때의 처리
    print("사용자가 없습니다")
else:
    # 사용자를 찾았을 때의 처리
    print(f"환영합니다, {user}님!")
```

### 💡 왜 에러를 발생시키지 않고 None을 반환하나요?

두 가지 접근 방식이 있습니다:

```python
# 방법 1: None 반환 (더 유연함)
def get_user(user_id):
    if user_not_found:
        return None  # 호출자가 None을 체크해서 처리

# 방법 2: 예외 발생 (더 명확함)
def get_user(user_id):
    if user_not_found:
        raise ValueError("사용자를 찾을 수 없습니다")  # 에러 발생
```

**None 반환이 좋은 경우**:
- "찾지 못함"이 정상적인 상황일 때
- 호출자가 None을 체크해서 다양하게 처리할 수 있을 때

**예외 발생이 좋은 경우**:
- "찾지 못함"이 심각한 오류일 때
- 반드시 처리되어야 할 때

## 5. FastAPI 스타일 - API 매개변수 패턴

### 실무에서 사용하는 검색/필터링 함수 패턴

드디어 FastAPI에서 사용하는 패턴을 미리 배워봅니다! 쇼핑몰의 상품 검색 기능을 생각해보세요:

- **필수**: 검색 키워드 (없으면 검색 불가)
- **선택**: 최소 가격, 최대 가격 (필터링 옵션)

이런 패턴을 함수로 만들어봅시다!

````python
def search_products(keyword, min_price=None, max_price=None):
    """FastAPI 쿼리 파라미터처럼 동작하는 함수"""
    # 검색 조건 문자열 생성
    # - 항상 키워드는 포함됩니다 (필수니까!)
    conditions = [f"키워드: {keyword}"]
    
    # min_price가 None이 아니면 (사용자가 제공했으면)
    # - 최소가격 조건을 추가합니다
    if min_price is not None:
        conditions.append(f"최소가격: {min_price}원")
    
    # max_price가 None이 아니면 (사용자가 제공했으면)
    # - 최대가격 조건을 추가합니다
    if max_price is not None:
        conditions.append(f"최대가격: {max_price}원")
    
    # 조건들을 쉼표로 연결해서 반환
    return ", ".join(conditions)

# 키워드만 검색
# - 최소/최대 가격은 None으로 남아있음
# - 가격 필터 없이 검색
result8 = search_products(keyword="노트북")
print(result8)  # 출력: 키워드: 노트북

# 키워드 + 최소가격
# - min_price만 제공
# - max_price는 None으로 남음
result9 = search_products(keyword="노트북", min_price=500000)
print(result9)  # 출력: 키워드: 노트북, 최소가격: 500000원

# 키워드 + 최소가격 + 최대가격
# - 모든 필터 조건 제공
# - 50만원 ~ 150만원 사이의 노트북 검색
result10 = search_products(keyword="노트북", min_price=500000, max_price=1500000)
print(result10)  # 출력: 키워드: 노트북, 최소가격: 500000원, 최대가격: 1500000원
````

**실행 결과:**
```
키워드: 노트북
키워드: 노트북, 최소가격: 500000원
키워드: 노트북, 최소가격: 500000원, 최대가격: 1500000원
```

### 🚀 FastAPI에서는 이렇게 사용됩니다!

위의 함수 패턴이 FastAPI에서는 실제 API로 변환됩니다:

```python
from fastapi import FastAPI

app = FastAPI()

@app.get("/products")  # GET 요청을 받는 엔드포인트
def search_products(keyword: str, min_price: int = None, max_price: int = None):
    # 우리가 방금 만든 함수와 똑같은 패턴!
    conditions = [f"키워드: {keyword}"]
    
    if min_price is not None:
        conditions.append(f"최소가격: {min_price}원")
    
    if max_price is not None:
        conditions.append(f"최대가격: {max_price}원")
    
    return {"search_conditions": ", ".join(conditions)}

# 실제 API 호출 예시:
# GET /products?keyword=노트북
# GET /products?keyword=노트북&min_price=500000
# GET /products?keyword=노트북&min_price=500000&max_price=1500000
```

### 💡 이 패턴의 핵심

1. **필수 매개변수**: 기본값 없음 → `keyword: str`
2. **선택적 매개변수**: `None` 기본값 → `min_price: int = None`
3. **조건부 처리**: `if ... is not None:` 체크
4. **키워드 인수**: 명확한 호출 → `search_products(keyword="노트북", min_price=500000)`

이 4가지 패턴이 FastAPI의 핵심입니다! 오늘 배운 내용이 그대로 사용됩니다. 🎉

## 📌 핵심 정리

이번 시간에 배운 내용:

1. **키워드 인수(Keyword Arguments)**: 매개변수 이름을 명시하며 함수 호출 → 가독성 향상, 순서 자유
2. **None의 의미**: "값이 없음"을 표현하는 Python의 특별한 값

5. **FastAPI 패턴**: 필수 매개변수 + 선택적 매개변수(None) → API 쿼리 파라미터의 기본 구조
