## 📖 함수(Function)란 무엇인가?

### 함수는 "재사용 가능한 코드 블록"입니다

여러분이 계산기 프로그램을 만든다고 생각해봅시다. 덧셈을 10번 해야 한다면 어떻게 하시겠습니까?

```python
# 함수 없이
result1 = 10 + 20
result2 = 30 + 40
result3 = 50 + 60
# ... 계속 반복?
```

**함수를 사용하면**:

```python
def add(a, b):
    return a + b

result1 = add(10, 20)
result2 = add(30, 40)
result3 = add(50, 60)
```

훨씬 간결하고 읽기 쉽죠! 🎯

---

### 왜 함수가 필요할까요?

#### 1. **코드 중복 제거**
같은 작업을 여러 번 해야 할 때, 함수 하나만 만들어두면 계속 재사용할 수 있습니다.

#### 2. **유지보수가 쉬워집니다**
만약 할인율을 10%에서 15%로 바꿔야 한다면?
- 함수 없이: 코드 10군데를 모두 수정
- 함수 있을 때: 함수 1곳만 수정! ✨

#### 3. **코드를 이해하기 쉬워집니다**
```python
# 무슨 코드인지 바로 이해 가능!
user_grade = calculate_member_grade(total_purchase)
```

### 함수의 구조: 입력 → 처리 → 출력

```
┌─────────────────────────┐
│  입력 (매개변수)         │
│  ↓                      │
│  처리 (함수 내부 코드)   │
│  ↓                      │
│  출력 (return 값)       │
└─────────────────────────┘
```

예시:
```python
def calculate_total(price, quantity):  # ← 입력 (매개변수)
    total = price * quantity           # ← 처리
    return total                       # ← 출력 (반환값)
```

### 매개변수(Parameter)와 반환값(Return Value)

#### 매개변수: 함수의 입력
- 함수에 전달하는 값
- 함수 정의 시 `def function(parameter):` 형태로 선언
- 여러 개 가능: `def function(param1, param2, param3):`

#### 반환값: 함수의 출력
- `return` 키워드로 반환
- 함수 실행 결과를 호출한 곳으로 돌려줌
- **여러 개도 반환 가능!** (튜플로 반환됨) ← 오늘의 핵심! ⭐

```python
# 여러 개 반환 예시 (Day 2의 튜플 언패킹 활용!)
def get_user_info():
    name = "Alice"

# 언패킹으로 받기
user_name, user_age = get_user_info()
```

## 1. def 키워드로 함수 정의하기

함수를 만드는 것을 "함수를 정의한다"고 표현합니다. Python에서는 `def` 키워드를 사용합니다.

### 기본 문법

```python
def 함수명(매개변수1, 매개변수2, ...):
    # 실행할 코드
    return 결과값
```

- `def`: "define(정의하다)"의 약자
- 함수명: 동사로 시작하는 것이 좋음 (get_user, calculate_total, check_password 등)
- 콜론(:) 필수!
- 들여쓰기로 함수 내부 코드 구분

---

### 1-1. 가장 간단한 함수 (매개변수 없음, 반환값 없음)

````python
# 가장 간단한 함수 정의
def say_hello():
    """인사말을 출력하는 함수"""
    print("안녕하세요!")

# 함수 호출 - 함수명()로 실행
say_hello()  # 출력: 안녕하세요!
````

**실행 결과:**
```
안녕하세요!
```

**설명:**
- `def say_hello():`: 매개변수가 없는 함수 정의 (괄호 안이 비어있음)
- `print("안녕하세요!")`: 함수가 실행할 코드
- `say_hello()`: 함수 호출 (함수명 뒤에 괄호 필수!)
- 이 함수는 return이 없어서 None을 반환합니다 (곧 배울 예정)

> **💡 Tip**: 함수 이름 바로 아래의 `"""..."""`는 **docstring**이라고 하며, 함수의 설명을 적는 곳입니다. 생략 가능하지만 쓰는 습관을 들이면 좋습니다!

---

### 1-2. 매개변수가 있는 함수

이제 입력값을 받는 함수를 만들어봅시다!

````python
# 매개변수가 있는 함수
def greet(name):  # name: 매개변수
    """이름을 받아서 인사하는 함수"""
    print(f"안녕하세요, {name}님!")

# 함수 호출 시 실제 값(인자)을 전달
greet("홍길동")  # 출력: 안녕하세요, 홍길동님!
greet("김철수")  # 출력: 안녕하세요, 김철수님!
````

**실행 결과:**
```
안녕하세요, 홍길동님!
안녕하세요, 김철수님!
```

**설명:**
- `def greet(name):`: `name`이라는 매개변수를 받는 함수
- `greet("홍길동")`: "홍길동"이라는 **인자(argument)**를 전달
- 매개변수(parameter)는 **함수 정의 시 사용**, 인자(argument)는 **함수 호출 시 전달**
- 같은 함수를 다른 값으로 여러 번 호출 가능!

> **🔍 용어 정리**:
> - **매개변수(Parameter)**: 함수 정의 시 괄호 안의 변수 (`name`)
> - **인자(Argument)**: 함수 호출 시 전달하는 실제 값 (`"홍길동"`)

## 2. return으로 값 반환하기

지금까지 만든 함수는 `print()`로 화면에 출력만 했습니다. 
하지만 실무에서는 **결과값을 반환**해서 다른 곳에서 활용하는 경우가 훨씬 많습니다!

### return이 없으면?

먼저 return이 없을 때 어떻게 되는지 봅시다.

````python
# return이 없으면 None을 반환
def add_no_return(a, b):
    """두 숫자를 더하지만 반환하지 않는 함수"""
    a + b  # 계산만 하고 반환하지 않음!

result = add_no_return(10, 20)
print(result)  # 출력: None (반환값이 없음!)
print(type(result))  # 출력: <class 'NoneType'>
````

**실행 결과:**
```
None
<class 'NoneType'>
```

**설명:**
- `a + b`를 계산하지만 `return`이 없어서 결과가 사라짐!
- Python 함수는 return이 없으면 자동으로 `None`을 반환
- `None`: "값이 없음"을 나타내는 특별한 값 (Day 4-4교시에서 자세히 배움)

> **⚠️ 주의**: 계산만 하고 return하지 않으면 결과를 활용할 수 없습니다!

---

### 2-1. return으로 값을 반환하기

````python
# return으로 값을 반환
def add(a, b):
    """두 숫자를 더해서 결과를 반환하는 함수"""
    return a + b  # 계산 결과를 반환!

result = add(10, 20)
print(result)  # 출력: 30

# 반환값을 다른 계산에도 활용 가능!
total = add(100, 200) + add(50, 30)
print(total)  # 출력: 380 (300 + 80)
````

**실행 결과:**
```
30
380
```

**설명:**
- `return a + b`: 계산 결과를 호출한 곳으로 **반환**
- `result = add(10, 20)`: 반환된 값(30)이 result에 저장됨
- 반환값을 변수에 저장하거나, 다른 연산에 바로 사용 가능!
- `add(100, 200) + add(50, 30)`: 두 함수 호출 결과를 더할 수 있음

> **✅ 핵심**: return은 함수의 결과를 "돌려준다"는 의미입니다. 이 결과를 받아서 다양하게 활용할 수 있습니다!

---

### 2-2. return 후에는 함수가 즉시 종료됨

`return`을 만나면 함수는 **즉시 종료**되고, return 아래 코드는 실행되지 않습니다.

````python
# return 후에는 함수가 즉시 종료됨
def check_adult(age):
    """나이를 받아 성인 여부를 반환하는 함수"""
    if age >= 18:
        return "성인"  # 여기서 함수 종료!
    # if가 False면 여기까지 옴
    return "미성년자"  # if가 False일 때만 실행

print(check_adult(20))  # 출력: 성인
print(check_adult(15))  # 출력: 미성년자
````

**실행 결과:**
```
성인
미성년자
```

**설명:**
- `age >= 18`이 True면 첫 번째 return 실행 → 함수 즉시 종료
- `age >= 18`이 False면 첫 번째 return을 건너뛰고 두 번째 return 실행
- 이렇게 조건에 따라 다른 값을 반환할 수 있습니다

> **💡 Tip**: 함수 안에 여러 개의 return이 있어도 되지만, **실제로 실행되는 return은 하나**입니다!

### 왜 여러 개를 반환해야 할까요?

### 3-1. 여러 개 반환하기 - 튜플로 자동 묶임

````python
# 여러 개 반환 - 자동으로 튜플로 묶여서 반환됨!
def get_user_info():
    """사용자 정보를 한 번에 반환하는 함수"""
    name = "Alice"
    age = 25
    email = "alice@example.com"
    return name, age, email  # 쉼표로 구분하면 튜플로 반환!

# 반환값 확인
result = get_user_info()
print(result)  # 출력: ('Alice', 25, 'alice@example.com')
print(type(result))  # 출력: <class 'tuple'>
````

**실행 결과:**
```
('Alice', 25, 'alice@example.com')
<class 'tuple'>
```

**설명:**
- `return name, age, email`: 3개 값을 쉼표로 구분하면 자동으로 튜플로 묶임
- 실제로는 `return (name, age, email)`과 같은 의미 (괄호 생략 가능)
- 튜플로 받아서 인덱싱으로 접근 가능: `result[0]`, `result[1]`, `result[2]`

하지만 이렇게 받으면 불편합니다! 더 좋은 방법이 있습니다 👇

### 3-2. 언패킹으로 받기 (Day 2 복습!) ⭐

Day 2에서 배운 **튜플 언패킹**을 여기서 활용합니다!

````python
# 언패킹으로 각각의 변수로 받기 (Day 2 복습!)
user_name, user_age, user_email = get_user_info()

# 이제 각각의 변수로 쉽게 사용 가능!
print(f"이름: {user_name}")  # 출력: 이름: Alice
print(f"나이: {user_age}")  # 출력: 나이: 25
print(f"이메일: {user_email}")  # 출력: 이메일: alice@example.com
````

**실행 결과:**


**설명:**
- `user_name, user_age, user_email = get_user_info()`: **언패킹!**
- 반환된 튜플의 값들이 순서대로 각 변수에 할당됨
- 인덱싱(`result[0]`)보다 훨씬 읽기 쉽고 명확함!

### 3-3. 실전 예제: 계산 결과를 여러 개 반환

````python
# 실전 예제: 여러 계산 결과를 한 번에 반환
def calculate(a, b):
    """두 숫자의 덧셈, 뺄셈, 곱셈 결과를 모두 반환"""
    plus = a + b
    minus = a - b
    multiply = a * b
    return plus, minus, multiply  # 3개 반환!

# 언패킹으로 각각 받기
result_add, result_sub, result_mul = calculate(10, 3)

print(f"덧셈: {result_add}")  # 출력: 덧셈: 13
print(f"뺄셈: {result_sub}")  # 출력: 뺄셈: 7
print(f"곱셈: {result_mul}")  # 출력: 곱셈: 30
````

**실행 결과:**
```
덧셈: 13
뺄셈: 7
곱셈: 30
```

**설명:**
- 함수 하나로 3가지 계산을 한 번에 처리!
- 필요한 결과만 골라서 사용할 수도 있음:
  ```python
  add_result, _, _ = calculate(10, 3)  # 덧셈 결과만 필요할 때
  ```
- `_`: "사용하지 않는 값"을 나타내는 관례적 변수명

> **✨ FastAPI에서는?**
> FastAPI에서도 함수가 여러 값을 처리한 후 딕셔너리로 반환하는 패턴을 많이 사용합니다:
> ```python
> @app.get("/user/{user_id}")
> def get_user_detail(user_id: int):
>     # DB에서 여러 정보 조회
>     return {"name": name, "age": age, "email": email}
> ```

## 4. 함수 없이 vs 함수 있을 때 비교

### 시나리오: 여러 상품의 할인 가격 계산하기

### 4-1. ❌ 함수 없이 - 코드 중복!

````python
# ❌ 함수 없이 - 코드 중복 발생!
price1 = 10000
quantity1 = 3
total1 = price1 * quantity1 * 0.9  # 10% 할인
print(f"상품1 총액: {total1}원")

price2 = 15000
quantity2 = 2
total2 = price2 * quantity2 * 0.9  # 똑같은 계산 반복!
print(f"상품2 총액: {total2}원")

# 만약 할인율을 15%로 바꾸려면?
# → 모든 0.9를 0.85로 일일이 수정해야 함! 😱
````

**실행 결과:**
```
상품1 총액: 27000.0원
상품2 총액: 27000.0원
```

**문제점:**
- 똑같은 계산 코드(`price * quantity * 0.9`)가 반복됨
- 할인율을 바꾸려면? 모든 0.9를 찾아서 수정해야 함
- 상품이 100개라면? 😱😱😱

### 4-2. ✅ 함수 있을 때 - 재사용 가능!

````python
# ✅ 함수 있을 때 - 재사용 가능!
def calculate_total(price, quantity):
    """가격과 수량을 받아 10% 할인된 총액 반환"""
    return price * quantity * 0.9  # 10% 할인 적용

# 여러 상품에 쉽게 적용!
total1 = calculate_total(10000, 3)
total2 = calculate_total(15000, 2)
total3 = calculate_total(8000, 5)  # 새 상품도 쉽게 추가!

print(f"상품1 총액: {total1}원")  # 출력: 상품1 총액: 27000.0원
print(f"상품2 총액: {total2}원")  # 출력: 상품2 총액: 27000.0원
print(f"상품3 총액: {total3}원")  # 출력: 상품3 총액: 36000.0원

# 할인율을 15%로 바꾸려면?
# → 함수 안의 0.9를 0.85로 한 곳만 수정! ✨
````


**장점:**
- 코드 중복 제거 ✅

## 5. 실전 예제: 회원 등급 판정 시스템

지금까지 배운 내용을 모두 활용한 실전 예제입니다!

### 요구사항
- 총 구매 금액에 따라 회원 등급 결정
- 등급별 할인율도 함께 제공
- 여러 회원에게 적용 가능

### 등급 기준
- VIP: 100만원 이상 (20% 할인)
- Gold: 50만원 이상 (15% 할인)
- Silver: 10만원 이상 (10% 할인)
- Bronze: 10만원 미만 (5% 할인)

````python
def get_member_grade(total_purchase):
    """총 구매 금액에 따른 회원 등급과 할인율을 반환
    
    Args:
        total_purchase: 총 구매 금액 (정수)
    
    Returns:
        grade: 회원 등급 (문자열)
        discount: 할인율 (실수, 0.0~1.0)
    """
    # if/elif로 등급 판정
    if total_purchase >= 1000000:  # 100만원 이상
        grade = "VIP"
        discount = 0.2  # 20% 할인
    elif total_purchase >= 500000:  # 50만원 이상
        grade = "Gold"
        discount = 0.15  # 15% 할인
    elif total_purchase >= 100000:  # 10만원 이상
        grade = "Silver"
        discount = 0.1  # 10% 할인
    else:  # 10만원 미만
        grade = "Bronze"
        discount = 0.05  # 5% 할인
    
    return grade, discount  # 등급과 할인율 함께 반환! (튜플)

# 여러 회원의 등급 판정
print("=== 회원 등급 판정 시스템 ===")

# 회원 1: 150만원 구매
member_grade, member_discount = get_member_grade(1500000)
print(f"등급: {member_grade}, 할인율: {member_discount * 100}%")
# 출력: 등급: VIP, 할인율: 20.0%

# 회원 2: 35만원 구매
member_grade, member_discount = get_member_grade(350000)
print(f"등급: {member_grade}, 할인율: {member_discount * 100}%")
# 출력: 등급: Silver, 할인율: 10.0%

# 회원 3: 5만원 구매
member_grade, member_discount = get_member_grade(50000)
print(f"등급: {member_grade}, 할인율: {member_discount * 100}%")
# 출력: 등급: Bronze, 할인율: 5.0%
````

**실행 결과:**
```
=== 회원 등급 판정 시스템 ===
등급: VIP, 할인율: 20.0%
등급: Silver, 할인율: 10.0%
등급: Bronze, 할인율: 5.0%
```

**코드 분석:**
- `if/elif/else`: 조건에 따라 다른 등급 부여 (Day 3 복습)
- `return grade, discount`: **2개 값을 튜플로 반환** ⭐
- `member_grade, member_discount = get_member_grade(...)`: **언패킹으로 받기** ⭐
- `member_discount * 100`: 0.2 → 20.0으로 변환 (퍼센트 표시)

**이 예제가 보여주는 것:**
1. ✅ 복잡한 로직을 함수로 캡슐화
2. ✅ 여러 개 값을 반환하여 한 번에 여러 정보 제공
3. ✅ 같은 함수를 다른 입력값으로 재사용
4. ✅ 등급 기준이 바뀌면 함수 한 곳만 수정하면 됨!

> **💡 실무 Tip**: 이런 등급 판정 로직은 실제 쇼핑몰이나 서비스에서 자주 사용됩니다. FastAPI에서도 비슷한 패턴으로 구현하게 될 것입니다!
