## 📖 매개변수와 인수: 차이를 알아야 합니다!

함수를 배우면서 가장 헷갈리는 용어가 **매개변수(Parameter)**와 **인수(Argument)**입니다. 명확히 구분해봅시다!

### 매개변수 (Parameter)

함수를 **정의할 때** 괄호 안에 작성하는 변수입니다. 함수가 어떤 입력을 받을지 미리 정해놓는 것이죠.

```python
def greet(name):  # ← name이 매개변수
    return f"안녕하세요, {name}님!"
```

### 인수 (Argument)

함수를 **호출할 때** 실제로 전달하는 값입니다. 매개변수에 들어갈 구체적인 데이터죠.

```python
greet("홍길동")  # ← "홍길동"이 인수
```

### 왜 이걸 구분해야 하나요?

이제부터 배울 **위치 인수**, **기본값**, **키워드 인수** 등의 개념을 정확히 이해하려면 이 차이를 알아야 합니다. 매개변수는 "설계도"이고, 인수는 "실제 재료"라고 생각하면 쉽습니다! 🏗️

## 📖 위치 인수 (Positional Arguments)

위치 인수는 가장 기본적인 함수 호출 방식입니다. **순서대로** 매개변수에 값이 전달됩니다.

### 왜 "위치" 인수일까요?

함수를 호출할 때 인수를 전달하는 **위치(순서)**가 중요하기 때문입니다!

```python
def introduce(name, age):
    return f"이름: {name}, 나이: {age}"

introduce("김철수", 25)  # ✅ 올바른 순서
introduce(25, "김철수")  # ❌ 순서가 바뀌면 이상한 결과!
```

첫 번째 인수는 첫 번째 매개변수에, 두 번째 인수는 두 번째 매개변수에 자동으로 매칭됩니다.

### 실무에서는?

FastAPI를 배우게 되면 이런 패턴을 자주 보게 됩니다:

```python
# FastAPI 엔드포인트 예시 (미리보기)
@app.get("/users/{user_id}")
def get_user(user_id: int, db: Session):  # 위치에 따라 다른 역할!
    # user_id는 경로에서, db는 의존성에서 자동 주입
    pass
```

매개변수의 순서와 의미를 정확히 이해하는 것이 중요합니다! 💡

## 📖 기본값 (Default Values): 선택적 매개변수 만들기

기본값은 Python 함수의 **핵심 기능** 중 하나입니다!

### 기본값이 없으면?

모든 매개변수가 필수가 됩니다. 하나라도 빠뜨리면 에러가 발생하죠.

```python
def calculate_price(price, tax_rate):
    return price * (1 + tax_rate)

calculate_price(10000)  # ❌ TypeError: missing 1 required argument
```

### 기본값이 있으면?

해당 매개변수를 **선택적(optional)**으로 만들 수 있습니다!

```python
def calculate_price(price, tax_rate=0.1):  # 기본값 10%
    return price * (1 + tax_rate)

calculate_price(10000)         # ✅ tax_rate는 0.1로 자동 설정
calculate_price(10000, 0.05)   # ✅ 원하면 다른 값 전달 가능
```

### 왜 기본값이 중요한가요?

1. **유연성**: 대부분의 경우 기본값을 사용하되, 필요할 때만 변경 가능
2. **사용자 편의**: 함수 사용자가 매번 모든 인수를 전달할 필요가 없음
3. **하위 호환성**: 나중에 매개변수를 추가해도 기존 코드가 깨지지 않음

### FastAPI에서의 활용

FastAPI에서 쿼리 파라미터를 만들 때 기본값을 많이 사용합니다:

```python
# FastAPI 예시 (미리보기)
@app.get("/items/")
def read_items(skip: int = 0, limit: int = 10):  # 기본값으로 페이지네이션
    # skip과 limit를 전달하지 않으면 기본값 사용
    pass
```

### ⚠️ 중요한 규칙!

**기본값이 있는 매개변수는 반드시 기본값이 없는 매개변수보다 뒤에 와야 합니다!**

```python
# ❌ 잘못된 예시
def wrong(name="홍길동", age):  # SyntaxError!
    pass

# ✅ 올바른 예시
def correct(age, name="홍길동"):  # OK!
    pass
```

왜 그럴까요? 위치 인수로 호출할 때 Python이 어떤 값을 어디에 넣어야 할지 알 수 없기 때문입니다! 🤔

## 1. 위치 인수 기본 예제

가장 기본적인 형태의 함수입니다. 순서대로 인수를 전달하면 해당 매개변수에 값이 할당됩니다.

````python
# 함수를 호출할 때 인수는 순서대로 매개변수에 전달됩니다
def introduce(name, age):
    """
    자기소개 함수
    
    매개변수:
        name: 이름 (첫 번째 매개변수)
        age: 나이 (두 번째 매개변수)
    """
    return f"제 이름은 {name}이고, {age}살입니다."

# 위치 인수로 호출 - 순서가 중요합니다!
result = introduce("김철수", 25)  # "김철수"는 name에, 25는 age에 전달
print(result)  # 출력: 제 이름은 김철수이고, 25살입니다.

# 순서가 바뀌면 이상한 결과가 나옵니다!
# result = introduce(25, "김철수")  # 이렇게 하면 안 됩니다!
````

**실행 결과:**
```
제 이름은 김철수이고, 25살입니다.
```

**핵심 포인트:**
- 첫 번째 인수 `"김철수"`는 첫 번째 매개변수 `name`에 할당
- 두 번째 인수 `25`는 두 번째 매개변수 `age`에 할당
- 순서를 바꾸면 타입이 맞지 않아 문제가 발생합니다

### 💡 팁

매개변수가 많아질수록 순서를 기억하기 어려워집니다. 이럴 때는 다음 교시에서 배울 **키워드 인수**를 사용하면 됩니다!

## 2. 기본값 - 숫자 타입

숫자 타입의 기본값을 설정하는 예제입니다. 세금 계산처럼 **일반적으로 사용되는 값**을 기본값으로 설정하면 편리합니다.

````python
# 매개변수에 기본값을 설정할 수 있습니다
def calculate_price(price, tax_rate=0.1):
    """
    상품 가격 계산 함수
    
    매개변수:
        price: 원가 (필수)
        tax_rate: 세율 (선택, 기본값: 0.1 = 10%)
    
    반환값:
        세금이 포함된 최종 가격
    """
    # 원가 × (1 + 세율)로 최종 가격 계산
    return int(price * (1 + tax_rate))

# 기본값 사용 (tax_rate를 전달하지 않음)
total1 = calculate_price(10000)  # tax_rate는 자동으로 0.1 사용
print(f"세금 포함 가격: {total1}원")  # 출력: 세금 포함 가격: 11000원

# 기본값 대신 다른 값 전달 (5% 세율)
total2 = calculate_price(10000, 0.05)  # tax_rate를 0.05로 지정
print(f"세금 포함 가격: {total2}원")  # 출력: 세금 포함 가격: 10500원
````

**실행 결과:**
```
세금 포함 가격: 11000원
세금 포함 가격: 10500원
```

**핵심 포인트:**
- `tax_rate=0.1`: 기본 세율을 10%로 설정
- 대부분의 경우 10% 세율을 사용하므로 매번 입력할 필요 없음
- 특별한 경우(5% 세율)에만 값을 전달하면 됨

### 🔍 실무 활용

쇼핑몰 API를 만들 때 이런 패턴을 자주 사용합니다:
```python
# 대부분 일반 배송(기본값)을 사용하지만,
# 필요할 때만 특급 배송 요금을 계산
def calculate_shipping(weight, is_express=False):
    base_cost = weight * 1000
    return base_cost * 2 if is_express else base_cost
```

## 3. 기본값 - 문자열 타입

문자열도 기본값으로 설정할 수 있습니다. 인사말처럼 **자주 사용되는 문구**를 기본값으로 두면 코드가 간결해집니다.

````python
def greet(name, greeting="안녕하세요"):
    """
    인사 함수
    
    매개변수:
        name: 이름 (필수)
        greeting: 인사말 (선택, 기본값: "안녕하세요")
    
    반환값:
        완성된 인사 메시지
    """
    return f"{greeting}, {name}님!"

# 기본값 사용 - 일반적인 인사
message1 = greet("홍길동")  # greeting은 "안녕하세요" 사용
print(message1)  # 출력: 안녕하세요, 홍길동님!

# 다른 인사말 사용 - 특별한 경우
message2 = greet("이영희", "반갑습니다")  # greeting을 "반갑습니다"로 변경
print(message2)  # 출력: 반갑습니다, 이영희님!
````

**실행 결과:**
```
안녕하세요, 홍길동님!
반갑습니다, 이영희님!
```

**핵심 포인트:**
- 기본값은 숫자뿐만 아니라 문자열도 가능
- 대부분 "안녕하세요"를 사용하므로 기본값으로 설정
- 특별한 상황에만 다른 인사말 전달

### 💡 다국어 지원 예시

    greetings = {

## 4. 위치 인수와 기본값 조합

**실전에서 가장 많이 사용하는 패턴**입니다! 필수 정보는 위치 인수로, 선택 정보는 기본값으로 설정합니다.

````python
def create_user_profile(name, age, city="서울", hobby="독서"):
    """
    사용자 프로필 생성
    
    매개변수:
        name: 이름 (필수) - 기본값 없음
        age: 나이 (필수) - 기본값 없음
        city: 거주 도시 (선택, 기본값: "서울")
        hobby: 취미 (선택, 기본값: "독서")
    
    설계 철학:
        - 이름과 나이는 반드시 필요한 정보 → 필수 매개변수
        - 도시와 취미는 없어도 괜찮은 정보 → 기본값 제공
    """
    return f"{name}({age}세)는 {city}에 살며 {hobby}를 좋아합니다."

# 케이스 1: 필수 인수만 전달 (가장 간단)
profile1 = create_user_profile("박민수", 30)
print(profile1)  # 출력: 박민수(30세)는 서울에 살며 독서를 좋아합니다.

# 케이스 2: 일부 기본값 변경 (city만 변경)
profile2 = create_user_profile("최지은", 25, "부산")
print(profile2)  # 출력: 최지은(25세)는 부산에 살며 독서를 좋아합니다.

# 케이스 3: 모든 인수 전달 (완전히 커스터마이징)
profile3 = create_user_profile("정우성", 28, "인천", "운동")
print(profile3)  # 출력: 정우성(28세)는 인천에 살며 운동을 좋아합니다.
````

**실행 결과:**
```
박민수(30세)는 서울에 살며 독서를 좋아합니다.
최지은(25세)는 부산에 살며 독서를 좋아합니다.
정우성(28세)는 인천에 살며 운동을 좋아합니다.
```

**핵심 포인트:**
- **필수 매개변수** (`name`, `age`): 반드시 전달해야 함
- **선택 매개변수** (`city`, `hobby`): 생략 가능, 기본값 사용
- 유연성: 필요한 만큼만 인수를 전달할 수 있음

### 🎯 설계 원칙

함수를 설계할 때 이렇게 생각해보세요:

1. **반드시 필요한 정보는?** → 기본값 없는 필수 매개변수
2. **대부분 같은 값을 사용하는 정보는?** → 기본값 있는 선택 매개변수
3. **순서는?** → 필수 매개변수를 앞에, 선택 매개변수를 뒤에

## 5. 실전 예제: 할인 계산 함수

쇼핑몰에서 실제로 사용할 법한 복잡한 계산 함수입니다. 여러 개의 기본값을 활용하여 **유연한 할인 시스템**을 만들어봅시다!

````python
def calculate_discount(price, quantity, discount_rate=0.0, member_discount=0.0):
    """
    할인 적용 가격 계산
    
    매개변수:
        price: 상품 가격 (필수)
        quantity: 수량 (필수)
        discount_rate: 일반 할인율 (선택, 기본값: 0.0 = 할인 없음)
        member_discount: 회원 할인율 (선택, 기본값: 0.0 = 할인 없음)
    
    계산 방식:
        1. 소계 = 가격 × 수량
        2. 일반 할인 적용
        3. 회원 할인 추가 적용 (중복 할인 가능!)
    
    반환값:
        최종 가격 (정수)
    """
    # 1단계: 소계 계산
    subtotal = price * quantity
    print(f"소계: {subtotal}원")
    
    # 2단계: 일반 할인 적용 (예: 20% 할인 = 80% 지불)
    after_discount = subtotal * (1 - discount_rate)
    print(f"일반 할인 후: {int(after_discount)}원 (할인율: {discount_rate*100}%)")
    
    # 3단계: 회원 할인 추가 적용
    final_price = after_discount * (1 - member_discount)
    print(f"회원 할인 후: {int(final_price)}원 (추가 할인: {member_discount*100}%)")
    
    return int(final_price)

print("=== 시나리오 1: 할인 없이 구매 ===")
price1 = calculate_discount(10000, 3)
print(f"최종 가격: {price1}원\n")

print("=== 시나리오 2: 20% 할인만 적용 ===")
price2 = calculate_discount(10000, 3, 0.2)
print(f"최종 가격: {price2}원\n")

print("=== 시나리오 3: 20% 할인 + 10% 회원 할인 ===")
price3 = calculate_discount(10000, 3, 0.2, 0.1)
print(f"최종 가격: {price3}원")
````

**실행 결과:**
```
=== 시나리오 1: 할인 없이 구매 ===
소계: 30000원
일반 할인 후: 30000원 (할인율: 0.0%)
회원 할인 후: 30000원 (추가 할인: 0.0%)
최종 가격: 30000원

=== 시나리오 2: 20% 할인만 적용 ===
소계: 30000원
일반 할인 후: 24000원 (할인율: 20.0%)
회원 할인 후: 24000원 (추가 할인: 0.0%)
최종 가격: 24000원

=== 시나리오 3: 20% 할인 + 10% 회원 할인 ===
소계: 30000원
일반 할인 후: 24000원 (할인율: 20.0%)
회원 할인 후: 21600원 (추가 할인: 10.0%)
최종 가격: 21600원
```

**핵심 포인트:**
- 기본값 `0.0`으로 "할인 없음" 상태를 표현
- 필요한 할인만 선택적으로 적용 가능
- 단계별 계산 과정을 출력하여 디버깅 쉽게
