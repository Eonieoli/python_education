## 📖 왜 고급 타입 힌트가 필요할까요?

지난 시간(6교시)에는 기본 타입 힌트(`int`, `str`, `float`, `bool`)를 배웠습니다. 하지만 실무에서는 **훨씬 더 복잡한 데이터**를 다루게 됩니다.

### 실무에서 마주하는 복잡한 데이터

예를 들어, FastAPI로 사용자 목록을 반환하는 API를 만든다고 생각해봅시다:

```python
def get_users():
    return [
        {"id": 1, "name": "Alice", "email": "alice@example.com"},
        {"id": 2, "name": "Bob", "email": "bob@example.com"}
    ]
```

이 함수의 반환값을 어떻게 표현해야 할까요?
- 리스트입니다 → `List`
- 리스트 안에는 딕셔너리가 들어있습니다 → `Dict`
- 딕셔너리의 키는 문자열이고 값은... 뭘까요? 숫자도 있고 문자열도 있네요!

**기본 타입만으로는 이런 복잡한 구조를 표현할 수 없습니다!** 😱

### typing 모듈이 해결해줍니다! ✨

Python의 `typing` 모듈은 이런 복잡한 타입을 표현할 수 있는 도구들을 제공합니다:

- **List[int]**: 정수가 들어있는 리스트
- **Dict[str, str]**: 문자열 키와 문자열 값의 딕셔너리
- **Optional[str]**: 문자열 또는 None
- **Union[int, str]**: 정수 또는 문자열

### FastAPI에서 왜 중요한가요?

FastAPI는 **Pydantic**이라는 라이브러리를 사용하는데, Pydantic은 타입 힌트에 **완전히 의존**합니다.

```python
# FastAPI에서 이렇게 쓰게 됩니다!
from pydantic import BaseModel
from typing import List, Optional

class User(BaseModel):
    id: int
    name: str
    email: Optional[str] = None  # 이메일은 선택사항!

@app.get("/users")
def get_users() -> List[User]:  # 사용자 리스트를 반환!
    return [...]
```

타입 힌트가 없으면 FastAPI를 제대로 사용할 수 없습니다! 그래서 오늘 배우는 내용은 **FastAPI의 핵심**입니다. 🎯

## 1. typing 모듈 import (미리보기)

### import란 무엇인가요?

`import`는 **다른 모듈(파일)에서 기능을 가져오는** Python 문법입니다. 

- **모듈(Module)**: 재사용 가능한 코드를 모아놓은 파일
- **typing**: Python에 내장된 모듈로, 타입 힌트 도구를 제공

### 왜 미리보기일까요?

원래 `import`는 내일 배울 내용이지만, **타입 힌트를 쓰려면 반드시 필요**해서 오늘 먼저 맛만 보는 겁니다! 😊

### typing 모듈에서 무엇을 가져올까요?

오늘 배울 고급 타입 도구들:
- **List**: 리스트 안의 타입 지정
- **Dict**: 딕셔너리의 키와 값 타입 지정
- **Optional**: None을 포함할 수 있는 타입
- **Union**: 여러 타입 중 하나를 허용

### 사용 방법

```python
from typing import List, Dict, Optional, Union
```

이 한 줄만 쓰면 됩니다! 간단하죠? 😄

> ⚠️ **주의**: `typing`은 Python에 **내장되어 있어서** 별도로 설치할 필요가 없습니다. 그냥 바로 쓰면 됩니다!

````python
# typing 모듈에서 고급 타입 도구들을 가져옵니다
# (내일 import를 자세히 배우지만, 오늘은 일단 따라 쓰세요)
from typing import List, Dict, Optional, Union

# Python에 내장되어 있어서 별도 설치 없이 사용 가능합니다
print("typing 모듈을 가져왔습니다!")
print("이제 List, Dict, Optional, Union을 사용할 수 있어요! ✨")
````

**실행 결과:**
```
typing 모듈을 가져왔습니다!
이제 List, Dict, Optional, Union을 사용할 수 있어요! ✨
```

### 💡 이해했나요?

- `from typing import ...`는 "typing 모듈에서 ... 도구를 가져와라"는 뜻
- 한 번만 쓰면 파일 전체에서 사용 가능
- 보통 **파일 맨 위**에 작성합니다

## 2. List 타입 힌트

### List[타입]의 의미

`List[int]`는 "정수가 들어있는 리스트"를 의미합니다.

- `List[int]`: 정수 리스트 (예: `[1, 2, 3]`)
- `List[str]`: 문자열 리스트 (예: `["a", "b", "c"]`)
- `List[float]`: 실수 리스트 (예: `[1.5, 2.7, 3.9]`)

### 왜 List[int]라고 써야 하나요?

기본 타입만으로는 "리스트"라는 것만 알 수 있고, **안에 무엇이 들어있는지** 모릅니다:

```python
# ❌ 이렇게 쓰면 리스트라는 것만 알 수 있음
def process(data: list) -> list:
    pass

# ✅ 이렇게 쓰면 "정수 리스트"라는 것을 명확히 알 수 있음
def process(data: List[int]) -> List[str]:
    pass
```

### FastAPI에서의 활용

FastAPI에서 여러 데이터를 반환할 때 **항상** 리스트를 사용합니다:

```python
# 사용자 ID 리스트
@app.get("/user-ids")
def get_user_ids() -> List[int]:
    return [1, 2, 3, 4, 5]

# 이름 리스트
@app.get("/user-names")
def get_user_names() -> List[str]:
    return ["Alice", "Bob", "Charlie"]
```

타입 힌트가 있으면 FastAPI가 자동으로 **API 문서를 생성**해줍니다! 📚

### 예제 2-1: 정수 리스트 처리하기

````python
# List[int]: 정수 리스트를 받아서 합계를 계산합니다
def sum_list(numbers: List[int]) -> int:
    """숫자 리스트의 합계를 계산합니다"""
    return sum(numbers)  # sum()은 Python 내장 함수

# 테스트
result = sum_list([10, 20, 30])
print(f"합계: {result}")
````

**실행 결과:**
```
합계: 60
```

**코드 분석:**
- `numbers: List[int]`: "정수가 들어있는 리스트를 받습니다"
- `-> int`: "정수를 반환합니다"
- `sum(numbers)`: Python 내장 함수로 리스트의 모든 숫자를 더함

### 💡 타입 힌트가 도움을 주는 순간

VS Code나 PyCharm 같은 IDE에서 이 함수를 사용하면:
- `numbers`에 마우스를 올리면 "정수 리스트"라고 알려줍니다
- 실수로 문자열 리스트를 넣으면 경고를 띄워줍니다
- 자동 완성이 더 정확해집니다

### 예제 2-2: 문자열 리스트를 대문자로 변환하기

````python
# List[str]: 문자열 리스트를 받아서 모두 대문자로 변환합니다
def process_names(names: List[str]) -> List[str]:
    """이름 리스트를 모두 대문자로 변환합니다"""
    # 리스트 컴프리헨션으로 각 이름을 대문자로 변환
    return [name.upper() for name in names]

# 테스트
names = ["alice", "bob", "charlie"]
upper_names = process_names(names)
print(f"원본: {names}")
print(f"대문자 변환: {upper_names}")
````

**실행 결과:**
```
원본: ['alice', 'bob', 'charlie']
대문자 변환: ['ALICE', 'BOB', 'CHARLIE']
```

**코드 분석:**
- `names: List[str]`: "문자열 리스트를 받습니다"
- `-> List[str]`: "문자열 리스트를 반환합니다"
- 입력과 출력이 모두 `List[str]`이므로 같은 형태의 데이터가 변환됨을 알 수 있습니다

> 📌 **패턴 발견**: `List[타입]`을 받아서 `List[타입]`을 반환하는 함수는 "변환" 함수가 많습니다!

## 3. Dict 타입 힌트

### Dict[키타입, 값타입]의 의미

`Dict[str, int]`는 "문자열 키와 정수 값을 가진 딕셔너리"를 의미합니다.

- `Dict[str, str]`: 문자열 키, 문자열 값 (예: `{"name": "Alice"}`)
- `Dict[str, int]`: 문자열 키, 정수 값 (예: `{"age": 25}`)
- `Dict[int, str]`: 정수 키, 문자열 값 (예: `{1: "first"}`)

### 왜 Dict[str, int]라고 써야 하나요?

딕셔너리는 **키와 값 모두 타입이 있기 때문에** 두 가지를 모두 명시해야 합니다:

```python
# ❌ 이렇게 쓰면 키와 값의 타입을 모름
def get_user() -> dict:
    pass

# ✅ 이렇게 쓰면 "문자열 키, 문자열 값"이라는 것을 알 수 있음
def get_user() -> Dict[str, str]:
    pass
```

### FastAPI에서의 활용 (매우 중요! 🌟)

FastAPI에서 **JSON 데이터는 딕셔너리**로 변환됩니다:

```python
# API가 이런 JSON을 반환한다면
{
  "name": "Alice",
  "email": "alice@example.com"
}

# 함수는 이렇게 작성
@app.get("/user")
def get_user() -> Dict[str, str]:  # 문자열 키, 문자열 값
    return {"name": "Alice", "email": "alice@example.com"}
```

**JSON의 핵심**은 딕셔너리이므로, `Dict` 타입 힌트는 FastAPI에서 **가장 자주 사용**됩니다! 🎯

### 예제 3-1: Dict[str, str] - 사용자 정보 반환하기

````python
# Dict[str, str]: 문자열 키와 문자열 값의 딕셔너리를 반환합니다
def get_user_info() -> Dict[str, str]:
    """사용자 정보를 딕셔너리로 반환합니다"""
    # 모든 값이 문자열입니다
    return {
        "name": "Alice",
        "email": "alice@example.com"
    }

# 테스트
user = get_user_info()
print(f"사용자 정보: {user}")
print(f"이름: {user['name']}")
print(f"이메일: {user['email']}")
````

**실행 결과:**
```
사용자 정보: {'name': 'Alice', 'email': 'alice@example.com'}
이름: Alice
이메일: alice@example.com
```

**코드 분석:**
- `-> Dict[str, str]`: "문자열 키, 문자열 값을 반환합니다"
- `"name"`, `"email"`: 키는 모두 문자열
- `"Alice"`, `"alice@example.com"`: 값도 모두 문자열

> 💡 **FastAPI 패턴**: 사용자 정보, 상품 정보 등 데이터를 반환할 때 자주 사용하는 패턴입니다!

### 예제 3-2: Dict[str, int] - 단어 빈도수 세기

````python
# Dict[str, int]: 문자열 키와 정수 값의 딕셔너리를 반환합니다
def count_items(items: List[str]) -> Dict[str, int]:
    """각 항목의 개수를 세어 딕셔너리로 반환합니다"""
    # set()으로 중복 제거 후, count()로 각 항목의 개수를 셉니다
    return {item: items.count(item) for item in set(items)}

# 테스트
fruits = ["사과", "바나나", "사과", "오렌지", "바나나", "사과"]
counts = count_items(fruits)
print(f"과일 목록: {fruits}")
print(f"과일 개수: {counts}")
print(f"사과: {counts['사과']}개")
````

**실행 결과:**
```
과일 목록: ['사과', '바나나', '사과', '오렌지', '바나나', '사과']
과일 개수: {'사과': 3, '바나나': 2, '오렌지': 1}
사과: 3개
```

**코드 분석:**
- `items: List[str]`: "문자열 리스트를 받습니다"
- `-> Dict[str, int]`: "문자열 키, 정수 값을 반환합니다"
- 키는 항목 이름(문자열), 값은 개수(정수)

### 💡 실무 활용

이런 패턴은 **통계**, **집계**, **분석** 작업에서 자주 사용됩니다:
- 사용자별 게시글 수: `Dict[str, int]` (이름: 개수)
- 상품별 재고: `Dict[str, int]` (상품명: 재고)
- 태그별 사용 횟수: `Dict[str, int]` (태그: 횟수)

## 4. Optional 타입 (중요! 🌟)

### Optional[타입]의 의미

`Optional[str]`은 "문자열 **또는** None"을 의미합니다.

- `Optional[int]` = `int 또는 None`
- `Optional[str]` = `str 또는 None`
- `Optional[Dict[str, str]]` = `Dict[str, str] 또는 None`

### 왜 Optional이 필요한가요?

실무에서는 **"값이 없을 수도 있는 상황"**이 정말 많습니다:

```python
# 사용자를 찾는 함수
def find_user(user_id: int):
    # 찾으면 사용자 정보를 반환
    # 못 찾으면? None을 반환!
    pass
```

이럴 때 `Optional`을 사용하면 "None을 반환할 수 있다"는 것을 **명확히** 표현할 수 있습니다!

### FastAPI에서 왜 중요한가요? (매우 중요! 🎯)

FastAPI/Pydantic에서 `Optional`은 **"선택적 필드"**를 의미합니다:

```python
from pydantic import BaseModel

class User(BaseModel):
    id: int               # 필수 필드
    name: str             # 필수 필드
    email: Optional[str] = None  # 선택 필드 (없어도 됨!)

# 이렇게 사용 가능
user1 = User(id=1, name="Alice", email="alice@example.com")  # 이메일 있음
user2 = User(id=2, name="Bob")  # 이메일 없음 (None)
```

**회원가입**을 생각해보세요:
- 이름은 **필수**입니다 → `name: str`
- 전화번호는 **선택**입니다 → `phone: Optional[str] = None`

이처럼 `Optional`은 FastAPI에서 **가장 자주 사용하는 타입 중 하나**입니다! 😊

> ⚠️ **주의**: `Optional[str]`과 `str = None`은 **거의 같은 의미**입니다. 하지만 명확성을 위해 `Optional`을 쓰는 것을 권장합니다!

### 예제 4-1: Optional[Dict] - 사용자 검색하기

````python
# Optional[Dict[str, str]]: 딕셔너리 또는 None을 반환합니다
def get_user(user_id: int) -> Optional[Dict[str, str]]:
    """
    사용자 ID로 사용자를 찾습니다.
    찾으면 사용자 정보를, 못 찾으면 None을 반환합니다.
    """
    # 간단한 예시: user_id가 1이면 찾음, 아니면 None
    if user_id == 1:
        return {"name": "Alice", "email": "alice@example.com"}
    return None  # 찾지 못함!

# 사용자를 찾았을 때
found_user = get_user(1)
print(f"찾은 사용자: {found_user}")

# 사용자를 못 찾았을 때
not_found = get_user(999)
print(f"찾은 사용자: {not_found}")
````

**실행 결과:**
```
찾은 사용자: {'name': 'Alice', 'email': 'alice@example.com'}
찾은 사용자: None
```

**코드 분석:**
- `-> Optional[Dict[str, str]]`: "딕셔너리를 반환하거나, None을 반환할 수 있습니다"
- `if user_id == 1: return {...}`: 찾으면 딕셔너리 반환
- `return None`: 못 찾으면 None 반환

### 💡 실무에서 이렇게 사용합니다

```python
user = get_user(user_id)

if user is not None:  # 또는 if user:
    print(f"환영합니다, {user['name']}님!")
else:
    print("사용자를 찾을 수 없습니다.")
```

`Optional`을 쓰면 "None 체크를 해야 한다"는 것을 **자동으로 알려주므로** 버그를 예방할 수 있습니다!

### 예제 4-2: Optional[str] - 리스트에서 항목 찾기

````python
# Optional[str]: 문자열 또는 None을 반환합니다
def find_item(items: List[str], target: str) -> Optional[str]:
    """
    리스트에서 항목을 찾습니다.
    있으면 그 항목을, 없으면 None을 반환합니다.
    """
    # target이 items에 있으면 target을, 없으면 None을 반환
    return target if target in items else None

# 테스트
fruits = ["사과", "바나나", "오렌지"]

result1 = find_item(fruits, "바나나")
print(f"검색 결과 (바나나): {result1}")  # 있음 → "바나나" 반환

result2 = find_item(fruits, "포도")
print(f"검색 결과 (포도): {result2}")  # 없음 → None 반환
````

**실행 결과:**
```
검색 결과 (바나나): 바나나
검색 결과 (포도): None
```

**코드 분석:**
- `-> Optional[str]`: "문자열을 반환하거나, None을 반환할 수 있습니다"
- `target if target in items else None`: 있으면 target, 없으면 None
- 이것은 **삼항 연산자** 패턴입니다 (Day 3에서 배웠죠!)

### 🔍 Optional의 핵심 패턴

```python
# 검색/조회 함수는 보통 Optional을 사용합니다
def find_xxx(...) -> Optional[타입]:
    if 찾았다:
        return 결과
    return None  # 못 찾음
```

이 패턴은 실무에서 **정말 자주** 사용하므로 꼭 익혀두세요! 📝

## 5. Union 타입 (간단히)

### Union[타입1, 타입2]의 의미

`Union[int, str]`은 "정수 **또는** 문자열"을 의미합니다.

- `Union[int, str]` = `int 또는 str`
- `Union[int, float]` = `int 또는 float` (숫자 전체)
- `Union[str, None]` = `str 또는 None` (이건 `Optional[str]`과 같음!)

### Optional과의 차이점

- `Optional[str]` = `Union[str, None]` (같은 의미!)
- `Union[int, str]` = int 또는 str (None은 포함 안 됨)

`Optional`은 "None 포함" 전용이고, `Union`은 "여러 타입 중 하나" 일반적인 경우에 사용합니다.

### 언제 사용하나요?

실무에서는 `Union`을 **많이 사용하지 않습니다**. 왜냐하면:
- 타입이 여러 개면 코드가 복잡해집니다
- 보통 한 가지 타입으로 통일하는 것이 좋습니다

하지만 **FastAPI에서 가끔** 유용합니다:

```python
# 숫자나 문자열 모두 받을 수 있는 API
@app.get("/item/{item_id}")
def get_item(item_id: Union[int, str]):
    # item_id가 123이어도 되고 "abc"여도 됨
    pass
```

> 💡 **권장**: 처음에는 `Union`보다 `Optional`에 집중하세요! `Union`은 필요할 때 찾아서 쓰면 됩니다.

### 예제 5-1: Union[int, str] 사용하기

````python
# Union[int, str]: int 또는 str 타입을 허용합니다
def process(value: Union[int, str]) -> str:
    """숫자나 문자열을 받아서 문자열로 변환합니다"""
    return str(value)  # 어떤 타입이든 문자열로 변환

# 테스트
result1 = process(123)  # 정수 전달
print(f"정수 변환: {result1} (타입: {type(result1)})")

result2 = process("hello")  # 문자열 전달
print(f"문자열 변환: {result2} (타입: {type(result2)})")
````

**실행 결과:**
```
정수 변환: 123 (타입: <class 'str'>)
문자열 변환: hello (타입: <class 'str'>)
```

**코드 분석:**
- `value: Union[int, str]`: "정수 또는 문자열을 받습니다"
- `-> str`: "항상 문자열을 반환합니다"
- `str(value)`: 어떤 값이든 문자열로 변환

### ⚠️ Union 사용 시 주의사항

```python
def calculate(value: Union[int, str]):
    # 이렇게 하면 안 됩니다! 타입에 따라 다른 동작을 하면 복잡해집니다
    if isinstance(value, int):
        return value * 2
    else:
        return value.upper()
```

가능하면 **한 가지 타입**으로 통일하는 것이 좋습니다!

## 6. *args, **kwargs에 타입 힌트

### 가변 인자에도 타입을 지정할 수 있습니다!

Day 4-5교시에서 `*args`, `**kwargs`를 배웠습니다. 이것들에도 타입 힌트를 줄 수 있습니다!

- `*args: int`: args의 각 요소가 int
- `**kwargs: str`: kwargs의 각 값이 str

### 주의사항

- `*args: int`는 "정수 여러 개"를 의미합니다 (List[int]가 아님!)
- `**kwargs: str`는 "문자열 값 여러 개"를 의미합니다 (Dict[str, str]가 아님!)

### 왜 유용한가요?

FastAPI에서는 **가변 인자를 자주 사용**하지 않지만, Python 함수를 작성할 때 타입 힌트를 주면:
- IDE가 더 정확한 자동 완성을 제공합니다
- 버그를 미리 찾을 수 있습니다
- 코드가 더 명확해집니다

### 예제 6-1: *args에 타입 힌트

````python
# *args: int - 각 인자가 int 타입입니다
def add_all(*args: int) -> int:
    """모든 숫자를 더합니다"""
    return sum(args)  # args는 튜플이지만, 각 요소는 int

# 테스트
total = add_all(1, 2, 3, 4, 5)
print(f"1+2+3+4+5 = {total}")

total2 = add_all(10, 20)
print(f"10+20 = {total2}")
````

**실행 결과:**
```
1+2+3+4+5 = 15
10+20 = 30
```

**코드 분석:**
- `*args: int`: "정수를 여러 개 받습니다"
- `-> int`: "정수를 반환합니다"
- `sum(args)`: 모든 정수를 더함

### 💡 타입 힌트의 효과

```python
add_all(1, 2, 3)      # ✅ OK - 모두 int
add_all(1, "2", 3)    # ❌ IDE가 경고! - "2"는 str
```

### 예제 6-2: **kwargs에 타입 힌트

````python
# **kwargs: str - 각 값이 str 타입입니다
def create_user(**kwargs: str) -> Dict[str, str]:
    """사용자 정보를 딕셔너리로 생성합니다"""
    return kwargs  # kwargs는 이미 딕셔너리

# 테스트
new_user = create_user(name="Bob", email="bob@example.com", city="서울")
print(f"새 사용자: {new_user}")
print(f"이름: {new_user['name']}")
print(f"도시: {new_user['city']}")
````

**실행 결과:**
```
새 사용자: {'name': 'Bob', 'email': 'bob@example.com', 'city': '서울'}
이름: Bob
도시: 서울
```

**코드 분석:**
- `**kwargs: str`: "문자열 값을 여러 개 받습니다"
- `-> Dict[str, str]`: "문자열 키, 문자열 값의 딕셔너리를 반환합니다"
- `return kwargs`: kwargs 자체가 딕셔너리이므로 그대로 반환

### 🔍 실무 패턴

```python
# 유연한 데이터 생성 함수
def create_data(**kwargs: Union[str, int]):
    # 문자열이나 정수 값을 자유롭게 받을 수 있음
    return kwargs

data = create_data(name="Alice", age=25, city="Seoul")
```

## 📌 핵심 정리

이번 시간에 배운 내용:

### 1. typing 모듈 import
```python
from typing import List, Dict, Optional, Union
```
- Python 내장 모듈이므로 설치 불필요
- 파일 맨 위에 한 번만 작성
- 내일 import를 자세히 배웁니다!

### 2. List 타입 힌트
```python
def process(data: List[int]) -> List[str]:
    pass
```
- `List[int]`: 정수 리스트
- `List[str]`: 문자열 리스트
- 리스트 안에 무엇이 들어있는지 명확히 표현

### 3. Dict 타입 힌트 (FastAPI 핵심! 🌟)
```python
def get_user() -> Dict[str, str]:
    return {"name": "Alice", "email": "alice@example.com"}
```
- `Dict[str, str]`: 문자열 키, 문자열 값
- `Dict[str, int]`: 문자열 키, 정수 값
- JSON 데이터 = 딕셔너리

### 4. Optional 타입 (매우 중요! 🎯)
```python
def find_user(user_id: int) -> Optional[Dict[str, str]]:
    if found:
        return user_data
    return None  # 못 찾으면 None!
```
- `Optional[str]` = `str 또는 None`
- "값이 없을 수도 있다"를 명확히 표현
- FastAPI/Pydantic에서 선택적 필드에 사용

### 5. Union 타입 (간단히)
```python
def process(value: Union[int, str]) -> str:
    return str(value)
```
- `Union[int, str]` = `int 또는 str`
- 여러 타입 중 하나를 허용
- 실무에서는 많이 사용 안 함

### 6. *args, **kwargs 타입 힌트
```python
def add_all(*args: int) -> int:
    return sum(args)

def create_user(**kwargs: str) -> Dict[str, str]:
    return kwargs
```
- `*args: int`: 각 인자가 int
- `**kwargs: str`: 각 값이 str
