### Q3. *args와 **kwargs

```python
def create_user(name: str, *args: str, **kwargs: str) -> None:
    print(name)      # "Alice"
    print(args)      # ???
    print(kwargs)    # ???

create_user("Alice", "engineer", "senior", city="Seoul", team="Backend")
```

`args`와 `kwargs`에는 각각 무엇이 들어있을까요?

<details>
<summary>정답 보기</summary>

**정답:**
- `args = ("engineer", "senior")` → 튜플
- `kwargs = {"city": "Seoul", "team": "Backend"}` → 딕셔너리

`*args`는 이름 없이 전달된 추가 인수들을 튜플로 받고,  
`**kwargs`는 키워드로 전달된 추가 인수들을 딕셔너리로 받습니다!
</details>

```python
def find_user_by_email(

    email: str
) -> Optional[Dict[str, str]]:
    for user in users:
        if user["email"] == email:
            return user
    return None  # 못 찾으면 None 반환
```

`Optional[T]`는 T 또는 None을 반환할 수 있다는 의미입니다!

## 📌 종합 정리

### 오늘 Day 4에서 배운 핵심 내용

1. **함수 기본**
   - `def` 키워드로 함수 정의
   - `return`으로 값 반환
   - 여러 값 반환 → 튜플로 자동 변환

2. **매개변수 종류**
   - 위치 인수: `func(a, b)`
   - 키워드 인수: `func(a=1, b=2)`
   - 기본값: `def func(a, b=10)`
   - None 처리: `def func(name, bio=None)`

3. **가변 인수**
   - `*args`: 여러 개의 위치 인수 → 튜플
   - `**kwargs`: 여러 개의 키워드 인수 → 딕셔너리
   - 언패킹: `*list`, `**dict`

4. **타입 힌트 기본**
   - 매개변수: `def func(a: int, b: str)`
   - 반환값: `def func() -> int:`
   - 기본 타입: `int`, `str`, `float`, `bool`

5. **타입 힌트 고급**
   - `List[int]`: 정수 리스트
   - `Dict[str, str]`: 문자열 딕셔너리
   - `Optional[str]`: 문자열 또는 None
   - `Union[int, str]`: int 또는 str
